## Aufgabe 1

Erstelle eine SQLite-Datenbank mit einer Tabelle namens schueler. Die Tabelle soll die folgenden Spalten enthalten:

- name (Text)
- vorname (Text)
- gebdatum (Datum)
- hobby (Text)
- lieblingsfach (Text)

Füge dann mindestens fünf Datensätze in die Tabelle ein.

Zur Vereinfachung der Aufgabe finden Sie in der nächsten Zelle ein paar Daten. Der SQL-Befehl müsste natürlich drum herum gebaut werden.

```python
import sqlite3
from pathlib import Path

db_path = Path(r"C:/Users/Administrator/Desktop/KursMaterial/04-IEDSCDB/Teil 2 - SQLite/telefonbuch.db")

print("CWD:", Path.cwd())
print("DB path:", db_path.resolve())
print("DB exists:", db_path.exists(), "size:", db_path.stat().st_size if db_path.exists() else None)

conn = sqlite3.connect(db_path)  # Verbindung wird hergestellt
c = conn.cursor()  # Ein Cursor wird erstellt (Ermöglicht den Zugriff auf die Datenbank)

# 1) Asegura tabla (si ya existe, no pasa nada)
c.execute("""
CREATE TABLE IF NOT EXISTS estudiantes (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    vorname TEXT,
    geb_datum TEXT,
    hobby TEXT,
    lieblingsfach TEXT,
    UNIQUE(name, vorname, geb_datum)
)
""")

# 2) Datos (fecha siempre YYYY-MM-DD)
daten = [
    ('Musterfrau', 'Lena', '2001-10-28', 'Musikverein', 'Mathe'),
    ('Mustermann', 'Max', '2002-01-01', 'Sportverein', 'Deutsch'),
    ('Mustermeier', 'Anna', '2003-02-02', 'Lesen', 'Englisch'),
    ('Mustersohn', 'Jan', '2004-03-03', 'Spielen', 'Physik'),
    ('Mustertochter', 'Lisa', '2005-04-04', 'Basteln', 'Chemie'),
    ('Müller', 'Hans', '1970-01-01', 'Bier trinken', 'einFach'),

]

sql = """
      INSERT INTO estudiantes (name, vorname, geb_datum, hobby, lieblingsfach) VALUES (?, ?, ?, ?, ?)
      """

# Der SQL-Befehl wird ausgeführt
c.executemany(sql, daten)
conn.commit()

# 3) Verificación visible ------ EXTRA
c.execute("SELECT id, name, vorname, geb_datum, hobby, lieblingsfach FROM estudiantes")
for row in c.fetchall():
    print(row)

conn.close() # Verbindung wird geschlossen
```

## Aufgabe 2

Schreibe ein Python-Programm, das alle Datensätze aus der Tabelle schuelern ausgibt, die das Hobby "Musikverein" haben.

```python
conn = sqlite3.connect(db_path)
c = conn.cursor()

sql = """
SELECT name, vorname, geb_datum, hobby, lieblingsfach
FROM estudiantes
WHERE hobby = ?
"""

c.execute(sql, ("Musikverein",))

rows = c.fetchall()

for row in rows:
    print(row)

conn.close()
```

## Aufgabe 3

Ändere den Lieblingsfach-Eintrag des Datensatzes mit der ID 1 von "Mathe" in "Informatik".

```python
conn = sqlite3.connect(db_path)
c = conn.cursor()

sql = """
UPDATE estudiantes
SET lieblingsfach = ?
WHERE id = ?
"""

c.execute(sql, ("Informatik", 1))
conn.commit()

 # Verificacion
c.execute("SELECT id, name, vorname, lieblingsfach FROM estudiantes WHERE id = 1")
print(c.fetchone())

conn.close()
```

## Aufgabe 4

Lösche den Datensatz mit der ID 2.

```python
conn = sqlite3.connect(db_path)
c = conn.cursor()

sql = """
DELETE FROM estudiantes
WHERE id = ?
"""

c.execute(sql, (2,))
conn.commit()

 # Verificacion
c.execute("SELECT * FROM estudiantes")
print(c.fetchall())

conn.close()
```

## Aufgabe 5

Schreibe ein Python-Programm, das alle Datensätze aus der Tabelle schueler in alphabetischer Reihenfolge nach dem Nachnamen ausgibt.

```python
conn = sqlite3.connect(db_path)
c = conn.cursor()

sql = """
SELECT * FROM estudiantes
ORDER BY name
"""

c.execute(sql)

rows = c.fetchall()
for row in rows:
    print(row)

conn.close()
```
