### NoSQL am Beispiel von Mongo-DB
A diferencia de las bases de datos SQL como SQLite, MySQL o MariaDB, MongoDB **NO es una base de datos relacional**. MongoDeb es una banco de datos basada en documentos con un lenguaje de consulta no estructurado**.

MongoDB es una popular base de datos NoSQL conocida por su flexibilidad y escalabilidad. En combinación con Python, uno de los lenguajes de programación más versátiles, ofrece una plataforma potente para gestionar y procesar datos.

### Wichtig zu beachten

MongoDB almacena datos en un formato llamado **BSON**, que es una versión binaria de JSON. MongoBB es **esquemática**, lo que significa que los documentos en la misma colección pueden tener campos diferentes. La interacción con MongoDEB a través de Python es principalmente mediante el trabajo con **Dictionaries** y **Cursors**.

#### Desmontar ObjectID y extraer Timestamp

La marca de tiempo se emite en UTC. Esto debe ser tenido en cuenta en la búsqueda.

```python
from pymongo import MongoClient
from bson import ObjectId

client = MongoClient("mongodb://localhost:27017/")
db = client["unsere_tool"]

collection = db["benutzer"]

daten = collection.find_one({"name": "Tony Tester"})

obj_id = ObjectId(daten["_id"])
print(obj_id)

print(obj_id.generation_time)
```

#### Suchen nach Erstellungsdatum

```python
from pymongo import MongoClient
from bson import ObjectId
from datetime import datetime

client = MongoClient("mongodb://localhost:27017/")
db = client["unsere_tool"]

collection = db["benutzer"]

daten = collection.find_one({"name": "Tony Tester"})

obj_id = ObjectId(daten["_id"])

# Startdatum:
start_date = datetime(2026, 1,26)

# umschlüsseln
start_object = ObjectId.from_datetime(start_date)

documents = collection.find({"_id": {"$gte": start_object}})

for doc in documents:
    print(doc)
```

Cambiar la hora en UTC.

```python
from pymongo import MongoClient
from bson import ObjectId
from datetime import datetime
import pytz


client = MongoClient("mongodb://localhost:27017/")
db = client["unsere_tool"]

collection = db["benutzer"]

berlin = pytz.timezone('Europe/Berlin')  # Deutsche Zeitzone einstellen


# Startdatum und Zeit:
start_date = berlin.localize(datetime(2026, 1,27, 8, 0, 0))
end_date = berlin.localize(datetime(2026, 1,27, 18, 0, 0))


# Umgewandelte Zeitstempel für die Suche in UTC
start_utc = start_date.astimezone(pytz.utc)
end_utc = end_date.astimezone(pytz.utc)


daten = collection.find({"_id": {
    "$gte": ObjectId.from_datetime(start_utc),
    "$lte": ObjectId.from_datetime(end_utc)
}})

for doc in daten:
    print(doc)
```

### Otras posibilidades de filtrado y comparación:

```python
collection.find({"alter": {"$eq": 25}})  # gleich
collection.find({"alter": {"$ne": 25}})  # nicht gleich
collection.find({"alter": {"$gt": 25}})  # größer als
collection.find({"alter": {"$gte": 25}})  # größer oder gleich
collection.find({"alter": {"$lt": 25}})  # kleiner als
collection.find({"alter": {"$lte": 25}})  # kleiner oder gleich
collection.find({"alter": {"$in": [20, 25, 30]}})  # in einer Liste enthalten
collection.find({"alter": {"$nin": [20, 25, 30]}})  # nicht in einer Liste enthalten
```

#### Ein kleines Skript zum Suchen eines Eintrages (gemeinsame Übung)

```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017/")
db = client["unsere_tool"]

collection = db["benutzer"]

nombre = input("escribe el nombre completo a buscar")

# busca si "nombre" esta de alguna forma dentro del str.
daten = collection.find({"name": {"$regex": f".*{nombre}.*"}})
print(daten)

for i in daten:
    for k, v in i.items():
        print(f"{k}: {v}")
```

.* bedeutet: beliebige Zeichenfolge beliebiger Länge <br>
(. = ein beliebiges Zeichen, * = beliebig oft).

#### Indizierung:
Al igual que con las bases de datos relacionales, la indexación en MongoDB es crucial para el rendimiento de las consultas. Puede crear índices en colecciones para mejorar la velocidad de la consulta.

```python
collection.create_index([("name", pymongo.ASCENDING)])
```

Cada colección de MongoDB puede tener varios índices que mejoran el rendimiento de la consulta. Puede utilizar el método **index_information()** para obtener detalles de los índices de una colección específica.

He aquí un ejemplo de cómo obtener información sobre los índices de una colección en MongoDB con Python y pymongo:

```python
indexes = collection.index_information()

# Informationen zu jedem Index ausgeben
for index in indexes:
    print(indexes[index])
```

- **index_information()** devuelve un diccionario en el que las claves son los nombres de los índices y los valores contienen detalles de cada índice.
- La salida contiene información como el nombre del índice, los campos indicados y si el índice está en orden ascendente (1) o descendente (-1).
- Por defecto, cada colección de MongoDB tiene un índice en el campo _id.

Este método es útil para entender cómo sus datos están organizados e indicados en MongoDB, lo que a su vez influye en el rendimiento de sus consultas.
