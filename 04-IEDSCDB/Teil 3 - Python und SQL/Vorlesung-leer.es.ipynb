# Breve introducción a SQL

SQL (Structured Query Language) es un lenguaje normalizado para la gestión y consulta de datos en bases de datos relacionales. Se pueden definir estructuras de datos (Data Definition Languaje, DDL), manipular datos (Data Manipulation Laguage, DML) y controlar los derechos de acceso y permisos (DATA Control Languace, DCL). Los conceptos básicos de SQl se basan en tablas en las que los datos se organizan en filas y columnas. Las relaciones entre tablas son creadas por claves (Keys).

## Wichtige Datentypen in SQL

### Numerische Datentypen
- **TINYINT**: 1 Byte, Wertebereich -128 bis 127 oder 0 bis 255 (unsigned).
- **SMALLINT**: 2 Bytes, Wertebereich -32.768 bis 32.767.
- **INT** (INTEGER): 4 Bytes, Wertebereich -2.147.483.648 bis 2.147.483.647.
- **BIGINT**: 8 Bytes, Wertebereich -9.223.372.036.854.775.808 bis 9.223.372.036.854.775.807.
- **FLOAT**(p): Número de coma flotante con precisión simple (tipo 4 bytes, aproximadamente 7 decimales de precisión).
- **DOUBLE**: Número de coma flotante con doble precisión (8 bytes, aproximadamente 15 decimales de precisión).
  - **Nota**: FLOAT y DOUBLE utilizan representaciones binarias de coma flotante, lo que puede dar lugar a errores de redondeo en números decimales.

- **DECIMAL**(p, s): Tipo de datos decimales exacto, ideal para cantidades de dinero y cálculos precisos.
  - **Precision (p)**: Número total de dígitos significativos (valor máximo, por ejemplo, DECIMAL [10,2) → a 99999999.99).
  - **Scale (s)**: número de decimales (por ejemplo, 2 en DECIMAL [10,2) ].
  - **Promoción**: Guarda valores como fracción decimal, sin errores de redondeo binario, resultados de cálculo exactos garantizados.

### Zeichenketten-Datentypen
- **CHAR**(n): Longitud fija n, llena de blanco, ideal para códigos o longitudes fijas de texto.
- **VARCHAR**(n): Longitud variable a n, sólo almacena longitud real de caracteres + 1 *2 bytes Overhead.
- **TEXT**-Typen: TINYTEXT (bis 255 Bytes), TEXT (bis 65 535 Bytes), MEDIUMTEXT (bis 16 MiB), LONGTEXT (bis 4 GiB).

### Tipos de fecha y hora
- **DATE**: YYYY-MM-DD, 3 Bytes (z. B. '2025-04-24').
- **TIME**: HH:MM:SS, 3 bytes, ideal para horas del día sin fecha.
- **DATETIME**: YYY-MM-DD HH:MM:SS, 8 bytes, guarda la fecha y la hora.
- **TIMESTAMP**: 4 bytes, guarda marcas de tiempo desde '1970-01-01 00:00:01' UTC, puede tener en cuenta zona horaria.
- **YEAR**: 1 byte, por ejemplo, valor anual entre 1901 y 2155.

### Boolescher Datentyp
- **BOOLEAN**: In vielen Systemen als TINYINT(1) realisiert, 0 = FALSE, 1 = TRUE.

### Tipos binarios de datos
- **BLOB** Tipos: TINYBLOb (hasta 255 bytes), BLOB (hasta 65 535 bytes) y MEDIUMBLLOB ( hasta 16 miB), LOGBLOB (4 GiB), para archivos binarios como imágenes o documentos.

## Restricciones importantes (Constraints) en SQL

- **PRIMARY KEY**
Un identificador único para cada línea de una tabla.
  - Garantiza que cada fila se identifica claramente (p0a).
  - Erlaubt keine `NULL`-Werte (`NOT NULL`).
  - A menudo se utiliza automáticamente como índice y, por ejemplo, se puede combinar con el factor de conversión P0 para generar ID continuos.

- **UNIQUE**
Asegura que todos los valores de una o más columnas dentro de la tabla son únicos.
  - Puede aplicarse a columnas individuales o múltiples (contraint Unique de varias columnas).
  - Útil para evitar duplicaciones (por ejemplo, direcciones de correo electrónico únicas, nombres de usuario).

- **FOREIGN KEY**
Define una columna (o varias) que se refiere al "P0" de otro cuadro.
  - Asegura la integridad de referencia entre tablas.
  - Permite realizar acciones en caso de supresión o actualización de los registros de referencia (por ejemplo , " P0 " y " P1 ").

- **NOT NULL**
Evita que una columna acepte valores de "P0"
  - Stellt sicher, dass verpflichtende Felder immer einen Wert enthalten (z. B. Geburtsdatum, Vorname).

- **CHECK**
Define una condición personalizada que cada valor registrado debe cumplir.
  - Por ejemplo, el valor de P0 solo permite valores para adultos.
  - Ayuda para validar la lógica empresarial directamente a nivel de datos.

- **DEFAULT**
Define un valor predeterminado para una columna si no se especifica ningún valor al insertarlo.
  - Ejemplo: P0 para las columnas de la marca temporal.
  - Ahorra la repetición de los valores utilizados con frecuencia y garantiza la coherencia.

### Erstellen einer neuen Tabelle
Mithilfe des Befehls "**CREATE TABLE**" wird eine neue Tabelle erstellt:

```python
import mysql.connector as mc

conn = None

# Verbindung herstellen:
try:
    conn = mc.connect(host = "localhost",
                      user = "root",
                      password = "12345",
                      database = "telefonbuch",
                      use_pure = True)

except mc.Error as e:
    print("Error connecting to MySQL Server:\n", e)


# Prüfen ob (noch) eine Verbindung besteht:
if conn.is_connected():
    print("Connection established successfully.")
else:
    print("Connection failed.")

# Cursor erstellen:
c = conn.cursor()

# Neue Tabelle erstellen:

sql = """CREATE TABLE IF NOT EXISTS eintraege (
      id INT(6) UNSIGNED AUTO_INCREMENT PRIMARY KEY,
      vorname VARCHAR(30) NOT NULL,
      nachname VARCHAR(30) NOT NULL,
      vorwahl VARCHAR (10) NOT NULL,
      telefonnummer VARCHAR (30) NOT NULL
      )"""

c.execute(sql)
conn.close()
```

### Añadir valores
Ahora agreguemos una entrada:

```python
import mysql.connector as mc

conn = None

# Verbindung herstellen:
try:
    conn = mc.connect(host = "localhost",
                      user = "root",
                      password = "12345",
                      database = "telefonbuch",
                      use_pure = True)

except mc.Error as e:
    print("Error connecting to MySQL Server:\n", e)

# Cursor erstellen:
c = conn.cursor()

var_vorname = input("Vorname: ")
var_nachname = input("Nachname: ")
var_vorwahl = input("Vorwahl: ")
var_telefonnummer = input("Telefonnummer: ")

params = (var_vorname, var_nachname, var_vorwahl, var_telefonnummer)
sql = """INSERT INTO eintraege (vorname, nachname, vorwahl, telefonnummer) VALUES (%s, %s, %s, %s)"""

try:
    c.execute(sql, params)
    conn.commit()

except Exception as e:
    print(e)

finally:
    # Prüfen ob (noch) eine Verbindung besteht:
    if conn and conn.is_connected():
        print("Verbindung zur Datenbank wird getrennt.")
        conn.close()
```

En la biblioteca **mysql-connector-python de Python**, **%s** se utiliza como marcador de posición en las declaraciones SQL. Se utiliza para insertar valores en el estado SQ L sin aumentar el riesgo de una inyección SQl. El comodín %s se utiliza independientemente del tipo de datos del valor que se debe insertar. Esto significa que **% se utiliza tanto para cadenas, números, fechas como otros tipos de datos**.

Es importante tener en cuenta que el marcador de posición %s en este contexto no tiene nada que ver con el operador de formato de cadena en Python. En el contexto del conector MySQL, %s es simplemente un marcador genérico para parámetros.

Si prepara una declaración SQL y utiliza %s como marcador de posición, debe pasar los valores reales como tupel (o lista) al método execute(). El conector MySQL garantiza que estos valores se incluyan de forma segura en la declaración S QL para evitar la inyección SQl.

En este contexto, no hay ningún otro tipo de comodines como %d o %f que se conozca desde el formato normal de cadena en Python. %s se utiliza universalmente para todos los tipos de datos.

#### Añadir varias entradas
En lugar de **c.execute()** se utiliza **c, executemany() **. No es necesario pasar parámetros:

```python
import mysql.connector as mc

conn = None

# Verbindung herstellen:
try:
    conn = mc.connect(host = "localhost",
                      user = "root",
                      password = "12345",
                      database = "telefonbuch",
                      use_pure = True)

except mc.Error as e:
    print("Error connecting to MySQL Server:\n", e)

# Cursor erstellen:
c = conn.cursor()

neue_eintraege = [
        ("Max", "Mustermann", "0123", "456789"),
        ("Erika", "Musterfrau", "098", "7654321"),
        ("Hans", "Meier", "0171", "1234567"),
    ]


sql = """INSERT INTO eintraege (vorname, nachname, vorwahl, telefonnummer) VALUES (%s, %s, %s, %s)"""


try:
    c.executemany(sql, neue_eintraege)
    conn.commit()

except Exception as e:
    print(e)

finally:
    # Prüfen ob (noch) eine Verbindung besteht:
    if conn and conn.is_connected():
        print("Verbindung zur Datenbank wird getrennt.")
        conn.close()
```

## Añadir otra columna de la tabla

Los planes cambian. Tal vez tenga que añadir una columna a su tabla ya creada.

Para ampliar una tabla existente en una base de datos MySQL a otra columna, utilice el estado SQL ALTER TABLE. En Python, puede ejecutar esta declaración con un conector MySqL.

Así es como lo haces.

```python
import mysql.connector as mc

def create_connection():
    return mc.connect(
        host = "localhost",
        user = "root",
        password = "FCbarcelona2025 !",
        database = "Warenkataloge",
        use_pure = True,
    )

def create_stock_col():
    conn = create_connection()
    cursor = conn.cursor()

    try:
        cursor.execute("SELECT DATABASE()")
        print("DB actual:", cursor.fetchone()[0])

        cursor.execute("""
            ALTER TABLE inventario
            ADD COLUMN ort VARCHAR(100) NOT NULL DEFAULT 'Leipzig'
        """)
        conn.commit()
        print("🆕 Columna ort creada")

    except mc.Error as e:
        if e.errno == 1060:
            print("ℹ️ columna ya existía")
        else:
            raise

    finally:
        cursor.execute("SHOW COLUMNS FROM inventario")
        print("Columnas:", [r[0] for r in cursor.fetchall()])
        cursor.close()
        conn.close()

create_stock_col()
```

## Wildcards

## Datapipelines

#### Eine Sache noch zum Schluss

Los oleoductos de datos en el contexto de SQL y las bases de datos suelen referirse a procesos **automatizados** utilizados para la transmisión y transformación sistemáticas de los datos de una fuente a un solo objetivo. Estos oleoductos desempeñan un papel crucial en el procesamiento de datos y la inteligencia empresarial. Permiten mover, procesar y utilizar grandes cantidades de datos de forma eficiente y fiable para fines de análisis.

Hier sind einige Kernaspekte von Data Pipelines:

- **Extracción de datos**: Este es el primer paso en el que los datos se extraen de una o más fuentes. Estas fuentes pueden ser bases de datos, archivos, API, servicios externos y otros.

- **Procesamiento/transformación de datos**: Después de la extracción, los datos se transforman a menudo. Estas transformaciones pueden incluir la limpieza de datos, la aplicación de lógica empresarial, la agregación, la clasificación, las joinas con otros registros, etc. El objetivo es introducir los datos en un formato adecuado para el análisis o el siguiente paso en la tubería.

- ** Almacenamiento de datos (Loading)**: En el último paso, los datos procesados se almacenan en un sistema de destino que puede ser un almacén de datos, una base de datos u otro sistema de almacenamiento. A menudo se llama proceso ETL (Extraction, Transformation, Loading).

- **Automatización**: Los oleoductos de datos a menudo se automatizan para funcionar de forma continua o programada. Esto es especialmente importante para garantizar la actualidad de los datos en entornos empresariales de vida rápida.

- ** Supervisión y mantenimiento**: Dado que los oleoductos de datos pueden apoyar procesos empresariales críticos, su seguimiento y mantenimiento periódicos son importantes para evitar problemas como pérdida de datos, retrasos o errores en el procesamiento de datos.

- **SQL en los oleoductos de datos**: SQL desempeña un papel crucial en muchas fases de un gasoducto de datos, especialmente en la extracción de datos de bases de datos relacionales y en su transformación. SQl se utiliza a menudo en estos escenarios debido a su eficiencia y eficacia en la manipulación y consulta de datos.

Los oleoductos de datos son un componente esencial de soluciones de mantenimiento de datos, sistemas de inteligencia empresarial, implementaciones de lago de datos y otras aplicaciones intensivas en datos basadas en el análisis y la notificación de datos.
