## Kurzeinführung in MySQL

In diesem Notebook lernst du, wie man eine View (Sicht) in MySQL erstellt,
verwendet und wieder löscht. Eine View ist eine gespeicherte SQL-Abfrage,
die wie eine virtuelle Tabelle funktioniert.

#### Eine Datenbank erstellen:
sql:

```python
CREATE DATABASE IF NOT EXISTS firma;
USE firma;
```

```python
CREATE TABLE mitarbeiter (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(100),
    abteilung VARCHAR(100),
    gehalt DECIMAL(10,2)
);
```

#### Ein paar Testdaten erstellen
sql:

```python

INSERT INTO mitarbeiter (name, abteilung, gehalt) VALUES
('Anna Müller', 'Vertrieb', 4500),
('Bernd Schulz', 'IT', 5200),
('Claudia Meier', 'Vertrieb', 4600),
('Dieter Huber', 'IT', 5400),
('Eva Schneider', 'Personal', 4100);
```

#### Eine erste Abfrage erstellen

```python
SELECT * FROM mitarbeiter;
```

#### Alle Mitarbeiter aus der IT-Abteilung

```python
SELECT * FROM mitarbeiter WHERE abteilung = "IT";
```

### Was ist eine View?

Eine View ist eine gespeicherte Abfrage, die wie eine virtuelle Tabelle behandelt wird.
Sie eignet sich besonders, um Daten gezielt gefiltert oder reduziert bereitzustellen.

```python
CREATE VIEW it_mitarbeiter AS
SELECT id, name, gehalt
FROM mitarbeiter
WHERE abteilung = 'IT';
```

```python
%%sql
SELECT * FROM it_mitarbeiter;
```

### Beispiel: View ohne Gehalt zur Datenschutz-Zwecken

Diese View zeigt nur Name und Abteilung – ideal für Benutzer, die keine Gehaltsdaten sehen sollen.

```python
CREATE VIEW public_mitarbeiter AS
SELECT name, abteilung
FROM mitarbeiter;
```

```python
SELECT * FROM public_mitarbeiter;
```

```python
DROP VIEW IF EXISTS public_mitarbeiter;
SELECT * FROM public_mitarbeiter;
```
