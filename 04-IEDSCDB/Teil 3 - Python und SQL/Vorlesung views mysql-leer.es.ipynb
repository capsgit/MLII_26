## Introducción rápida en MySQL

En este portátil aprenderás a crear, usar y borrar una vista en MySQL. Una vista es una consulta SQL guardada que funciona como una tabla virtual.

#### Eine Datenbank erstellen:
sql:

```python
CREATE DATABASE IF NOT EXISTS firma;
USE firma;
```

```python
CREATE TABLE mitarbeiter (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(100),
    abteilung VARCHAR(100),
    gehalt DECIMAL(10,2)
);
```

#### Ein paar Testdaten erstellen
sql:

```python

INSERT INTO mitarbeiter (name, abteilung, gehalt) VALUES
('Anna Müller', 'Vertrieb', 4500),
('Bernd Schulz', 'IT', 5200),
('Claudia Meier', 'Vertrieb', 4600),
('Dieter Huber', 'IT', 5400),
('Eva Schneider', 'Personal', 4100);
```

#### Eine erste Abfrage erstellen

```python
SELECT * FROM mitarbeiter;
```

#### Todos los empleados del departamento de TI

```python
SELECT * FROM mitarbeiter WHERE abteilung = "IT";
```

### ¿Qué es una vista?

Una vista es una consulta guardada que se trata como una tabla virtual. Es especialmente adecuada para proporcionar datos filtrados o reducidos.

```python
CREATE VIEW it_mitarbeiter AS
SELECT id, name, gehalt
FROM mitarbeiter
WHERE abteilung = 'IT';
```

```python
%%sql
SELECT * FROM it_mitarbeiter;
```

### Beispiel: View ohne Gehalt zur Datenschutz-Zwecken

Esta vista sólo muestra el nombre y el departamento ideal para los usuarios que no quieren ver los datos de salario.

```python
CREATE VIEW public_mitarbeiter AS
SELECT name, abteilung
FROM mitarbeiter;
```

```python
SELECT * FROM public_mitarbeiter;
```

```python
DROP VIEW IF EXISTS public_mitarbeiter;
SELECT * FROM public_mitarbeiter;
```
