### Methoden in einer Klasse definieren und aufrufen

Mit weiteren Methoden können wir das Verhalten eines Objektes definieren.
<i>Alle Hund, die mit der Klasse erstellt werden, können bellen.<i>

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


    def bellen(self):
        print("Wuff!")

anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")

anrei.bellen()
````

### Werte beim Methodenaufruf übergeben

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


    def bellen(self, anzahl):
        print(anzahl * "Wuff! ")

anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")

anrei.bellen(3)
````

### Weiters Beispiel für Methoden

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


    def bellen(self, anzahl):
        print(anzahl * "Wuff! ")

    def spielen(self, dauer):
        print("Der Hund spielt jetzt", dauer, "Minuten.")

anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")
taiko = Hund("Collie", "Taiko", "weiß")

anrei.bellen(3)
anrei.spielen(10)

taiko.bellen(10)
taiko.spielen(6)
````

Wie könnte man ändern, dass dort steht welcher Hund bellt?

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


    def bellen(self, anzahl):
        print(self.name, "sagt: ", anzahl * "Wuff! ")

    def spielen(self, dauer):
        print(f"{self.name} spielt jetzt", dauer, "Minuten.")

anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")
taiko = Hund("Collie", "Taiko", "weiß")

anrei.bellen(3)
anrei.spielen(10)

taiko.bellen(10)
taiko.spielen(6)
````

## Instanzbasierte Attribute und Methoden

#### Instanzbasierte Attribute (Eigenschaften):

- Diese werden auch als Instanzvariablen bezeichnet.
- Sie sind spezifisch für jede Instanz einer Klasse.
- Jedes Objekt einer Klasse kann unterschiedliche Werte für diese Attribute haben.
- Sie werden in der Regel innerhalb der \_\_init\_\_-Methode (Konstruktor) der Klasse initialisiert.

Beispiel:

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse  # Instanzbasiertes Attribut
        self.name = name   # Instanzbasiertes Attribut
        self.farbe = farbe   # Instanzbasiertes Attribut

    # Instanzbasierte (Objektbasierte) Methoden:
    def bellen(self, anzahl):
        """Instanzbasierte Methode"""
        print(self.name, "sagt: ", anzahl * "Wuff! ")

    def spielen(self, dauer):
        """Instanzbasierte Methode"""
        print(f"{self.name} spielt jetzt", dauer, "Minuten.")

# Instanzen erstellen:
anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")
taiko = Hund("Collie", "Taiko", "weiß")
````

#### Instanzbasierte Methoden

- Diese Methoden können auf die instanzbasierten Attribute zugreifen und sie modifizieren.
- Sie benötigen eine Referenz auf eine Instanz (üblicherweise self in Python) und werden für und mit dieser spezifischen Instanz aufgerufen.

## Klassenbasierte Attribute und Methoden

#### Klassenbasierte Attribute

- Auch als Klassenvariablen bekannt.
- Sie sind für die gesamte Klasse und somit für alle Instanzen derselben Klasse gleich.
- Änderungen an einem klassenbasierten Attribut beeinflussen alle Instanzen dieser Klasse.

Beispiel:

````python
class Auto:
    anzahl_autos = 0

    def __init__(self, marke, modell):
        self.marke = marke
        self.modell = modell

        Auto.anzahl_autos += 1
````

#### Klassenbasierte Methoden

- Diese Methoden sind mit der Klasse selbst verbunden, nicht mit einer spezifischen Instanz.
- In Python werden sie mit dem Dekorator @classmethod markiert und nehmen als ersten Parameter die Klasse selbst (üblicherweise cls genannt) anstelle einer Instanz.

````python
class Auto:
    anzahl_autos = 0

    def __init__(self, marke, modell):
        self.marke = marke
        self.modell = modell

        Auto.anzahl_autos += 1

    def __del__(self):

        Auto.anzahl_autos -= 1

    @classmethod
    def get_anzahl_autos(cls):
        print("Anzahl der Autos:", cls.anzahl_autos)


mein_auto = Auto("VW", "Tiguan")
ihr_auto = Auto("Opel", "Mokka")

Auto.get_anzahl_autos()

print("Auto wird entfernt")
del ihr_auto

Auto.get_anzahl_autos()
# ihr_auto.marke  # geht nicht mehr
````

#### \_\_del\_\_- Methode

Bei dieser Methode handelt es sich um eine magische Methode. Kennen Sie nicht? nicht schlimm, auch das wird in den folgenden Baustein erklärt.
Im folgenden ist diese aber sehr nützlich, da der Destruktor (__del__) sicher stellt, dass die Verbindung zur Datenbank geschlossen wird, wenn die Instanz der Klasse gelöscht wird.

Beispiel:

#### Klassenbasierte Methoden

- Diese Methoden sind mit der Klasse selbst verbunden, nicht mit einer spezifischen Instanz.
- In Python werden sie mit dem Dekorator @classmethod markiert und nehmen als ersten Parameter die Klasse selbst (üblicherweise cls genannt) anstelle einer Instanz.

````python
class Auto:
    anzahl_autos = 0

    def __init__(self, marke, modell):
        self.marke = marke
        self.modell = modell

        Auto.anzahl_autos += 1


    @classmethod
    def get_anzahl_autos(cls):
        print("Anzahl der Autos:", cls.anzahl_autos)


mein_auto = Auto("VW", "Tiguan")
ihr_auto = Auto("Opel", "Mokka")

Auto.get_anzahl_autos()
````

Zusammenfassend sind instanzbasierte Attribute und Methoden spezifisch für einzelne Objekte einer Klasse, während klassenbasierte Attribute und Methoden für die ganze Klasse und somit für alle ihre Instanzen gelten.

#### Gemeinsames Übungsprojekt: Bibliothek

**Funktionen**:

1. Hinzufügen neuer Bücher zur Bibliothek.
2. Anzeigen aller Bücher in der Bibliothek.
3. Suchen nach Büchern basierend auf dem Titel.

Beim Starten des Programms soll unter Umständen eine neue Datenbank erstellt werde, wenn noch keine vorhanden ist.

Klassenstruktur:
- Eine Klasse Buch, die die Attribute eines Buches wie Titel und Autor repräsentiert.
- Eine Klasse Bibliothek, die die Funktionalitäten zum Hinzufügen, Anzeigen und Suchen von Büchern bereitstellt und mit der SQLite-Datenbank interagiert.
- Eine Klasse datenbank, die die Datenbank repräsentiert und eine Methode enthält, die eine neue Datenbank erstellt, sollte keine Vorhanden sein

````python
import sqlite3
import os
from pathlib import Path
from pyfiglet import Figlet

class Buch:
    def __init__(self, titel, autor, ort):
        self.titel = titel
        self.autor = autor
        self.lagerort = ort

class Bibliothek:
    def __init__(self, database):
        self.datenbank = database

    def buch_hinzufuegen(self, buch):
        c = self.datenbank.verbindung.cursor()
        params = (buch.titel, buch.autor, buch.lagerort)
        c.execute("INSERT INTO buecher VALUES (NULL, ?, ?, ?)", params)
        self.datenbank.verbindung.commit()

    def alle_bucher_ausgeben(self):
        c = self.datenbank.verbindung.cursor()
        c.execute("SELECT * FROM buecher")
        for i, titel, author, ort in c:
            print(f"{i}  {titel} - {author}  {ort}")

    def buch_suchen(self, titel):
        c = self.datenbank.verbindung.cursor()
        params = ("%" + titel + "%",)
        c.execute("SELECT * FROM buecher WHERE titel like ?", params)

        for i, titel, author, ort in c:
            print(f"{i}  {titel} - {author}  {ort}")


class Datenbank:
    def __init__(self, pfad):
        self.pfad = pfad
        self.verbindung = None
        self.ist_neu = not os.path.exists(self.pfad)
        self.verbindung_herstellen()

    def verbindung_herstellen(self):
        """Verbindung zur Datenbank herstellen"""
        self.verbindung = sqlite3.connect(self.pfad)

        if self.ist_neu:
            self.datenbank_erstellen()

    def datenbank_erstellen(self):
        """Eine passende Datenbank wird erstellt"""
        c = self.verbindung.cursor()
        c.execute("""CREATE TABLE IF NOT EXISTS buecher (id integer constraint id primary key,
                                                         titel TEXT,
                                                         autor TEXT,
                                                         lagerort TEXT)""")
        self.verbindung.commit()

    def __del__(self):
        if hasattr(self, "verbindung"):
            self.verbindung.close()


def main():
    """HauptFUNKTION unseres Bibliothekprogramms"""
    f = Figlet(font='slant')
    print(f.renderText('Bibliothek'))

    items = ["Buch suchen", "Buch hinzufügen", "Inventarliste ausgeben", "Programmende"]

    for n, i in enumerate(items, 1):
        print(f"{n}\t{i}")

    auswahl = input("Ihre Auswahl: ")

    match auswahl:
        case "1":
            meine_bibliothek.buch_suchen(input("Titel: "))

        case "2":
            buch = Buch(input("Titel: "), input("Autor: "), input("Lagerort: "))
            meine_bibliothek.buch_hinzufuegen(buch)



ordner = Path(__file__).parent  # Ordner wird ermittelt
db_name = ordner / "bibliothek.db"  # Pfad wird erstellt

db = Datenbank(db_name)  # Instanz der Klasse Datenbank wird erstellt, der Datenbankname mit Pfad wird übergeben
meine_bibliothek = Bibliothek(db)  # Instanz der Klasse Bibliothek wird erstellt, und die Instanz der Datenbank wird übergeben.

# Buch hinzufügen:
# buch2 = Buch("Brave New World", "Aldous Huxley", "B4-001") # Objekt der Klasse Buch erstellen
# buch3 = Buch("Daten- und Prozessanalyse", "Sascha Kersken", "B10-011") # Objekt der Klasse Buch erstellen
#
# for i in buch2, buch3:
#     meine_bibliothek.buch_hinzufuegen(i)

# Alle Bücher ausgeben:
# meine_bibliothek.alle_bucher_ausgeben()

# Ein Buch suchen:
meine_bibliothek.buch_suchen("Brave")


if __name__ == "__main__":
    main()
````
