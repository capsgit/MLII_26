### Transferir valores al llamar al método

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


    def bellen(self, anzahl):
        print(anzahl * "Wuff! ")

anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")

anrei.bellen(3)
````

### Siguiente ejemplo de métodos

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe


    def bellen(self, anzahl):
        print(anzahl * "Wuff! ")

    def spielen(self, dauer):
        print("Der Hund spielt jetzt", dauer, "Minuten.")

anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")
taiko = Hund("Collie", "Taiko", "weiß")

anrei.bellen(3)
anrei.spielen(10)

taiko.bellen(10)
taiko.spielen(6)
````

## Atributos y métodos basados en la instancia

#### Instanzbasierte Attribute (Eigenschaften):

- Diese werden auch als Instanzvariablen bezeichnet.
- Son específicos para cada instancia de una clase.
- Cada objeto de una clase puede tener valores diferentes para estos atributos.
- Normalmente se inicializan dentro del método \_\_init\\__ (diseñador) de la clase.

Beispiel:

````python
class Hund:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse  # Instanzbasiertes Attribut
        self.name = name   # Instanzbasiertes Attribut
        self.farbe = farbe   # Instanzbasiertes Attribut

    # Instanzbasierte (Objektbasierte) Methoden:
    def bellen(self, anzahl):
        """Instanzbasierte Methode"""
        print(self.name, "sagt: ", anzahl * "Wuff! ")

    def spielen(self, dauer):
        """Instanzbasierte Methode"""
        print(f"{self.name} spielt jetzt", dauer, "Minuten.")

# Instanzen erstellen:
anrei = Hund("Germanischer Bärenhund", "Anrei", "Kupfer")
taiko = Hund("Collie", "Taiko", "weiß")
````

#### Instanzbasierte Methoden

- Estos métodos pueden acceder a los atributos basados en la instancia y modificarlos.
- Usted necesita una referencia a una instancia (normalmente self in Python) y se llama para y con esta instancia específica.

## Atributos y métodos basados en clases

#### Klassenbasierte Attribute

- Auch als Klassenvariablen bekannt.
- Son iguales para toda la clase y, por tanto, para todas las instancias de la misma clase.
- Änderungen an einem klassenbasierten Attribut beeinflussen alle Instanzen dieser Klasse.

Beispiel:

````python
class Auto:
    anzahl_autos = 0

    def __init__(self, marke, modell):
        self.marke = marke
        self.modell = modell

        Auto.anzahl_autos += 1
````

#### Klassenbasierte Methoden

- Estos métodos están vinculados a la propia clase, no a una instancia específica.
- En Python, se marcan con el decorador @classmetod y son los primeros parámetros que toman la clase misma (normalmente llamada cls) en lugar de una instancia.

````python
class Auto:
    anzahl_autos = 0

    def __init__(self, marke, modell):
        self.marke = marke
        self.modell = modell

        Auto.anzahl_autos += 1

    def __del__(self):

        Auto.anzahl_autos -= 1

    @classmethod
    def get_anzahl_autos(cls):
        print("Anzahl der Autos:", cls.anzahl_autos)


mein_auto = Auto("VW", "Tiguan")
ihr_auto = Auto("Opel", "Mokka")

Auto.get_anzahl_autos()

print("Auto wird entfernt")
del ihr_auto

Auto.get_anzahl_autos()
# ihr_auto.marke  # geht nicht mehr
````

#### Klassenbasierte Methoden

- Estos métodos están vinculados a la propia clase, no a una instancia específica.
- En Python, se marcan con el decorador @classmetod y son los primeros parámetros que toman la clase misma (normalmente llamada cls) en lugar de una instancia.

````python
class Auto:
    anzahl_autos = 0

    def __init__(self, marke, modell):
        self.marke = marke
        self.modell = modell

        Auto.anzahl_autos += 1


    @classmethod
    def get_anzahl_autos(cls):
        print("Anzahl der Autos:", cls.anzahl_autos)


mein_auto = Auto("VW", "Tiguan")
ihr_auto = Auto("Opel", "Mokka")

Auto.get_anzahl_autos()
````

En resumen, los atributos y métodos basados en la instancia son específicos de los objetos individuales de una clase, mientras que los atributos basados en clases y los métodos se aplican a toda la clase y, por tanto, a todas sus instancias.

#### Gemeinsames Übungsprojekt: Bibliothek

**Funktionen**:

1. Añadir nuevos libros a la biblioteca.
2. Anuncios de todos los libros de la biblioteca.
3. Buscar libros basados en el título.

Al iniciar el programa, es posible que se cree una nueva base de datos si aún no existe.

Estructura de la clase:
- Un libro de clase que representa los atributos de un libro como el título y el autor.
- Una biblioteca de clase que proporciona las funcionalidades para agregar, mostrar y buscar libros e interactúa con la base de datos SQLite.
- No debe existir una base de datos de clase que represente la base y que contenga un método que cree una nueva base de conocimientos.

````python
import sqlite3
import os
from pathlib import Path
from pyfiglet import Figlet

class Buch:
    def __init__(self, titel, autor, ort):
        self.titel = titel
        self.autor = autor
        self.lagerort = ort

class Bibliothek:
    def __init__(self, database):
        self.datenbank = database

    def buch_hinzufuegen(self, buch):
        c = self.datenbank.verbindung.cursor()
        params = (buch.titel, buch.autor, buch.lagerort)
        c.execute("INSERT INTO buecher VALUES (NULL, ?, ?, ?)", params)
        self.datenbank.verbindung.commit()

    def alle_bucher_ausgeben(self):
        c = self.datenbank.verbindung.cursor()
        c.execute("SELECT * FROM buecher")
        for i, titel, author, ort in c:
            print(f"{i}  {titel} - {author}  {ort}")

    def buch_suchen(self, titel):
        c = self.datenbank.verbindung.cursor()
        params = ("%" + titel + "%",)
        c.execute("SELECT * FROM buecher WHERE titel like ?", params)

        for i, titel, author, ort in c:
            print(f"{i}  {titel} - {author}  {ort}")


class Datenbank:
    def __init__(self, pfad):
        self.pfad = pfad
        self.verbindung = None
        self.ist_neu = not os.path.exists(self.pfad)
        self.verbindung_herstellen()

    def verbindung_herstellen(self):
        """Verbindung zur Datenbank herstellen"""
        self.verbindung = sqlite3.connect(self.pfad)

        if self.ist_neu:
            self.datenbank_erstellen()

    def datenbank_erstellen(self):
        """Eine passende Datenbank wird erstellt"""
        c = self.verbindung.cursor()
        c.execute("""CREATE TABLE IF NOT EXISTS buecher (id integer constraint id primary key,
                                                         titel TEXT,
                                                         autor TEXT,
                                                         lagerort TEXT)""")
        self.verbindung.commit()

    def __del__(self):
        if hasattr(self, "verbindung"):
            self.verbindung.close()


def main():
    """HauptFUNKTION unseres Bibliothekprogramms"""
    f = Figlet(font='slant')
    print(f.renderText('Bibliothek'))

    items = ["Buch suchen", "Buch hinzufügen", "Inventarliste ausgeben", "Programmende"]

    for n, i in enumerate(items, 1):
        print(f"{n}\t{i}")

    auswahl = input("Ihre Auswahl: ")

    match auswahl:
        case "1":
            meine_bibliothek.buch_suchen(input("Titel: "))

        case "2":
            buch = Buch(input("Titel: "), input("Autor: "), input("Lagerort: "))
            meine_bibliothek.buch_hinzufuegen(buch)



ordner = Path(__file__).parent  # Ordner wird ermittelt
db_name = ordner / "bibliothek.db"  # Pfad wird erstellt

db = Datenbank(db_name)  # Instanz der Klasse Datenbank wird erstellt, der Datenbankname mit Pfad wird übergeben
meine_bibliothek = Bibliothek(db)  # Instanz der Klasse Bibliothek wird erstellt, und die Instanz der Datenbank wird übergeben.

# Buch hinzufügen:
# buch2 = Buch("Brave New World", "Aldous Huxley", "B4-001") # Objekt der Klasse Buch erstellen
# buch3 = Buch("Daten- und Prozessanalyse", "Sascha Kersken", "B10-011") # Objekt der Klasse Buch erstellen
#
# for i in buch2, buch3:
#     meine_bibliothek.buch_hinzufuegen(i)

# Alle Bücher ausgeben:
# meine_bibliothek.alle_bucher_ausgeben()

# Ein Buch suchen:
meine_bibliothek.buch_suchen("Brave")


if __name__ == "__main__":
    main()
````
