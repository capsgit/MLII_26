# Teil 4: Modularisierung


Modularisierung in Python bezieht sich auf die Praxis, den Code in separate Module aufzuteilen, um die Wiederverwendbarkeit, Wartbarkeit und Organisation des Codes zu verbessern. Ein Modul in Python ist einfach eine Datei, die Python-Code enthält, und kann Funktionen, Klassen, Variablen oder ausführbaren Code enthalten. Hier sind die Schlüsselaspekte der Modularisierung in Python zusammengefasst:

Vorteile der Modularisierung:

1. **Wiederverwendbarkeit**: Code in Modulen kann leicht in verschiedenen Teilen des Programms oder in verschiedenen Programmen wiederverwendet werden.
2. **Wartbarkeit**: Modularisierter Code ist leichter zu verstehen, zu testen und zu warten.
3. **Namensraum-Organisation**: Module helfen dabei, Namenskonflikte zu vermeiden, indem sie einen separaten Namensraum für Funktionen, Klassen usw. bereitstellen.
4. **Erleichterte Zusammenarbeit**: In größeren Projekten ermöglicht die Modularisierung es mehreren Entwicklern, gleichzeitig an verschiedenen Modulen zu arbeiten.

Die folgenden Übungen können am einfachsten in Python-Dateien und nicht in den Jupyter-Notebooks ausgeführt werden.

# Globale Variablen

Globale Variablen in Python sind Variablen, die außerhalb von Funktionen definiert werden und daher im gesamten Skript oder Modul sichtbar sind. Sie können von jeder Funktion im selben Skript oder Modul gelesen werden. Das macht sie nützlich für Werte, die von verschiedenen Teilen des Programms geteilt werden sollen, aber sie sollten mit Vorsicht verwendet werden, da sie die Lesbarkeit und Wartbarkeit des Codes beeinträchtigen können. #### Definition und Verwendung

#### Vorsichtsmaßnahmen und Best Practices

- Vermeidung von Seiteneffekten: Die Verwendung globaler Variablen kann zu unerwarteten Seiteneffekten führen, da sie von jeder Funktion im Skript geändert werden können.
- Wartbarkeit und Lesbarkeit: Zu viele globale Variablen können den Code schwer lesbar und wartbar machen. Es ist oft besser, Werte als Parameter an Funktionen zu übergeben.
- Namenskonflikte: Globale Variablen können Namenskonflikte mit lokalen Variablen verursachen, was zu Fehlern führen kann.

Im Allgemeinen ist es eine gute Praxis, die Verwendung globaler Variablen zu minimieren und stattdessen auf lokale Variablen und Parameterübergabe zu setzen, um die Klarheit und Wartbarkeit des Codes zu verbessern.


## Globale Namensräume

**Globale Namensräume** in Python beziehen sich auf den Bereich, in dem globale Variablen und Funktionen definiert sind. Jedes Mal, wenn ein Python-Script ausgeführt wird, wird ein globaler Namensraum erstellt. Dieser Namensraum enthält alle Definitionen, die auf der obersten Ebene des Scripts, also nicht innerhalb einer Funktion oder Klasse, vorgenommen werden. Hier sind einige Schlüsselaspekte der globalen Namensräume:
Charakteristika globaler Namensräume

- **Definitionsebene**: Globale Namensräume enthalten alle Namen, die auf der obersten Ebene eines Scripts oder Moduls definiert sind.
- **Lebensdauer**: Der globale Namensraum existiert, solange das Script läuft. Sobald das Script beendet wird, wird auch der globale Namensraum (und alle darin enthaltenen Objekte) gelöscht.
- **Zugänglichkeit**: Namen im globalen Namensraum sind im gesamten Script zugänglich, außer sie werden durch lokale Namen in Funktions- oder Klassendefinitionen verdeckt.

## Der Begriff "**Scope**"

n Python umfasst der Begriff "Scope" den Bereich eines Programms, in dem ein Name (wie eine Variable oder eine Funktion) sichtbar und direkt zugänglich ist. **Erklärung**:

- **Globaler Scope**: globale_variable ist im gesamten Script verfügbar. Jede Funktion oder Klasse im Script kann auf sie zugreifen.
- **Umgebender (Enclosing) Scope**: Innerhalb der aeußere_funktion wird umgebende_variable definiert. Diese Variable ist sowohl in der aeußere_funktion als auch in jeder inneren Funktion wie innere_funktion zugänglich, aber nicht außerhalb der aeußere_funktion.
- **Lokaler Scope**: lokale_variable ist nur innerhalb der innere_funktion definiert und zugänglich.
- **Built-in Scope**: Nicht im Beispiel gezeigt, aber Python bietet auch einen Built-in Scope, der Namen wie len, print usw. enthält, die in jedem Teil des Programms verwendet werden können.

## Wichtiger Hinweis zu Scopes:

Der Zugriff auf Variablen folgt der **LEGB-Regel**: **Local**, **Enclosing**, **Global**, **Built-in**. 
Python sucht zuerst im lokalen Scope, dann im umgebenden Scope, danach im globalen Scope und zuletzt im Built-in Scope.
