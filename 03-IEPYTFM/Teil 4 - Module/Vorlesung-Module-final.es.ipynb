# Teil 4: Modularisierung


La modularización en Python se refiere a la práctica de dividir el código en módulos separados para mejorar la reutilizabilidad, la durabilidad y la organización del código. Un módulo en Pithon es simplemente un archivo que contiene código Pythion y puede contener funciones, clases, variables u código ejecutable. Aquí se resumen los aspectos clave de la modularización de Pytón:

Ventajas de la modularización:

1. **Wiederverwendbarkeit**: Code in Modulen kann leicht in verschiedenen Teilen des Programms oder in verschiedenen Programmen wiederverwendet werden.
2. **Mantenimiento**: El código modularizado es más fácil de entender, probar y mantener.
3. ** Organización de Sala de Nombres**: Los módulos ayudan a evitar conflictos de nombres proporcionando un espacio de nombres separado para funciones, clases, etc.
4. **Cooperación facilitada**: En proyectos más grandes, la modularización permite a varios desarrolladores trabajar simultáneamente en diferentes módulos.

Los siguientes ejercicios se pueden realizar con más facilidad en archivos Python y no en los portátiles de Jupyter.

## Importlib

"**importlib**" es un módulo integrado en Python que ofrece una funcionalidad rica y dinámica para importar módulos. Es especialmente útil si necesita importar módulos de forma dinámica, es decir, en tiempo de ejecución, y ofrece más flexibilidad que el estado de importación tradicional. importalib está disponible a partir de Pyton 3.1.

Estas son algunas de las funciones principales y los casos de aplicación de importlib:

### Import eines Moduls zur Laufzeit

Puede utilizar importlib. import_module() para importar un módulo en tiempo de ejecución basado en una cadena que contenga el nombre del módulo. Esto es especialmente útil si el nombre de un módulo que se va a importar no está disponible hasta el momento de ejecución.

```python
import importlib

modul = importlib.import_module("math")

print(modul.sqrt(25))  # Importiert math erst, wenn es gebraucht wird.
```

### Neu laden eines Moduls

A veces necesita actualizar o recargar un módulo ya importado, por ejemplo, si el código del módulo ha cambiado. Puede usar **importlib.reload()** para ello.

# Globale Variablen

Las variables globales en Python son variables que se definen fuera de las funciones y que, por tanto, son visibles en todo el script o módulo. Pueden ser leídas por cualquier función en el mismo guión o módulo, lo que las hace útiles para los valores que deben ser compartidos por diferentes partes del programa, pero deben ser utilizadas con precaución, ya que pueden afectar a la legibilidad y durabilidad del código. Aquí hay algunos puntos importantes sobre variables globales:

#### Definición y uso

**Definición**: Una variable global se define fuera de una función. **Acceso**: puede ser leída por cualquier función en el mismo script o módulo.

```python
globale_variable = "Ich bin global"

def demonstriere_globale_variable():
    print(globale_variable)  # Zugriff auf die globale Variable

demonstriere_globale_variable()
```

## Ändern globaler Variablen innerhalb einer Funktion

- Para cambiar una variable global dentro de una función, debe usar la palabra clave global.
- Si no fuera global, se crearía una nueva variable local con el mismo nombre dentro de la función, lo que puede causar confusión.

*Ejemplo para cambiar una variable global:

```python
globale_variable = 5


def veraendere_globale_variable():
    global globale_variable  # mit dem Keyword global kann man eine globale Variable ändern

    globale_variable = 10

    print(globale_variable)

veraendere_globale_variable()
print(globale_variable)
```

#### Precauciones y mejores prácticas

- Evitar efectos de página: el uso de variables globales puede dar lugar a efectos de páginas inesperados, ya que cualquier función del script puede cambiarlos.
- Perseverancia y legibilidad: Demasiadas variables globales pueden hacer que el código sea difícil de leer y guardar. A menudo es mejor pasar valores a funciones como parámetros.
- Conflictos de nombres: Las variables globales pueden causar conflictos de nombres con variables locales, lo que puede conducir a errores.

En general, es una buena práctica minimizar el uso de variables globales y, en su lugar, apostar por variables locales y transferencia de parámetros para mejorar la claridad y la durabilidad del código.

#### Solución potencial:

```python

globale_variable = 5


def veraendere_globale_variable(globale_variable):
    return globale_variable

neu = veraendere_globale_variable(15)
print(globale_variable)
print(neu)
```

## Espacios de nombres globales

**Los espacios globales de nombres** en Python se refieren al área en la que se definen las variables y funciones globales. Cada vez que se ejecuta un script de Pythin, se crea un espacio global de nombres. Este espacio de nombres contiene todas las definiciones que se realizan en el nivel superior del script, es decir, no dentro de una función o clase. Aquí están algunos aspectos clave de los espacios globales del nombre: Características de los dominios globales de nombre.

- **Nivel de definición**: Los espacios globales de nombres contienen todos los nombres definidos en el nivel superior de un script o módulo.
- **Tiempo de vida**: El espacio de nombres global existe mientras se ejecuta el script. Una vez que el guión se termina, también se elimina el espacio de nombre global (y todos los objetos que contiene).
- **Accesibilidad**: Los nombres en el espacio de nombres global son accesibles en todo el script, a menos que los nombres locales los oculten en definiciones funcionales o de clase.

## El término "**Scope**"

En Python, el término "Scope" incluye el área de un programa en el que un nombre (como una variable o una función) es visible y accesible directamente. Pythion utiliza tres tipos principales de Scopes: local, rodeado (enclosing) y Global. Además, existe el built-in Scope. Aquí hay un ejemplo que ilustra estos Scope:

```python
# Globale Variable (Global Scope)
globale_variable = "global"

def aeußere_funktion():
    # Umgebende (Enclosing) Variable
    umgebende_variable = "umgebend"

    def innere_funktion():
        # Lokale Variable
        lokale_variable = "lokal"

        # Zugriff auf alle Scopes
        print(lokale_variable)  # lokal
        print(umgebende_variable)  # umgebend
        print(globale_variable)  # global, nur lesend!

    innere_funktion()

# Aufruf der äußeren Funktion
aeußere_funktion()

# Zugriff auf die globale Variable, aber nicht auf umgebende oder lokale Variablen
print(globale_variable)  # global
```

**Declaración**:

- **Escopeta global**: global_variable está disponible en todo el script. Cualquier función o clase del script puede acceder a ella.
- **Enclosing Scope**: Dentro de la función aeustere_se define la variable_que rodea. Esta variable es accesible tanto en la función außere como en cualquier función interna, como en la_función interna, pero no fuera de la_externa.
- **Scope local**:_variable local sólo está definido y accesible dentro de la función_interna.
- **Built-in Scope**: No mostrado en el ejemplo, pero Python también ofrece una escala de construcción que contiene nombres como len, print, etc. que se pueden utilizar en cualquier parte del programa.

## Wichtiger Hinweis zu Scopes:

El acceso a variables sigue la regla **LEGB**: **Local**, **Enclosing**,**Global**,***Built-in**. Python busca primero en la escala local, luego en la escopeta circundante, después en la scope global y, por último, en el built in Scope.
