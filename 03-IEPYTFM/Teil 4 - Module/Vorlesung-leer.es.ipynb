# Teil 4: Modularisierung


La modularización en Python se refiere a la práctica de dividir el código en módulos separados para mejorar la reutilizabilidad, la durabilidad y la organización del código. Un módulo en Pithon es simplemente un archivo que contiene código Pythion y puede contener funciones, clases, variables u código ejecutable. Aquí se resumen los aspectos clave de la modularización de Pytón:

Ventajas de la modularización:

1. **Wiederverwendbarkeit**: Code in Modulen kann leicht in verschiedenen Teilen des Programms oder in verschiedenen Programmen wiederverwendet werden.
2. **Mantenimiento**: El código modularizado es más fácil de entender, probar y mantener.
3. ** Organización de Sala de Nombres**: Los módulos ayudan a evitar conflictos de nombres proporcionando un espacio de nombres separado para funciones, clases, etc.
4. **Cooperación facilitada**: En proyectos más grandes, la modularización permite a varios desarrolladores trabajar simultáneamente en diferentes módulos.

# Globale Variablen

Las variables globales en Python son variables que se definen fuera de las funciones y que, por tanto, son visibles en todo el script o módulo. Pueden ser leídas por cualquier función en el mismo guión o módulo, lo que las hace útiles para los valores que deben ser compartidos por diferentes partes del programa, pero deben ser utilizadas con precaución, ya que pueden afectar a la legibilidad y durabilidad del código. Aquí hay algunos puntos importantes sobre variables globales:

#### Definición y uso

#### Precauciones y mejores prácticas

- Evitar efectos de página: el uso de variables globales puede dar lugar a efectos de páginas inesperados, ya que cualquier función del script puede cambiarlos.
- Perseverancia y legibilidad: Demasiadas variables globales pueden hacer que el código sea difícil de leer y guardar. A menudo es mejor pasar valores a funciones como parámetros.
- Conflictos de nombres: Las variables globales pueden causar conflictos de nombres con variables locales, lo que puede conducir a errores.

En general, es una buena práctica minimizar el uso de variables globales y, en su lugar, apostar por variables locales y transferencia de parámetros para mejorar la claridad y la durabilidad del código.

## Espacios de nombres globales

**Los espacios globales de nombres** en Python se refieren al área en la que se definen las variables y funciones globales. Cada vez que se ejecuta un script de Pythin, se crea un espacio global de nombres. Este espacio de nombres contiene todas las definiciones que se realizan en el nivel superior del script, es decir, no dentro de una función o clase. Aquí están algunos aspectos clave de los espacios globales del nombre: Características de los dominios globales de nombre.

- **Nivel de definición**: Los espacios globales de nombres contienen todos los nombres definidos en el nivel superior de un script o módulo.
- **Tiempo de vida**: El espacio de nombres global existe mientras se ejecuta el script. Una vez que el guión se termina, también se elimina el espacio de nombre global (y todos los objetos que contiene).
- **Accesibilidad**: Los nombres en el espacio de nombres global son accesibles en todo el script, a menos que los nombres locales los oculten en definiciones funcionales o de clase.
