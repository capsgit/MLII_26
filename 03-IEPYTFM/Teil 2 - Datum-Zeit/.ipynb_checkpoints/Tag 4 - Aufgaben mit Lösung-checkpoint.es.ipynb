## Aufgabe 2:
A partir de la hora actual, desvíese una hora:

```python
from datetime import datetime, timedelta

jetzt = datetime.now()
frueher = jetzt + timedelta(minutes=-60)
frueher = frueher.strftime("%H:%M")


print("Zeit jetzt:", jetzt)
print("Zeit - 60 Minuten:", frueher)
```

## Aufgabe 4:
Escriba una función que calcula el volumen de un cubo. Devuelva la longitud, anchura, altura y pregunta antes de hacerlo.

```python
def berechnung(a, b, c):
    vol = a * b * c
    return vol

def main():
    a = int(input("Bitte geben Sie die Länge ein: "))
    b = int(input("Bitte geben Sie die Breite ein: "))
    c = int(input("Bitte geben Sie die Höhe ein: "))
    vol = berechnung(a, b, c)
    print("Das Volumen beträgt: ", vol)

main()
```

## Aufgabe 5:
Escriba una función que resuma las entradas en el carrito de la compra: •P0 • **Sugerencia:** Usted necesita un bucle **for** para realizar la tarea.

```python
#  Warenkorb
warenkorb = [19.99, 2.99, 1.99, 3.99, 14.99, 12.99]

#  Aufrufen der Funktion:
warenpreis(warenkorb)


#  Musterlösung:

def warenpreis(summe):
    total = 0
    for i in summe:
        total = total + i
    print(total)    

warenpreis(warenkorb)
```

##  Aufgabe 6:
Escriba un juego de adivinanzas en el que tendrá que agregar números generados al azar. Utilice funciones para ello.

```python
import random


def zufallszahlen():
    random.seed()
    a = random.randint(1, 100)
    b = random.randint(1, 100)
    return a, b


def eingabe():
    a, b = zufallszahlen()

    print("Bitte geben Sie das Ergebnis von ", a, "+", b, "ein.")
    eing = input("--> ")
    eing = int(eing)
    return eing, a, b


def main():
    eing, a, b = eingabe()
    erg = a + b
    if eing == erg:
        print("Die Eingabe ist richtig!")
    else:
        print("Leider falsch!")


if __name__ == '__main__':
    main()
```

## Aufgabe 7:
Amplíe el juego con un contador de puntos y vida. Utilice un bucle **while** para ello.

```python
import random
import sys


def zufallszahlen():
    random.seed()
    a = random.randint(1, 100)
    b = random.randint(1, 100)
    return a, b


def eingabe(lives, score):

    while lives > 0:
        a, b = zufallszahlen()
        print("Bitte geben Sie das Ergebnis von ", a, "+", b, "ein, noch", lives, "Leben verfügbar.")
        eing = input(" --> ")
        eing = int(eing)
        auswertung(a, b, eing, lives, score)


def auswertung(a, b, eing, lives, score):
    erg = a + b

    if eing == erg:
        print("Die Eingabe ist richtig!")
        score = score + 1
        eingabe(lives, score)

    else:
        print("Leider falsch!")
        lives = lives - 1
        eingabe(lives, score)

    if lives >= 0:
        print("Spielende! Sie haben", score, " Punkt(e)")
        sys.exit(0)


def main():
    lives = 3
    score = 0
    eingabe(lives, score)


if __name__ == '__main__':
    main()
```
