### Zusatzaufgabe 
Por favor, subdivida el código de su máquina expendedora de bebidas ya creada en funciones. Verá. El código será más legible directamente y evitará el doble código por la reutilizabilidad del código.

```python
# Getränkeautomat Version mit Funktionen
# -*- coding: utf-8 -*-
import sys

def main():
    menu()

def menu():
    # Hier wird für jedes Getränk eine Liste angelegt, in der der Name und der Preis festgelegt wird.
    getraenk1 = ["Cola", 1.00]
    getraenk2 = ["Fanta", 1.00]
    getraenk3 = ["Sprite", 1.00]
    getraenk4 = ["Saft", 1.10]
    getraenk5 = ["Wasser", 0.80]
    getraenk6 = ["Red Bull", 3.00]
    # Der Preis wird aus der Liste genommen. Das ":.2f" rundet auf 2 Stellen hinter dem Komma auf/ab.
    print("** Getränk -O- Mat***")
    print(f"1 - Cola: {getraenk1[1]:.2f} Euro")
    print(f"2 - Fanta: {getraenk2[1]:.2f} Euro")
    print(f"3 - Sprite: {getraenk3[1]:.2f} Euro")
    print(f"4 - Saft: {getraenk4[1]:.2f} Euro")
    print(f"5 - Wasser: {getraenk5[1]:.2f} Euro")
    print(f"6 - Red Bull: {getraenk6[1]:.2f} Euro")
    print("0 - beenden")

    auswahl = int(input("Bitte wählen Sie ein Getränk: (1-6) "))
    if auswahl == 0:
        print("Auf Wiedersehen!")
        sys.exit(0) # Das Programm wird mit dem Status 0 (erfolgreich) sauber beendet

    if auswahl == 1:
        betrag = getraenk1[1]
    if auswahl == 2:
        betrag = getraenk2[1]
    if auswahl == 3:
        betrag = getraenk3[1]
    if auswahl == 4:
        betrag = getraenk4[1]
    if auswahl == 5:
        betrag = getraenk5[1]
    if auswahl == 6:
        betrag = getraenk6[1]

    geldeinwurf(betrag)

def geldeinwurf(betrag):
    geld = float(input("Bitte werfen Sie Geld ein! "))

    if geld == 0:
        print("Auf Wiedersehen!")
        sys.exit(0)

    if betrag > geld:
        print()
        geldnachwerfen = input("Leider reicht ihr Geld nicht aus, möchten Sie noch Geld nachwerfen? (j/n) ")
        if geldnachwerfen != "j":
            geldauswurf(geld)
        else:
            fehlbetrag = betrag - geld
            nachwurf(fehlbetrag, geld, betrag)
    if betrag < geld:
        rueckgabe = geld - betrag
        ausgabe(rueckgabe)

    if betrag == geld:
        rueckgabe = 0
        ausgabe(rueckgabe)


def nachwurf(fehlbetrag, geld, betrag):
    print("Fehlbetrag: ", fehlbetrag)
    neuerbetrag = float(input(f"Bitte werfen Sie den fehlenden Betrag von mindestens {fehlbetrag:.2f} Euro ein: "))
    if neuerbetrag + geld >= betrag:
        geld = geld + fehlbetrag
        if geld - betrag >= 0:
            rueckgabe = geld - betrag
            ausgabe(rueckgabe)

    else:
        print("Geld reicht immer noch nicht")
        geldauswurf(geld)


def geldauswurf(geld):
    print(f"Sie erhalten {geld:.2f0} Euro zurück *klöter*")
    sys.exit(0)


def ausgabe(rueckgabe):
    print()
    print("*Klapper*, das Getränkt fällt in das Ausgabefach...")
    print(f"Sie erhalten {rueckgabe:.2f} Euro zurück *klöter*")  # Die Ausgabe wird auf 2 Kommastellen erweitert


if __name__ == '__main__':
    main()
```
