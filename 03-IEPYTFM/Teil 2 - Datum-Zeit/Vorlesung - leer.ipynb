### Funktion strftime()
Bessere Informationen bietet die Funktion **strftime()**. Diese benötigt 2 Parameter:
* Einen Formatierungsstring
* ein Zeittupel, der z. B. von **localtime()** geliefert wird

Eine Dokumentation von strftime() finden Sie unter der folgenden Adresse:<br> 
https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes
<br>


```python
import time

print (time.strftime("Tag.Monat.Jahr: -----> %d.%m.%Y"))
print (time.strftime("Stunde:Minute:Sekunde: -----> %H:%M:%S"))
print (time.strftime("Im 12h-Format: -----> %I:%M:%S Uhr %p"))
print (time.strftime("Datum und Zeit: -----> %c"))
print (time.strftime("nur Datum: -----> %x, nur Zeit: -----> %X"))
print (time.strftime("Wochentag abgekürzt: -----> %a, ganz: -----> %A, Nr. (Sonntag=0): -----> %w"))
print (time.strftime("Monat abgekürzt: -----> %b, ganz: -----> %B"))
print (time.strftime("Tag des jahres: -----> %j"))
print (time.strftime("Woche des Jahres, Start bei Sonntag: -----> %U"))
print (time.strftime("Woche des Jahres, Start bei Montag: -----> %W"))
print (time.strftime("Zeitzone: -----> %Z"))
```

### Mit Zeiten rechnen
Wieder eine stets beliebte Funktion ist es, mit Zeiten zu rechnen. Dazu wird **unter Umständen** das Modul **"datetime"** benötigt.
Mehr Informationen zu **datetime** finden Sie unter: https://docs.python.org/3/library/datetime.html

Wann ist Weihnachten?

```python
from datetime import datetime

hoy = datetime.now()
hoy2 = datetime.today()
xmas = datetime(2026, 12, 24)

print(f"Weihnachten ist in {(xmas - hoy).days} Tagen.")
```

In den folgenden Beispiel wird einfach die **aktuelle** Zeit genommen, und **45 Minuten** dazu gerechnet:

```python
from datetime import datetime, timedelta

ahora = datetime.now()
mas_tarde = ahora + timedelta(minutes=45)

print(f"ahora son las {ahora.strftime('%H:%M:%S')}")
print(f"en 45 minutos seran las {mas_tarde.strftime('%H:%M:%S')}")
```

Für eine vernünftige Zeitmessung, oder Logdateien etc. ist das vorherige Beispiel natürlich recht ungeeignet. Das folgende Beispiel zeigt eine Möglichkeit auf, eine gewisse Zeitspanne messen zu lassen:

Eine ziemlich umständliche, aber sehr flexible Art, die Dauer auszurechnen, ist die, die benötigte Zeit in Sekunden zu messen,und danach in Stunden und Minuten umzurechnen. Dazu wird die **Ganzzahlendivision** (siehe Tag 1) benötigt:

```python
from datetime import datetime, timedelta
from time import sleep

ahora = datetime.now()
print(f"ahora son las {ahora.strftime('%H:%M:%S')}")
mas_tarde = ahora + timedelta(minutes=45)

sleep(3)
ahora_mas_sleep = datetime.now()
tiempo_sleep = ahora_mas_sleep - ahora

print(f"ahora despues de {tiempo_sleep.seconds} segundos, son las {ahora_mas_sleep.strftime('%H:%M:%S')}")

print(f"en 45 minutos seran las {mas_tarde.strftime('%H:%M:%S')}")

#-------------------------------------------------------------------------------------
print("-----------------------------------------------------------------------------")
jetzt = datetime.now()  # Zeit jetzt wird genommen
print(jetzt.strftime("%H:%M:%S"))  # formatierte Zeitausgabe

sleep(3)  # Wartet 3 Sekunden

gleich = datetime.now()  # Zeit wird erneut genommen
print(gleich.strftime("%H:%M:%S"))  # formatierte Zeitausgabe

dauer = gleich - jetzt  # Dauer wird berechnet

print(f"Dauer: {dauer.seconds} Sekunden")  # Ausgabe der Sekunden (gerundet)
```

```python
from datetime import datetime
from time import sleep

jetzt = datetime.now()
print(jetzt)

sleep(2) # Wartezeit in Sekunden

gleich = datetime.now()
print(gleich)

dauer = gleich - jetzt

dauer_sek = dauer.total_seconds()
print("Dauer in Sekunden: ", dauer_sek)

# Stunden
stunden = dauer_sek // 3600
# verbleibende Sekunden
dauer_sek = dauer_sek - (stunden * 3600)
# Minuten
minuten = dauer_sek // 60  # Ganzzahlendivision
# übrig gebliebene Sekunden:
sekunden = dauer_sek - (minuten * 60)
# formatierte Ausgabe:
print('{:02}:{:02}:{:02}'.format(int(stunden), int(minuten), int(sekunden)))
```

#### Zeitangabe aus String erstellen:

```python
from datetime import datetime

t1 = "2019-12-24 18:00:00"  # Datum nach ISO-8601-Format (normaler String, nur formatiert halt)
dateobject_from_iso = datetime.fromisoformat(t1)
print("Ausgabe der Datetime-Objekte im Rohformat:")
print("Dateobject --------> ISO:", dateobject_from_iso)

t2 = "2019 01 11"

# mit "%Y %m %d" wird festgelegt, wie sich das Datum aus dem String zusammensetzt:
dateobject_from_str = datetime.strptime(t2, "%Y %m %d")
print("Dateobject -------> String:", dateobject_from_str)

print("Ausgabe der FORMATIERTEN Datetime-Objekte")
# Formatiert vom ISO-String:
print("ISO formatiert:", dateobject_from_iso.strftime("%d.%m.%Y"))

# oder vom String:
print(f"String formatiert: {dateobject_from_str:%d.%m.%Y}")
```

## Zeitgesteuerte Aufgaben

🕒 Einführung in das schedule-Modul in Python

Das schedule-Modul ist eine einfache und intuitive Bibliothek zur zeitgesteuerten Ausführung von Python-Funktionen. Es eignet sich besonders gut für kleine Automatisierungsaufgaben, wie z. B.:

* regelmäßiges Versenden von E-Mails,
* automatisiertes Daten-Backup,
* periodisches Abrufen von API-Daten,
* zeitlich gesteuerte Skriptaktionen.

🎯 Vorteile:
* einfache Syntax,
* keine komplexen Cronjobs notwendig,
* gut geeignet für Skripte und einfache Hintergrundprozesse.

Mit schedule wird eine Funktion geplant (z. B. „alle 10 Minuten ausführen“), und in einer Schleife regelmäßig geprüft, ob diese Aufgabe fällig ist.

Das Modul ist rein auf Python-Ebene und nutzt keine Systemdienste wie den Task-Scheduler unter Windows oder cron unter Linux.

🔧 Typische Einsatzbereiche
Beispiel	Beschreibung

* schedule.every(10).seconds	Alle 10 Sekunden
* schedule.every().hour	Jede volle Stunde
* schedule.every().day.at("08:00")	Täglich um 8:00 Uhr
schedule.every().monday.at("12:00")	Jeden Montag um 12:00 Uhr

### Job alle 10 Sekunden ausführen

```python
import schedule
import time

def job():
    print("Job läuft:", time.strftime("%H:%M:%S"))
schedule.run_pending()
schedule.every(5).seconds.do(job) #se crea un trabajo/tarea, cada 5 segundos se llamara a la funcion "job()"

print("holis")
"""
while False:
    schedule.run_pending() #hay un "trabajo" en espera?
    time.sleep(2)
"""

import schedule
import time
import shutil

def job():
    total, used, free = shutil.disk_usage("/")
    print(f"Gesamt: {total // (2 ** 30)} GB, Frei: {free // (2 ** 30)} GB")

    if free / total < 0.1:
        print("⚠️ Weniger als 10 % frei!")

schedule.every(5).seconds.do(job)  # Eine Aufgabe wird erstellt, alle 5 Sekunden, Funktion "job" ausführen

#Endlos-Schleife:
while True:
    schedule.run_pending()  # Hier wird überprüft, ob ein Job "fällig" ist
    time.sleep(1)  # Wartet 1 Sekunde. Kann angepasst werden, abhängig davon, wie oft ein Job erwartet wird.

```

Erklärung:

* <code>schedule.run_pending()</code> prüft, ob ein geplanter Job fällig ist.

* <code>time.sleep(1)</code> vermeidet zu hohe CPU-Auslastung.

### Täglicher Job zu einer bestimmten Uhrzeit

```python
import schedule
import time
import mss, mss.tools


def bildschirmaufnahme():
    """Screenshot von einem bestimmten Monitor (1, 2, 3, ...)"""
    monitor = 1
    filename = "screenshot.png"

    with mss.mss() as sct:
        monitor = sct.monitors[monitor]
        screenshot = sct.grab(monitor)
        mss.tools.to_png(screenshot.rgb, screenshot.size, output=filename)

        print("Screenshot gespeichert unter: " + filename)

#aca!!!
schedule.every().day.at("08:00").do(bildschirmaufnahme)

# Endlos-Schleife:
while True:
    schedule.run_pending()  # Hier wird überprüft, ob ein Job "fällig" ist
    time.sleep(1)  # Wartet 1 Sekunde. Kann angepasst werden, abhängig davon, wie oft ein Job erwartet wird.
```

Führt den Job täglich um 8 Uhr morgens aus. Die Zeitangabe muss im Format "HH:MM" (24h) sein.

### Montags und Mittwochs um 12:30 Uhr

```python
import schedule
import time

def mittagsjob():
    print("Job am Montag und Mittwoch um 12:30")

schedule.every().monday.at("12:30").do(mittagsjob)
schedule.every().friday.at("12:30").do(mittagsjob)

while True:
    schedule.run_pending()
    time.sleep(1)
```

Mögliche Zeitintervalle

   * schedule.every().second
   * schedule.every().minute
   * schedule.every().hour
   * schedule.every().day
   * schedule.every().monday (und andere Wochentage)

Man kann auch schedule.every(5).minutes oder schedule.every(2).hours usw. verwenden.
