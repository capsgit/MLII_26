## Aufgabe 1:
<br>
Bitte geben Sie das aktuelle Datum in der Form: Stunde:Minute:Sekunde aus.

```python
from datetime import datetime

aktuelle_zeit = datetime.now()

print(aktuelle_zeit)
```

## Aufgabe 2:
<br>
Ziehen Sie eine Stunde von der aktuellen Zeit ab:

```python
from datetime import datetime, timedelta

aktuelle_zeit = datetime.now()
delay = aktuelle_zeit - timedelta(minutes=60)
print("now: ", aktuelle_zeit.strftime("%H:%M:%S"))
print("delay", delay.strftime("%H:%M:%S"))
```
