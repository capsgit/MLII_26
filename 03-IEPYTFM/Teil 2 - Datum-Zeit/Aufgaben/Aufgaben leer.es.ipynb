## Aufgabe 1:
Por favor, introduzca la fecha actual en la forma: hora:minute:segundo.

```python
from datetime import datetime

aktuelle_zeit = datetime.now()

print(aktuelle_zeit)
```

## Aufgabe 2:
A partir de la hora actual, desvíese una hora:

```python
from datetime import datetime, timedelta

aktuelle_zeit = datetime.now()
delay = aktuelle_zeit - timedelta(minutes=60)
print("now: ", aktuelle_zeit.strftime("%H:%M:%S"))
print("delay", delay.strftime("%H:%M:%S"))
```
