### random.seed

Wenn **random.seed()** ohne Argument aufgerufen wird, initialisiert es den Zufallszahlengenerator (Random Number Generator, RNG) mit einem Wert, der auf einem Quellwert basiert, der von der aktuellen Systemzeit oder einem anderen zufälligen Wert abhängt.

Die Konsequenz davon ist, dass jedes Mal, wenn Sie Ihr Programm starten und random.seed() ohne Argument aufrufen, der RNG wahrscheinlich mit einem anderen Wert initialisiert wird. Das führt dazu, dass die Sequenz der generierten Zufallszahlen **bei jedem Programmlauf unterschiedlich** ist.

```python
import random

random.seed()  # Die Zufallszahl wird initialisiert

zufallszahl = random.randint(1, 10)  # Zufällige Ganzzahl zwischen 1 und 10

print(zufallszahl)  # Ausgabe
```

Wenn Sie **random.seed()** mit einem bestimmten Wert aufrufen, beginnt der RNG immer mit dem gleichen Startwert, was bedeutet, dass er jedes Mal, wenn das Programm ausgeführt wird, dieselbe Folge von Zufallszahlen generiert. Dies ist nützlich, wenn Sie **reproduzierbare** Ergebnisse benötigen, beispielsweise für Tests oder Simulationen.

```python
import random

random.seed(6)  # Die Zufallszahl wird initialisiert

zufallszahl = random.randint(1, 10)  # Zufällige Ganzzahl zwischen 1 und 10

print(zufallszahl)  # Ausgabe
```

## Zufallszahlen

#### Zufällige Gleitkommzahl zwischen 0 und 1:

```python
import random

zufalls_float = random.random()
print(zufalls_float)
```

#### Zufällige Gleitkommazahl in einem bestimmten Bereich:

```python
import random

zufalls_float = random.uniform(1.5, 7.5)
print(zufalls_float)
```

### Auswahl aus einer Sequenz

#### Zufälliges Element aus einer Liste auswählen

```python
meine_liste = ["uno", "dos", "tres", "cuatro", "cinco"]
element = random.choice(meine_liste)
print(element)
```

#### Mehrere zufällige Elemente aus einer Liste auswählen:

```python
elements = random.choices(meine_liste, k=3)
print(elements)
```

#### Listenelemente zufällig anordnen:

```python
import random

zufalls_float = random.uniform(1.5, 7.5)
print(zufalls_float)
```

 #### Gewichtete Zufallsauswahl

```python
elements = random.choices(meine_liste, k=3, weights=[99, 0, 0, 99, 0] )
print(elements)
```

Diese Beispiele zeigen einige der grundlegenden Möglichkeiten, wie Sie das random-Modul verwenden können, um verschiedene Arten von Zufälligkeit in Ihren Python-Programmen zu erzeugen

## Eine kleine Aufgabe:
#### Übungsaufgabe: Zufälliges Abendessen

**Ziel**: Schreiben Sie ein Programm, das ein zufälliges Abendessen aus einer Liste von möglichen Optionen auswählt.

**Aufgabenstellung**:

1. Erstellen Sie eine Liste mit verschiedenen Abendessen-Optionen (z.B. Pizza, Pasta, Salat, Burger, Sushi).
2. Verwenden Sie random.choice, um zufällig eines der Abendessen auszuwählen.
3. Geben Sie die ausgewählte Option aus.

4. **Optional**: Erlauben Sie dem Benutzer, eigene Optionen zur Liste hinzuzufügen, bevor die Auswahl getroffen wird.

```python
import random

cena = ["Pizza", "Sushi", "Hot dog", "Pho", "Pasta", "Brot", "Dönner"]

def mostrar_menu(menu):
    print("\n📋 Menú actualizado:")
    for i, plato in enumerate(menu, start=1):
        print(f"  {i}. {plato}")

while True:
    nuevo_plato = input(
        "\n➕ Escribe un plato para añadir al menú\n"
        "🎲 O presiona Enter para elegir una cena aleatoria:\n> "
    ).strip()

    if nuevo_plato == "":
        print("\n" + "-" * 35)
        print(f"🍴 Hoy hay {random.choice(cena)} para cenar!!! 😋")
        print("-" * 35 + "\n")
        break

    nuevo_plato_formato = nuevo_plato.title()

    if nuevo_plato_formato not in cena:
        cena.append(nuevo_plato_formato)
        mostrar_menu(cena)
    else:
        print(f"\n⚠️ Ya tenemos {nuevo_plato_formato} en el menú. Escribe otro o presiona Enter.")
```
