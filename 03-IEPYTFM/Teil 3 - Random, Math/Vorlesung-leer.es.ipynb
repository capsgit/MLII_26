### random.seed

Cuando **random.seed()** es llamado sin argumento, inicializa el generador de números aleatorios (RNG) con un valor basado en un valor de origen que depende del tiempo actual del sistema o de otro valor aleatorio.

La consecuencia es que cada vez que inicia su programa y llama al azar.seed() sin ningún argumento, es probable que el RNG se inicialice con otro valor, lo que hace que la secuencia de números aleatorios generados ** sea diferente en cada ejecución**.

```python
import random

random.seed()  # Die Zufallszahl wird initialisiert

zufallszahl = random.randint(1, 10)  # Zufällige Ganzzahl zwischen 1 und 10

print(zufallszahl)  # Ausgabe
```

Cuando llama **random.seed() ** con un valor determinado, el RNG comienza siempre con el mismo valor inicial, lo que significa que genera la misma secuencia de números aleatorios cada vez que se ejecuta el programa. Esto es útil si necesita resultados **reproducibles, por ejemplo para pruebas o simulaciones.

```python
import random

random.seed(6)  # Die Zufallszahl wird initialisiert

zufallszahl = random.randint(1, 10)  # Zufällige Ganzzahl zwischen 1 und 10

print(zufallszahl)  # Ausgabe
```

### Auswahl aus einer Sequenz

#### Seleccionar elemento aleatorio de una lista

```python
meine_liste = ["uno", "dos", "tres", "cuatro", "cinco"]
element = random.choice(meine_liste)
print(element)
```

#### Seleccionar varios elementos aleatorios de una lista:

```python
elements = random.choices(meine_liste, k=3)
print(elements)
```

#### Ordenar los elementos de la lista al azar:

```python
import random

zufalls_float = random.uniform(1.5, 7.5)
print(zufalls_float)
```

 #### Gewichtete Zufallsauswahl

```python
elements = random.choices(meine_liste, k=3, weights=[99, 0, 0, 99, 0] )
print(elements)
```

Estos ejemplos muestran algunas de las formas básicas de cómo utilizar el módulo aleatorio para crear diferentes tipos de aleabilidad en sus programas de Python

## Eine kleine Aufgabe:
#### Tarea de ejercicio: Cena aleatoria

**Objetivo**: Escriba un programa que elija una cena aleatoria de una lista de opciones posibles.

**Aufgabenstellung**:

1. Crear una lista de diferentes opciones de cena (p. ej. pizza, pasta, ensalada, hamburguesa, sushi).
2. Utilice random.choice para seleccionar al azar una de las cenas.
3. Escriba la opción seleccionada.

4. **Opcional**: Permite al usuario añadir sus propias opciones a la lista antes de seleccionar.

```python
import random

cena = ["Pizza", "Sushi", "Hot dog", "Pho", "Pasta", "Brot", "Dönner"]

def mostrar_menu(menu):
    print("\n📋 Menú actualizado:")
    for i, plato in enumerate(menu, start=1):
        print(f"  {i}. {plato}")

while True:
    nuevo_plato = input(
        "\n➕ Escribe un plato para añadir al menú\n"
        "🎲 O presiona Enter para elegir una cena aleatoria:\n> "
    ).strip()

    if nuevo_plato == "":
        print("\n" + "-" * 35)
        print(f"🍴 Hoy hay {random.choice(cena)} para cenar!!! 😋")
        print("-" * 35 + "\n")
        break

    nuevo_plato_formato = nuevo_plato.title()

    if nuevo_plato_formato not in cena:
        cena.append(nuevo_plato_formato)
        mostrar_menu(cena)
    else:
        print(f"\n⚠️ Ya tenemos {nuevo_plato_formato} en el menú. Escribe otro o presiona Enter.")
```

Es un número irracional, lo que significa que no puede ser representado como una fracción simple y que su representación decimal no termina ni se repite periódicamente. El valor de "e" es aproximadamente 2,71828.

El número Euleriano desempeña un papel central en muchas áreas de las matemáticas, especialmente en el análisis. Es la base de logaritmos naturales y actúa en diferentes contextos matemáticos, como la fórmula para la función exponencial y ecuaciones diferenciales. La función exex (la función exponential básica e) tiene la característica única de que su derivada es igual a la función misma.
