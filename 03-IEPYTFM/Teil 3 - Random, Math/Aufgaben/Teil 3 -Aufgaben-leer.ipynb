### Aufgabe 1: Umwandlung von Radiant in Grad

Schreiben Sie ein Programm, das einen Winkel in Radiant vom Benutzer annimmt und diesen in Grad umwandelt.

```python
import math

def convertir_radian_a_grados(radian):
    return math.degrees(radian)

try:
    radianes = float(input("Introduce un ángulo en radianes: "))
    grados = convertir_radian_a_grados(radianes)
    print(f"{radianes} radianes son {grados:.2f} grados")
except ValueError:
    print("Por favor, introduce un número válido.")
```

### Aufgabe 2: Flächeninhalt eines Kreises

Schreiben Sie ein Programm, das den Radius eines Kreises als Eingabe erhält und den Flächeninhalt des Kreises berechnet. Verwenden Sie die Konstante math.pi für die Berechnung.

Formel: pi * r<sup>2</sup>

```python
def calcular_area_circulo(radio):
    return math.pi * radio ** 2

try:
    radio_del_circulo = float(input("ingresa el radio del circulo (en metros): "))
    area_del_circulo = calcular_area_circulo(radio_del_circulo)
    print(f"El area del circulo es: {area_del_circulo:.2f} m².")
except ValueError:
    print("El valor ingresado no es un numero")
```

### Aufgabe 3: Prüfung auf Ganzzahligkeit

Schreiben Sie ein Programm, das eine Gleitkommazahl einliest und prüft, ob diese eine ganze Zahl ist (d.h., ob der dezimale Teil 0 ist).

```python
def verificar_numero_entero(numero):
    return numero.is_integer()

try:
    numero = float(input("Ingresa un número: "))
    if verificar_numero_entero(numero):
        print(f"El número {numero} es entero")
    else:
        print(f"El número {numero} no es entero")
except ValueError:
    print("El valor ingresado no es un número válido")
```

### Aufgabe 4: Zufällige Begrüßung

Erstellen Sie eine Liste von Begrüßungsformeln (wie "Hallo", "Guten Tag", "Willkommen") und wählen Sie mit random.choice zufällig eine aus, um den Benutzer zu begrüßen.

```python
import random

saludos = ["Hola", "Q hubo", "Q pasa", "Q dice", "Holis", "Buenas las tengas", "Cantalas"]

def crear_saludo(nombre: str) -> str:
    saludo = random.choice(saludos)
    return f"{saludo}, {nombre}!"

nombre = input("Ingresa tu nombre: ").strip().title()
print(crear_saludo(nombre))
```

### Aufgabe 5: Lottozahlen-Generator

Schreiben Sie ein Programm, das sechs eindeutige Lottozahlen zwischen 1 und 49 generiert und ausgibt. Verwenden Sie random.sample, um sicherzustellen, dass keine Zahl doppelt vorkommt.

```python
lottozahlen = random.sample(range(1, 50), 6)
print(lottozahlen)
```

### Aufgabe 6: Zufällige Passworterzeugung

Erstellen Sie ein Programm, das ein zufälliges Passwort generiert. Das Passwort sollte eine Mischung aus Groß- und Kleinbuchstaben sowie Zahlen und Sonderzeichen sein. Nutzen Sie **random.choice** in Kombination mit einer entsprechenden Zeichenliste.

Tipp:
Erinnern Sie sich an die String-Funktionen **string.ascii_letters**, **string.digits** und **string.punctuation**.

```python
import string

MIN_LEN = 12 #???
MAX_LEN = 16 #???

caracteres = (
    string.ascii_lowercase +
    string.ascii_uppercase +
    string.digits +
    string.punctuation
)

def create_password():
    longitud = random.randint(MIN_LEN, MAX_LEN)
    return "".join(random.choice(caracteres) for _ in range(longitud))

password = create_password()
print(f"Tu nueva contraseña: {password}")
```

### Bonus:
Lassen Sie den Benutzer entscheiden, wie lang sein Passwort sein soll, und wie sich das Passwort zusammen setzt.

```python
def create_password(longitud, mayus, minus, numeros, simbolos):
    caracters = ""

    if mayus:
        caracters += string.ascii_uppercase
    if minus:
        caracters += string.ascii_lowercase
    if numeros:
        caracters += string.digits
    if simbolos:
        caracters += string.punctuation

    if not caracters:
        raise ValueError("Debe seleccionar al menos un tipo de carácter.")

    return "".join(random.choice(caracters) for _ in range(longitud))

try:
    longitud = int(input("Longitud del password: "))
    mayus = input("¿Incluir mayúsculas? (s/n): ").lower() == "s"
    minus = input("¿Incluir minúsculas? (s/n): ").lower() == "s"
    numeros = input("¿Incluir números? (s/n): ").lower() == "s"
    simbolos = input("¿Incluir símbolos? (s/n): ").lower() == "s"

    password = create_password(longitud, mayus, minus, numeros, simbolos)
    print(f"Password generado: {password}")

except ValueError as e:
    print(e)
```
