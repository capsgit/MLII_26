## Funktionen

En el caso de los programas más pequeños y sencillos, no ha habido ningún problema en escribir el llamado "código Spagetti." Por lo tanto, código que se trabaja simplemente desde arriba hacia abajo. En los proyectos grandes o en crecimiento, el **modularización** del código con la ayuda de **funciones** es mucho más eficaz. El P0 es un código reutilizable y mucho más claro. Además, los programas útiles pueden reutilizarse más fácilmente en otros programas. Algunas funciones ya están definidas en Python, por ejemplo **print()**, **input() **, etc. Se reconocen funciones en los paréntesis finales.

Lo que es increíblemente importante para las funciones** son los **remociones**. Al igual que para los lijados. De acuerdo con el estándar PEP8 (guide de estilo Python) Se supone que un **reemplazamiento** es de **4 espacios**. IDEs modernos como PyCharm transforman automáticamente la presión sobre el tabulador en 4 espacios. Además, después de PEP8, **2 líneas en blanco** según la definición de una función.
* Los nombres funcionales no se ajustarán al nombre de las funciones Python prefabricadas o incorporadas.
* Las funciones no deben comenzar con un dígito
* Las funciones pueden incluir los caracteres a-z, A-Z, números y el signo "_."
* De acuerdo con el estándar PEP8, los nombres de funciones deben ser escritos en pequeña escala, no como en el caso de C en Camel-Case.
