# Teil 1: Funktionen

Bei den bisherigen recht kleinen und übersichtlichen Programmen war es kein Problem sogenannten “Spagetticode” zu schreiben. Also Code, der einfach von oben nach unten abgearbeitet wird. Bei umfangreichen oder wachsenen Projekten ist das modularisieren von Code mit Hilfe von Funktionen deutlich effektiver. Dadurch wird Code wiederverwendbar und viel übersichtlicher. Nützliche Programme können außerdem leichter in anderen Programmen wiederverwendet werden. Einige Funktionen sind unter Python schon vorgegeben, zum Beispiel print(), input() usw. Man erkennt Funktionen an den abschließenden Klammern.

#### Funktionsobjekte

Funktionsobjekte in Python sind Instanzen von Funktionen. Wenn Sie eine Funktion in Python definieren, erstellen Sie tatsächlich ein Funktionsobjekt. Dieses Konzept ist zentral für das Verständnis der funktionalen Programmierung in Python, da es ermöglicht, Funktionen wie jede andere Art von Objekten zu behandeln. Hier sind einige Schlüsselaspekte von Funktionsobjekten:

Erste-Klasse-Objekte: In Python sind Funktionen "Erste-Klasse-Objekte". Das bedeutet, dass sie wie jede andere Art von Wert (wie String, Integer, Liste usw.) behandelt werden können. Sie können als Argumente an andere Funktionen übergeben, von Funktionen zurückgegeben, in Variablen zugewiesen und in Datenstrukturen wie Listen oder Wörterbücher gespeichert werden.

### Einfache Funktionen
Einfache Funktionen führen beim Aufruf immer dieselbe Funktion, ohne Übergabe von irgendwelchen Parametern aus.

```python
def menue():
    """
    Zeigt dem Benutzer ein Menü an.
    """
    print("=====================")
    print("=   Hauptmenü       =")
    print("=====================")
    print("= 1. Neues Spiel    =")
    print("= 2. Spiel laden    =")
    print("= 3. Einstellungen  =")
    print("= 0. Ende           =")
    print("=====================")

menue()
```

Was bei Funktionen  unglaublich **wichtig** ist, sind die **Einrückungen**. Genau wie bei Schleifen. Laut PEP8-Standard (Python-Styleguide) Soll eine **Einrückung** aus **4 Leerzeichen** bestehen. Moderne IDEs wie PyCharm wandeln einen Druck auf Tabulator automatisch in 4 Leerzeichen um. Außerdem sollen nach PEP8 **2 Leerzeilen** nach der Definition einer Funktion erfolgen. <br>
Weitere wichtige Informationen:
* Funktionsnamen dürfen nicht den Namen vorgefertigter oder eingebauter Python-Funktionen entsprechen.
* Funktionen dürfen nicht mit einer Ziffer beginnen
* Funktionen dürfen die Zeichen a-z, A-Z, aus Ziffern und dem"_"-Zeichen bestehen.
* Laut PEP8-Standard sollen Funktionsnamen klein geschrieben werden, nicht wie bei Beispielsweise C in Camel-Case.


### Funktionen mit einem Parameter
Diese Funktionen ermöglichen es, das bei Aufruf Informationen an die Funktion weitergegeben werden können. Diese Information nennt sich **Parameter**. Diese Parameter werden dann innerhalb der Funktion ausgewertet oder weiterbearbeitet.<br>


```python
def quadrat(zahl):
    print(f"{zahl} * {zahl} = {zahl * zahl}")

def addiere(zahl):
    print(f"{zahl} + {zahl} = {zahl + zahl}")

eingabe = int(input("Bitte geben Sie eine Zahl ein: "))
quadrat(eingabe)

addiere(eingabe)
```

### Funktionen mit mehreren Parametern
Manchmal reicht ein Parameter nicht aus. Da müssen schonmal 2 oder mehr her.

Entpacken von Parameterlisten:

Das "Entpacken einer Parameterliste" in Python bezieht sich auf die Technik, bei der eine Liste oder ein Tupel von Werten in separate Variable aufgeteilt wird. Dies wird häufig verwendet, um Argumente an eine Funktion zu übergeben oder um Werte aus einer iterierbaren Sequenz in einzelne Variablen zu entpacken. Es gibt zwei Hauptformen des Entpackens: das Entpacken in einzelne Variablen und das Entpacken beim Funktionsaufruf.

**Entpacken in einzelne Variablen***: Hierbei wird eine Liste oder ein Tupel in mehrere einzelne Variablen aufgeteilt. Zum Beispiel:

```python
werte = [1, 2, 3]
a, b, c = werte

print(a)  # Gibt 1 aus
print(b)  # Gibt 2 aus
print(c)  # Gibt 3 aus
```

**Entpacken beim Funktionsaufruf**: Sie können Listen oder Tupel verwenden, um mehrere Argumente an eine Funktion zu übergeben. Dies geschieht durch Voranstellen eines Sterns * vor der Liste oder dem Tupel. Zum Beispiel:



```python

def addiere(zahl1, zahl2, zahl3):
    print(zahl1 + zahl2 + zahl3)



zahlen = [1, 2, 3]  # Wichtig! Die Länge der Liste muss der Anzahl der Parameter entsprechen!

# Anstatt die Liste VORHER in Variablen zu entpacken, machen wir das mit *zahlen
addiere(*zahlen)  # Durch das "*" in *zahlen wird die Liste in den Parametern der Funktion "entpackt"
```

Das Entpacken von Parameterlisten ist besonders nützlich, um Code flexibler und lesbarer zu gestalten, insbesondere wenn die Anzahl der Elemente in der Liste oder im Tupel genau der Anzahl der erforderlichen Argumente einer Funktion entspricht.

#### Normale Schlüsselwortparameter:
 Diese können sowohl positionell als auch als Schlüsselwortargumente übergeben werden. Das bedeutet, dass Sie den Parameter entweder durch seine Position in der Funktionsdefinition oder durch explizites Nennen seines Namens übergeben können. Zum Beispiel:

```python
def addiere(zahl1, zahl2, zahl3):
    print(zahl1, zahl2, zahl3)

# addiere(1, 2, 3)  # Übergabe der Parameter positionell

addiere(zahl3=3, zahl2=2, zahl1=1)  # Übergabe als Keyword
```

#### Reine Schlüsselwortparameter:

Diese müssen explizit als Schlüsselwortargumente übergeben werden. Sie werden in der Funktionsdefinition nach einem Stern * angegeben. Ein reiner Schlüsselwortparameter kann nicht positionell übergeben werden. Zum Beispiel:

```python
def addiere(*, zahl1, zahl2, zahl3):
    print(zahl1, zahl2, zahl3)

# addiere(1, 2, 3)  # Übergabe der Parameter positionell  (funktioniert HIER NICHT!)

addiere(zahl3=3, zahl2=2, zahl1=1)  # Übergabe als Keyword. (Ist in diesem Fall zwingend!)
```

### Funktionen mit Rückgabewert
Wir haben gelernt, wie wir Werte übergeben, um diese weiter zu bearbeiten. Aber wie bekommen wir die bearbeiten Werte zurück?

```python
def addiere(a, b):
    return a + b

zahl1 = addiere(2, 3)  # Zahlen werden addiert und in Variable gespeichert (5)
zahl2 = addiere(4, 5)  # Zahlen werden addiert und in Variable gespeichert (9)

ergibt = addiere(zahl1, zahl2)  # Beide Ergebnisse werden nochmals addiert und in Variable gespeichert (14)

print(ergibt)  # Inhalt der Variable wird ausgegeben
```

### Funktionen mit einem Standardwert
Es besteht die Möglichkeit Parametern von Optionen Standardwerte zu geben, die genutzt werden, sollten keine Werte übergeben werden. 
Man nennt diese Technik auch "**optionale Parameter**".

```python
def addiere(a=3, b=5):
    return a + b

print(addiere())  # Beide Standardwerte werden genutzt (8)
print(addiere(5))  # Der Standardwert "b" wird genutzt, weil Position 1 mit 5 überschrieben wird. (10)
print(addiere(b=10))  # Standardwert "a" wird genutzt (13)
print(addiere(10, 20))  # Beide Werte werden ersetzt
```

Mit *args können eine variable Zahl von Argumenten abgefangen werden. Das ist in vielerlei Hinsicht nützlich. Unter anderen auch dafür, wenn Funktionen Argumente übergeben werden, die entweder überhaupt nicht benötigt werden (Beispielsweise bei der Nutzung von Frameworks), oder sich die Anzahl der übergebenen Argumente von Aufruf zu Aufruf verschieden sein können.

### **kwargs
Bei **kwargs (Keyword Arguments) werden die Argumente in Form von Schlüsseln übergeben:

```python
def key_arguments(**kwargs):
    print(kwargs)  # Demo, dass es sich um Dictionary handel

    # Dictionary entpacken
    for schluessel, wert in kwargs.items():
        # print(schluessel, wert)
        print(f"{schluessel} enthält {wert}")

key_arguments(a=1, b=2)
key_arguments(vorname="Dirk", nachname="Wagner", telefonnummer="08154")
```

### assert-Anweisung
Die assert-Anweisung wird verwendet, um eine Bedingung zu überprüfen und einen AssertionError zu werfen, wenn die Bedingung False ist. Sie dient hauptsächlich zur Fehlersuche und zum Testen von Code. Der AssertionError kann abgefangen werden. Dazu aber später mehr. <br>
Syntax: assert Bedingung, Fehlermeldung

**Beispiel 1: Grundlegende Verwendung:**

```python
def dividiere(a=5, b=6):
    assert b != 0, "Division durch 0 ist nicht erlaubt"
    return (a/b)

print(dividiere())
```

```python
def dividiere(a,b):
    assert b != 0, "Division durch 0 ist nicht erlaubt"
    return (a/b)

try:
    print(dividiere(10,0))

except AssertionError as e:
    print("ups")
```

**Beispiel 2: Überprüfung von Eingabeparametern**:

```python
def berechne_durchschnitt(liste):
    assert isinstance(liste, list), "Eingabe muss eine Liste sein"
    assert len(liste) > 0, "Liste darf nicht leer sein"
    assert all(isinstance(x, (int, float)) for x in liste), "Alle Elemente müssen Zahlen sein"

    return sum(liste) / len(liste)

# Funktioniert
print(berechne_durchschnitt([1, 2, 3, 4, 5]))  # Ausgabe: 3.0

# Würde AssertionError werfen:
# berechne_durchschnitt("nicht eine liste")  # AssertionError: Eingabe muss eine Liste sein
# berechne_durchschnitt([])  # AssertionError: Liste darf nicht leer sein
# berechne_durchschnitt([1, 2, "drei"])  # AssertionError: Alle Elemente müssen Zahlen sein
```

Die assert-Anweisung ist besonders nützlich während der Entwicklung, um Annahmen über deinen Code zu überprüfen und Fehler früh zu entdecken. In Produktionsumgebungen werden Assertions oft deaktiviert (mit dem -O-Flag beim Starten von Python), da sie hauptsächlich für die Entwicklung gedacht sind.

### Funktionen in der Praxis: Beispiel eines Logins

```python
import getpass

name = ""
passwort = ""
counter = 0

def login_ok(name, passwort):
    db_name = "dirk".lower()  # das kommt natürlich aus einer Datenbank
    db_passwort = "geheim"  # das kommt natürlich aus einer Datenbank

    return name.lower() == db_name and passwort == db_passwort

# print(login_ok("Hans", "geheim"))  # legitimer Unittest (debugging)

def interner_bereich():
    print("Hier passiert magisches!")

while not login_ok(name, passwort):
    # Eingabe:
    name = input("Bitte geben Sie Ihren Namen ein: ")
    passwort = getpass.getpass("Bitte geben Sie Ihr Passwort ein: ")


    # Auswertung:
    if login_ok(name, passwort):
        print("Login erfolgreich!")
        interner_bereich()
        break

    else:
        print("Login fehlgeschlagen!")
        counter += 1
        if counter >= 3:
            print("Sie haben drei falsche Login-Versuche gemacht. Das Programm wird beendet.")
            break
```

### Rekursion

Eine Rekursion in der Programmierung, insbesondere in Bezug auf Funktionen, bezeichnet den Prozess, bei dem eine Funktion sich selbst innerhalb ihres eigenen Codes aufruft. Diese Technik wird verwendet, um Probleme zu lösen, die auf natürliche Weise in kleinere, ähnliche Probleme aufgeteilt werden können. Rekursive Funktionen sind ein grundlegendes Konzept in der Informatik und werden oft in der algorithmischen Programmierung eingesetzt.

Um eine rekursive Funktion zu erstellen, müssen Sie zwei Hauptkomponenten berücksichtigen:

1. Rekursiver Fall: Dies ist der Teil der Funktion, in dem die Funktion sich selbst aufruft. Der rekursive Fall reduziert das Problem in kleinere Teile und ruft die Funktion mit diesen neuen Parametern wieder auf.

2. Basisfall: Um eine unendliche Rekursion zu verhindern, muss jede rekursive Funktion einen Basisfall haben. Der Basisfall definiert eine Bedingung, unter der die Rekursion endet und keine weiteren Selbstaufrufe der Funktion erfolgen.

```python
def fakultät(n):
    if n == 0: # Basisfall
        return 1

    else:
        return n * fakultät(n-1)  # Rekursiver Fall

print(fakultät(5))  # 5 * 4 * 3 * 2 * 1 = 120
```

In diesem Beispiel wird fakultät(**n**) rekursiv aufgerufen, indem sie sich selbst mit dem Argument **n - 1** aufruft, bis **n 0** erreicht. Bei **n = 0** tritt der Basisfall ein, der die Rekursion beendet.

## Lambda-Funktionen
Lambda-Funktionen, auch bekannt als **anonyme Funktionen**, sind kleine Funktionen in Python, die ohne einen Namen definiert werden können. Sie werden häufig in Situationen verwendet, in denen eine Funktion als Argument an eine andere Funktion übergeben werden soll, oder wenn eine kurze Funktion benötigt wird, die nicht unbedingt eine eigene benannte Funktion erfordert.

Der Syntax einer Lambda-Funktion:
<code>lambda argumente: ausdruck</code>

Hier ist eine Aufschlüsselung der verschiedenen Teile einer Lambda-Funktion:

* Das Schlüsselwort lambda leitet die Definition einer Lambda-Funktion ein.
* Die arguments stehen für die Eingabeparameter der Funktion. Es können mehrere Argumente durch Kommata getrennt werden, aber es ist auch möglich, keine Argumente zu haben.
* Der Doppelpunkt : trennt die Argumente von der Ausdrucks- oder Rückgabewertseite.
* Der expression-Teil enthält den Ausdruck, der evaluiert wird und den Rückgabewert der Lambda-Funktion darstellt.

Hier ist ein einfaches Beispiel, das verdeutlicht, wie man eine Lambda-Funktion in Python verwendet, um die Quadrate von Zahlen zu berechnen:

```python
power = lambda x, y: x**y

print(power(2, 3))
```
