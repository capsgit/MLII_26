## Aufgabe 1:
Escriba una función para calcular el **circuito** de un círculo y pase el diámetro a la función para el cálculo. La fórmula necesaria: u = π·d. Como π es 3.14.

```python
import math

def perimetro_circulo(diametro):
    return math.pi * diametro

while True:
    try:
        diametro = float(input("Por favor, introduce el diámetro del círculo en metros: "))

        if diametro <= 0:
            print("❌ El diámetro debe ser un número mayor que 0.")
        else:
            break

    except ValueError:
        print("❌ Entrada inválida. Debes introducir un número.")

resultado = perimetro_circulo(diametro)

print(f"Con un diámetro de {diametro} m, el perímetro del círculo es {resultado} m")
```

## Aufgabe 2:
Escriba una función que calcula el volumen de un cubo. Devuelva la longitud, anchura, altura y pregunta antes de hacerlo.

```python
def volumen_quader(*, laenge, breite, hoehe):
    return laenge * breite * hoehe


while True:
    try:
        laenge = float(input("Geben Sie die Länge des Quaders in Metern an: "))
        breite = float(input("Geben Sie die Breite des Quaders in Metern an: "))
        hoehe  = float(input("Geben Sie die Höhe des Quaders in Metern an: "))

        if laenge <= 0 or breite <= 0 or hoehe <= 0:
            print("❌ Fehler: Alle Maße müssen positive Zahlen sein.")
        else:
            break

    except ValueError:
        print("❌ Fehler: Bitte geben Sie gültige Zahlen ein.")


volumen = volumen_quader(
    laenge=laenge,
    breite=breite,
    hoehe=hoehe
)

print(
    f"Ein Quader mit einer Länge von {laenge} m, "
    f"einer Breite von {breite} m und "
    f"einer Höhe von {hoehe} m "
    f"hat ein Volumen von {volumen} m³."
)
```

## Aufgabe 3:
Escriba una función que resuma las entradas en el carrito de la compra: •P0 • **Sugerencia:** Usted necesita un bucle **for** para realizar la tarea.

```python
import random

def generar_warenkorb(n):
    warenkorb = []
    for _ in range(n):
        warenkorb.append(round(random.uniform(1, 100), 2))
    return warenkorb


random.seed(1)
warenkorb = generar_warenkorb(55)

def formato_moneda(valor):
    return f"{valor:.2f}".replace(".", ",")


def warenkorb_summe(warenkorb):
    total = 0
    for preis in warenkorb:
        if preis < 0:
            raise ValueError("Preise dürfen nicht negativ sein.")
        total += preis

    return total

total_final = warenkorb_summe(warenkorb)
print(f"Gesamtpreis: {formato_moneda(total_final)} €")
```

##  Aufgabe 4:
Escriba un juego de adivinanzas en el que tendrá que agregar números generados al azar. Utilice funciones para ello.

```python
import random

def generate_numbers():
    return random.randint(1, 100), random.randint(1, 100)

def read_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("❌ Bitte geben Sie eine ganze Zahl ein.")

def play():
    a, b = generate_numbers()
    result = a + b

    response = read_int(f"Wie viel ist {a} + {b}? ")

    if response == result:
        print(f"✅ Richtig! Es ist {result}. Gut gemacht.")
    else:
        print(f"❌ Falsch. Die richtige Antwort ist {result}.")

play()
```

## Aufgabe 5:
Amplíe el juego con un contador de puntos y vida. Utilice un bucle **while** para ello.

```python
import random

def generate_numbers():
    return random.randint(1, 100), random.randint(1, 100)

def read_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("❌ Bitte geben Sie eine ganze Zahl ein.")

def play():
    punkte = 0
    leben = 3
    runde = 1

    print("🎮 Willkommen zum Additionsspiel!")
    print(f"Start: Punkte = {punkte}, Leben = {leben}\n")

    while leben > 0:
        print(f"--- Runde {runde} ---")
        print(f"Aktueller Stand: Punkte = {punkte}, Leben = {leben}")

        a, b = generate_numbers()
        result = a + b

        response = read_int(f"""
        Wie viel ist {a} + {b}?
        ⭐ = {punkte} | 💔 = {leben}
        """)

        if response == result:
            punkte += 1
            print(f"✅ Richtig! {a} + {b} = {result}.")
        else:
            leben -= 1
            print(f"❌ Falsch. {a} + {b} = {result}.")
            print(f"💔 Du hast ein Leben verloren.")

        print(f"➡️ Neuer Stand: Punkte = {punkte}, Leben = {leben}\n")
        runde += 1

    print("🏁 Spiel vorbei! Keine Leben mehr übrig.")
    print(f"⭐ Endpunktzahl: {punkte} Punkte")

play()
```
