### Funciones con un parámetro
Estas funciones permiten transmitir información a la función cuando se llama. Esta información se llama **Parámetros**. Estos parámetros se evalúan o reprocesan dentro de la función.

```python
def quadrat(zahl):
    print(f"{zahl} * {zahl} = {zahl * zahl}")

def addiere(zahl):
    print(f"{zahl} + {zahl} = {zahl + zahl}")

eingabe = int(input("Bitte geben Sie eine Zahl ein: "))
quadrat(eingabe)

addiere(eingabe)
```

### Funciones con múltiples parámetros
A veces un parámetro no es suficiente.

```python
def addiere(zahl1, zahl2):
    print(f"{zahl1} + {zahl2} = {zahl1 + zahl2}")


eingabe = int(input("Bitte geben Sie eine Zahl ein: "))
eingabe2 = int(input("Bitte geben Sie noch eine Zahl ein: "))
addiere(eingabe, eingabe2)
```

Entpacken von Parameterlisten:

El "descomprimir una lista de parámetros" en Python se refiere a la técnica en la que una lista o un bulto de valores se divide en variable separada. Esto se utiliza con frecuencia para pasar argumentos a una función o para descomprimir valores de una secuencia iterable en variables individuales. Existen dos formas principales de desembalaje: el desacoplamiento a variables individuales y el desenvuelto al llamar a la función.

**Entpacken in einzelne Variablen***: Hierbei wird eine Liste oder ein Tupel in mehrere einzelne Variablen aufgeteilt. Zum Beispiel:

```python
werte = [1, 2, 3]
a, b, c = werte

print(a)  # Gibt 1 aus
print(b)  # Gibt 2 aus
print(c)  # Gibt 3 aus
```

**Descomprimir cuando se llama a la función**: Puede utilizar listas o tupel para pasar varios argumentos a una función. Esto se hace preponiendo una estrella * antes de la lista o del tubo. Por ejemplo:


```python

def addiere(zahl1, zahl2, zahl3):
    print(zahl1 + zahl2 + zahl3)



zahlen = [1, 2, 3]  # Wichtig! Die Länge der Liste muss der Anzahl der Parameter entsprechen!

# Anstatt die Liste VORHER in Variablen zu entpacken, machen wir das mit *zahlen
addiere(*zahlen)  # Durch das "*" in *zahlen wird die Liste in den Parametern der Funktion "entpackt"
```

El descomprimir las listas de parámetros es especialmente útil para hacer que el código sea más flexible y legible, especialmente si el número de elementos en la lista o en el tubo corresponde exactamente al número de argumentos necesarios para una función.

#### Parámetros de palabra clave normales:
Esto significa que puede pasar el parámetro ya sea por su posición en la definición de función o por el nombre explícito de su nombre. Por ejemplo:

```python
def addiere(zahl1, zahl2, zahl3):
    print(zahl1, zahl2, zahl3)

# addiere(1, 2, 3)  # Übergabe der Parameter positionell

addiere(zahl3=3, zahl2=2, zahl1=1)  # Übergabe als Keyword
```

#### Parámetros de palabra clave puros:

Se deben pasar explícitamente como argumentos de palabras clave. Se indican en la definición de función según una estrella *. Un parámetro de palabras claves puras no se puede pasar posicionalmente. Por ejemplo:

```python
def addiere(*, zahl1, zahl2, zahl3):
    print(zahl1, zahl2, zahl3)

# addiere(1, 2, 3)  # Übergabe der Parameter positionell  (funktioniert HIER NICHT!)

addiere(zahl3=3, zahl2=2, zahl1=1)  # Übergabe als Keyword. (Ist in diesem Fall zwingend!)
```

### Funciones con valor de retorno
Aprendimos a pasar valores para editarlos, pero ¿cómo recuperamos los valores editados?

```python
def addiere(a, b):
    return a + b

zahl1 = addiere(2, 3)  # Zahlen werden addiert und in Variable gespeichert (5)
zahl2 = addiere(4, 5)  # Zahlen werden addiert und in Variable gespeichert (9)

ergibt = addiere(zahl1, zahl2)  # Beide Ergebnisse werden nochmals addiert und in Variable gespeichert (14)

print(ergibt)  # Inhalt der Variable wird ausgegeben
```

### Funciones con un valor predeterminado
Existe la posibilidad de dar parámetros de opciones a valores predeterminados que se usen si no se pasa ningún valor. Esta técnica también se llama "Parametros opcionales**."

```python
def addiere(a=3, b=5):
    return a + b

print(addiere())  # Beide Standardwerte werden genutzt (8)
print(addiere(5))  # Der Standardwert "b" wird genutzt, weil Position 1 mit 5 überschrieben wird. (10)
print(addiere(b=10))  # Standardwert "a" wird genutzt (13)
print(addiere(10, 20))  # Beide Werte werden ersetzt
```

### **kwargs
En **kwargs (Argumentos de palabras clave) los argumentos se pasan en forma de claves:

```python
def key_arguments(**kwargs):
    print(kwargs)  # Demo, dass es sich um Dictionary handel

    # Dictionary entpacken
    for schluessel, wert in kwargs.items():
        # print(schluessel, wert)
        print(f"{schluessel} enthält {wert}")

key_arguments(a=1, b=2)
key_arguments(vorname="Dirk", nachname="Wagner", telefonnummer="08154")
```

### assert-Anweisung
La instrucción assert se utiliza para comprobar una condición y lanzar un error de aserción si la condición es false. Se utiliza principalmente para la búsqueda de errores y para probar código. La asertionError puede ser interceptada. Pero más adelante. QTP0 Sintaxis: asert Condición, mensaje de error

**Beispiel 1: Grundlegende Verwendung:**

```python
def dividiere(a=5, b=6):
    assert b != 0, "Division durch 0 ist nicht erlaubt"
    return (a/b)

print(dividiere())
```

```python
def dividiere(a,b):
    assert b != 0, "Division durch 0 ist nicht erlaubt"
    return (a/b)

try:
    print(dividiere(10,0))

except AssertionError as e:
    print("ups")
```

** Ejemplo 2: Verificación de parámetros de entrada**:

```python
def berechne_durchschnitt(liste):
    assert isinstance(liste, list), "Eingabe muss eine Liste sein"
    assert len(liste) > 0, "Liste darf nicht leer sein"
    assert all(isinstance(x, (int, float)) for x in liste), "Alle Elemente müssen Zahlen sein"

    return sum(liste) / len(liste)

# Funktioniert
print(berechne_durchschnitt([1, 2, 3, 4, 5]))  # Ausgabe: 3.0

# Würde AssertionError werfen:
# berechne_durchschnitt("nicht eine liste")  # AssertionError: Eingabe muss eine Liste sein
# berechne_durchschnitt([])  # AssertionError: Liste darf nicht leer sein
# berechne_durchschnitt([1, 2, "drei"])  # AssertionError: Alle Elemente müssen Zahlen sein
```

La instrucción assert es especialmente útil durante el desarrollo para comprobar suposiciones sobre tu código y descubrir errores antes de tiempo. En entornos de producción, las aserciones a menudo se desactivan (con la bandera -O al iniciar Python), ya que están destinadas principalmente al desarrollo.

### Funciones prácticas: Ejemplo de inicio de sesión

```python
import getpass

name = ""
passwort = ""
counter = 0

def login_ok(name, passwort):
    db_name = "dirk".lower()  # das kommt natürlich aus einer Datenbank
    db_passwort = "geheim"  # das kommt natürlich aus einer Datenbank

    return name.lower() == db_name and passwort == db_passwort

# print(login_ok("Hans", "geheim"))  # legitimer Unittest (debugging)

def interner_bereich():
    print("Hier passiert magisches!")

while not login_ok(name, passwort):
    # Eingabe:
    name = input("Bitte geben Sie Ihren Namen ein: ")
    passwort = getpass.getpass("Bitte geben Sie Ihr Passwort ein: ")


    # Auswertung:
    if login_ok(name, passwort):
        print("Login erfolgreich!")
        interner_bereich()
        break

    else:
        print("Login fehlgeschlagen!")
        counter += 1
        if counter >= 3:
            print("Sie haben drei falsche Login-Versuche gemacht. Das Programm wird beendet.")
            break
```

### Rekursion

Una recursión en la programación, especialmente en relación con las funciones, indica el proceso en el que una función se llama a sí misma dentro de su propio código. Esta técnica se utiliza para resolver problemas que se pueden dividir naturalmente en problemas más pequeños y similares. Las funciones recursivas son un concepto básico en la informática y a menudo se utilizan en la planificación algorítmica.

Para crear una función recursiva, debe tener en cuenta dos componentes principales:

1. Caso recursivo: Esta es la parte de la función en la que la función se llama a sí misma. El caso recursiva reduce el problema en partes más pequeñas y recupera la función con estos nuevos parámetros.

2. Caso de base: Para evitar una recursión infinita, cada función recursiva debe tener un caso de base. El caso de referencia define una condición en la que la recursión termina y no hay más autollamadas de la función.

```python
def fakultät(n):
    if n == 0: # Basisfall
        return 1

    else:
        return n * fakultät(n-1)  # Rekursiver Fall

print(fakultät(5))  # 5 * 4 * 3 * 2 * 1 = 120
```

En este ejemplo, se llama recursivamente a la facultad(**n**) llamando a sí misma con el argumento **n - 1** hasta **n 0**. En **n = 0** ocurre el caso base que termina la recursión.

## Lambda-Funktionen
Las funciones Lambda, también conocidas como **Anónimos**, son pequeñas funciones en Python que se pueden definir sin un nombre. A menudo se utilizan en situaciones en las que una función debe pasarse como argumento a otra función, o cuando se necesita una función corta que no necesariamente requiere una función llamada propia.

La sintaxis de una función Lambda: Argumentos de lambda de lampP0: expresión P1

Aquí está un desglose de las diferentes partes de una función Lambda:

* La palabra clave lambda introduce la definición de una función Lambda.
* Los argumentos representan los parámetros de entrada de la función. Se pueden separar varios argumentos por comas, pero también es posible no tener argumentos.
* El doble punto: separa los argumentos de la página del valor de expresión o de retorno.
* La parte de expresión contiene la expresión que se evalúa y que representa el valor de retorno de la función Lambda.

He aquí un ejemplo sencillo que ilustra cómo usar una función Lambda en Python para calcular los cuadrados de números:

```python
power = lambda x, y: x**y

print(power(2, 3))
```
