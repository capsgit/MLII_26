# Teil 5: Fehlerbehandlung (Exceptions)

Por desgracia, como programador, es obligatorio realizar un gran esfuerzo para proteger el código de las entradas falsas.

### Einfache Fehlerbehandlung
La forma más fácil de interceptar errores es un simple bloque de Try:/Except. En el siguiente ejemplo se espera la entrada de un número entero. Si el usuario introduce una letra, esto provocaría un error y el programa se terminará:

```python
while True:
    eingabe = int(input("Bitte gib eine Zahl ein: "))
    print(eingabe)
```

Con la ayuda de un bloque de Try:/Except: el programador puede interceptar el error y así evitar la interrupción del programa: Ejemplo simple, pero no según el estándar PEP8, pero lo haremos más tarde.

```python
from sys import exit

while True:
    eingabe = input("Bitte gib eine Zahl ein (e für exit): ")
    if eingabe == "e":
        exit(0)

    try:
        eingabe = int(eingabe)
        print(eingabe ** 2)

    except ValueError:
        print("Bitte gib eine Zahl ein.")
```

### Mehrere Fehler abfangen
Dependiendo de los errores, se pueden llamar varios mensajes de error, o el programador puede distinguir qué tipo de mensaje de error debe pasar. El mayor problema es siempre un error lógico. Estos errores no podemos interceptarlos y por lo general son difíciles de encontrar.

```python
while True:
    zahl = input("Bitte gib eine Zahl ein (e für exit): ")
    teiler = int("bitte geben Sie einen Teiler ein")

    if zahl == "e":
        exit(0)

    try:
        zahl = int(zahl)
        teiler = int(teiler)

        ergebnis = zahl / teiler
        print(f"Das Ergebnis von {zahl} / {teiler} ist {ergebnis}.")

    except ValueError:
        print("Bitte nur Zahlen eingeben!")
        exit(1)

    except ZeroDivisionError:
        print("Der Teiler darf nicht 0 sein!")
        exit(2)
```


Ejemplo **División por 0**. Por supuesto, las divisiones por 0 no se pueden realizar. El programa abandonaría con un mensaje de error.

Im folgenden recht realen Beispiel werden 3 verschiedene Fehlermeldungen abgefangen:

```python
while True:

    try:
        zahl = input("Bitte geben Sie eine Zahl ein (e für Ende): ")
        teiler = input("Bitte geben Sie einen Teiler ein: ")

    except KeyboardInterrupt:
        print("Abgebrochen!")
        exit(0)

    if zahl == "e":
        exit(0)

    try:
        zahl = int(zahl)
        teiler = int(teiler)

        ergebnis = zahl / teiler
        print(f"Das Ergebnis von {zahl} / {teiler} ist {ergebnis}.")

    except ValueError:
        print("Bitte nur Zahlen eingeben!")
        exit(1)

    except ZeroDivisionError:
        print("Der Teiler darf nicht 0 sein!")
        exit(2)

    except KeyboardInterrupt:
        print("Abgebrochen!")
        exit(0)
```

#### Imprimir y tratar el mensaje de error

```python
while True:

    try:
        zahl = input("Bitte geben Sie eine Zahl ein (e für Ende): ")
        teiler = input("Bitte geben Sie einen Teiler ein: ")

    except KeyboardInterrupt:
        print("Abgebrochen!")
        exit(0)

    if zahl == "e":
        exit(0)

    try:
        zahl = int(zahl)
        teiler = int(teiler)

        ergebnis = zahl / teiler
        print(f"Das Ergebnis von {zahl} / {teiler} ist {ergebnis}.")

    except Exception as e:
        print("Es ist ein Fehler aufgetreten")
        print(e)
        continue
```

Puede que desee responder a un evento específico que no sea el tratamiento de errores. En el siguiente ejemplo, desee programar un formulario de correo electrónico. Si no se ha registrado una dirección de correo válida, se activará un error. Primero, cree una clase llamada por el mensaje de error. Esta clase no tiene que contener contenido. (Más información sobre clases en el tema OOP)

```python
class InvalidMailError(Exception):
    pass

def mailform(adresse, betreff, inhalt):

    if not "@" in adresse:
        raise InvalidMailError("Sie haben keine gültige eMailadresse eingegeben!")  # raise löst einen Fehler aus

    print("Eine eMail mit dem folgenden Inhalt wurde versendet:!")
    print("Empfänger: ", adresse)
    print("Betreff: ", betreff)
    print("Inhalt: ", inhalt)

mailform("dirk@wbs.de", "Test", "Hallo Welt!")
```

Al llamar a la función aparecerá el error deseado: "**InvalidMail**" que podemos interceptar ** ahora**.

```python
class InvalidMailError(Exception):
    pass

def mailform(adresse, betreff, inhalt):

    if not "@" in adresse:
        raise InvalidMailError("Sie haben keine gültige eMailadresse eingegeben!")  # raise löst einen Fehler aus

    print("Eine eMail mit dem folgenden Inhalt wurde versendet:!")
    print("Empfänger: ", adresse)
    print("Betreff: ", betreff)
    print("Inhalt: ", inhalt)

try:
    mailform("dirkwbs.de", "Test", "Hallo Welt!")

except InvalidMailError:
    print("Es ist ein Fehler aufgetreten!")
    exit(1)
```

#### Excursión Regex (expresiones regulares)

```python
import re

EMAIL_REGEX = re.compile(
    r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
)

class InvalidMailError(Exception):
    pass


def mailform(adresse, betreff, inhalt):
    if not EMAIL_REGEX.match(adresse):
        raise InvalidMailError("Sie haben keine gültige eMailadresse eingegeben!")  # raise löst einen Fehler aus

    print("Eine eMail mit dem folgenden Inhalt wurde versendet:!")
    print("Empfänger: ", adresse)
    print("Betreff: ", betreff)
    print("Inhalt: ", inhalt)


try:
    mailform("dirk@wbs.de", "Test", "Hallo Welt!")
except InvalidMailError as e:
    print(e)
```

## Error-Handling Hirarchie

La jerarquía de errores en Python se basa en la clase BaseException, de la que se derivan todas las demás clases de excepción y error. Aquí hay algunos puntos importantes de esta jerarquía:

- **BaseException**: En la parte superior de la Jerarquía está la clase BaseExcepción. Es la clase base para todas las demás clases de excepción. Por lo general, debe derivar excepciones a la exclusión, no directamente de BaseExcepción.

- **Excepción**: Esta clase es un descendiente directo de BaseException y la superclase para la mayoría de las clases de excepción estándar. Debería heredar Exception cuando cree sus propias clases de excepciones personalizadas.

- **Clasificaciones de error estándar**: En Exception, hay muchas clases de error predefinidas que representan diferentes tipos de errores:
- **ArithmeticError**: Clase básica para errores en operaciones aritméticas como **OverflowErrorror **, **ZeroDivisiónErrora**.
- **BufferError**: Error al acceder a operaciones relacionadas con Buffer.
- **LookupError**: Clase básica de errores en las operaciones de lookup, como IndexErrorror (índice falso en secuencias) y KeyErrord (la clave que falta en un diccionario).
- **EnvironmentError**: clase base para errores que se producen fuera del intérprete de Python (sustituido en Pytón 3 por **OSERROR**).
- **OSError**: Error relacionado con el sistema operativo, como archivo no encontrado (**FileNotFoundErroR**) o incumplimiento de acceso (**PermissionErrord**).

- **Clasificaciones de excepción**: Puede crear sus propias clases de excepción heredando Exception o cualquiera de sus subclases. Esto es útil para mostrar condiciones de error específicas en su aplicación.

**Herarquía y herencia**: El principio de herencia en la jerarquía de excepción significa que con un bloque de except puede interceptar un grupo de errores relacionados mediante el tratamiento de la clase base. Por ejemplo, except Exception: interceptaría todos los errores no relacionados con el sistema.

**Excepciones de salida del sistema**: **SystemExit**, **KeyboardInterrupt** y ** GeneratorExit ** son excepciones especiales derivadas de **BaseException** y no de Exception. Estos son para eventos relacionados con el sistema y normalmente no deben ser capturados en un código típico de tratamiento de errores.

El conocimiento exacto y el uso de esta jerarquía le permiten crear aplicaciones Python más robustas y tolerantes al responder a tipos de errores específicos y tratarlas en consecuencia.

# Logging

**Logging** es un método normalizado en Python para registrar información sobre el funcionamiento de un programa. Es una técnica importante para la búsqueda de errores y la supervisión del comportamiento de las aplicaciones. Pyton ofrece una funcionalidad de registro flexible con el módulo **logging**.

## Grundlagen des Python logging-Moduls

1. **Importar** del módulo: Para utilizar el sistema de registro, primero importe el módulo de registro.
2. **niveles de registro**: El módulo define diferentes niveles de registro estándar, como DEBUG, INFO, ADVERTENCIA, ERROR y CRITICAL, que indican la gravedad o el significado de los mensajes de registro.
3. **ConfigurarLogger**: Puede configurar el logger para especificar qué mensajes se registran y adónde se envían (por ejemplo, consola, archivo).

Ein einfaches Beispiel:

```python
import logging

# Grundkonfiguration des Loggings (Standard ist WARNING und höher)
logging.basicConfig(level=logging.DEBUG)

# Verschiedene Log-Nachrichten
logging.debug("Debug-Information")
logging.info("Allgemeine Info")
logging.warning("Warnung")
logging.error("Ein Fehler ist aufgetreten")
logging.critical("Kritischer Fehler")
```

In diesem Beispiel:

BasicConfig configura el sistema de registro en el nivel raíz con el nivel especificado. **level=logging.DEBUG** hace que todos los mensajes estén registrados. Varios mensajes se registran con diferentes grados, o en este caso se transmiten.

Para el siguiente ejemplo, cree un archivo **test_log.py** e inserte el siguiente contenido:

```python
import os
import logging
from pathlib import Path

# os.chdir(Path(__file__).parent)

# Erstellen unseres loggers

logger = logging.getLogger()  # root logger

logger.setLevel(logging.DEBUG)

formatter = logging.Formatter('%(name)s - %(levelname)s - %(asctime)s - %(message)s')  # Formatierung

# Filehandler:
file_handler = logging.FileHandler('app.log')
file_handler.setLevel(logging.DEBUG)
file_handler.setFormatter(formatter)

# Streamhandler (wird gebraucht, damit die Ausgabe nicht nur in der Datei, sondern auch im Terminal angezeigt wird)
stream_handler = logging.StreamHandler()
stream_handler.setLevel(logging.WARNING)
stream_handler.setFormatter(formatter)

# Die beiden Handler werden dem Logger hinzugefügt
logger.addHandler(file_handler)
logger.addHandler(stream_handler)


logger.debug("Das ist eine Debug-Information")
logger.info("Das ist eine information!")
logger.warning("Das ist eine Warnung")
logger.error("Ein Fehler ist aufgetreten! Biepub! Alarm! Alle in ihre Rettungskapseln")
logger.critical("Ein Kritischer Fehler ist aufgetreten, der Weg zu den Rettungskapseln lohnt sich auch nicht mehr!")
```

Después de ejecutar el programa, se crea un archivo "**app.log**" para ver el contenido.

En este ejemplo:
- **filename='app.log'** redirecciona los mensajes de registro al archivo app. log.
- **filemode='w'** establece el modo de escritura del archivo. 'w' sobrescribe el archivo cada vez que se inicia el programa.
- **format='%(name)s - %(levelname]s -%(message)s'** define el formato de los mensajes de registro.
- Sólo se registran los mensajes con el nivel INFO y superior.

Instrucciones de registro Utilice el registro en lugar de estados print() para obtener un sistema profesional de procesamiento de errores y registro de información. Para aplicaciones complejas, configure varios loggers con diferentes niveles y destinos de salida. Logging también puede configurarse para la salida en consolas, el envío de correos electrónicos, la escritura en bases de datos, etc.

El módulo de registro es extremadamente potente y adaptable y un componente esencial para cualquier desarrollo serio de Python.

## Registro avanzado

El módulo Python logging puede configurarse con un archivo de configuración, normalmente en formato **INI**. Este método permite separar la configuración de logger de su código y personalizarlo si es necesario sin cambios en el código fuente. Aquí hay un ejemplo de cómo utilizar el módulo de logging con un fichero .ini y un ejemplo práctico de un programa Pythion que utiliza logging.

### Paso 1: Crear un archivo de configuración INI para registro

Cree un archivo llamado logging_config.ini con el siguiente contenido:

```python
import os
import logging
from pathlib import Path
import sys  # Hinzugefügt für sys.stdout beim StreamHandler


# --- Konfigurationsfunktion ---
def setup_logging(log_file_name="app.log",
                  logger_level=logging.DEBUG,
                  file_handler_level=logging.DEBUG,
                  stream_handler_level=logging.INFO,
                  log_format="%(asctime)s - %(levelname)s - %(name)s - %(message)s"):
    """
    Konfiguriert das Logging-System.

    Args:
        log_file_name (str): Der Name der Log-Datei.
        logger_level (int): Das minimale Level für den Logger selbst (z.B. logging.DEBUG).
        file_handler_level (int): Das minimale Level für den FileHandler.
        stream_handler_level (int): Das minimale Level für den StreamHandler (Konsole).
        log_format (str): Das Formatierungs-String für die Log-Einträge.

    Returns:
        logging.Logger: Das konfigurierte Logger-Objekt.
    """
    # Setzt das Arbeitsverzeichnis auf den Ordner, in dem sich diese Datei befindet
    # Vorsicht: Dies ändert das globale Arbeitsverzeichnis. Sicherstellen, dass dies gewünscht ist.
    try:
        os.chdir(Path(__file__).parent)
    except Exception as e:
        print(f"Warnung: Konnte Arbeitsverzeichnis nicht ändern: {e}", file=sys.stderr)

    # Verwende einen spezifischen Logger statt des Root-Loggers für bessere Modularität
    # __name__ verwendet den Modulnamen (z.B. 'test_log'), was oft nützlich ist.
    logger = logging.getLogger(__name__)
    logger.setLevel(logger_level)

    # Verhindere das Hinzufügen von Handlern, falls die Funktion mehrmals aufgerufen wird
    if logger.hasHandlers():
        logger.handlers.clear()

    # Formatierung definieren
    formatter = logging.Formatter(log_format)

    # Konfiguration des Filehandlers
    try:
        file_handler = logging.FileHandler(log_file_name, encoding='utf-8')  # Encoding hinzugefügt
        file_handler.setLevel(file_handler_level)
        file_handler.setFormatter(formatter)
        logger.addHandler(file_handler)
    except Exception as e:
        print(f"Fehler beim Konfigurieren des FileHandlers ({log_file_name}): {e}", file=sys.stderr)
        # Optional: Hier könnte man entscheiden, ob das Programm ohne File-Logging weiterlaufen soll

    # Konfiguration des Streamhandlers (Konsole)
    try:
        stream_handler = logging.StreamHandler(sys.stdout)  # Explizit stdout verwenden
        stream_handler.setLevel(stream_handler_level)
        stream_handler.setFormatter(formatter)
        logger.addHandler(stream_handler)
    except Exception as e:
        print(f"Fehler beim Konfigurieren des StreamHandlers: {e}", file=sys.stderr)
        # Optional: Hier könnte man entscheiden, ob das Programm ohne Konsolen-Logging weiterlaufen soll

    # Wichtig: Propagation verhindern, wenn man NICHT möchte, dass auch der Root-Logger (falls konfiguriert) die Meldungen erhält.
    # logger.propagate = False # Oft nützlich in größeren Anwendungen, hier erstmal auskommentiert

    return logger


# --- Logging-Funktion ---
def log_message(logger: logging.Logger, level: int, message: str):
    """
    Loggt eine Nachricht mit dem angegebenen Level über den übergebenen Logger.

    Args:
        logger (logging.Logger): Das Logger-Objekt, das zum Loggen verwendet werden soll.
        level (int): Das Logging-Level (z.B. logging.DEBUG, logging.INFO, logging.WARNING, logging.ERROR, logging.CRITICAL).
        message (str): Die zu loggende Nachricht.
    """
    if logger:
        logger.log(level, message)
    else:
        # Fallback, falls kein Logger übergeben wurde (sollte nicht passieren bei korrekter Nutzung)
        print(f"FEHLER: Kein Logger zum Loggen der Nachricht vorhanden: {logging.getLevelName(level)} - {message}",
              file=sys.stderr)


# --- Hauptteil / Beispielverwendung ---
if __name__ == "__main__":
    # 1. Logging einmalig konfigurieren und Logger-Objekt erhalten
    #    Hier werden die Standardwerte aus der Funktion verwendet (app.log, DEBUG, DEBUG, INFO)
    main_logger = setup_logging()

    # Alternative Konfiguration (Beispiel):
    # main_logger = setup_logging(log_file_name="mein_anderes_log.txt",
    #                             logger_level=logging.INFO,
    #                             file_handler_level=logging.WARNING,
    #                             stream_handler_level=logging.INFO)

    # 2. Die log_message Funktion verwenden, um Nachrichten zu loggen
    log_message(main_logger, logging.DEBUG, "Das ist eine Debug-Information (via Funktion)")
    log_message(main_logger, logging.INFO, "Das ist eine Information! (via Funktion)")
    log_message(main_logger, logging.WARNING, "Das ist eine Warnung (via Funktion)")
    log_message(main_logger, logging.ERROR, "Ein Fehler ist aufgetreten! (via Funktion)")
    log_message(main_logger, logging.CRITICAL, "Ein Kritischer Fehler ist aufgetreten (via Funktion)")

    # Man kann natürlich auch weiterhin direkt die Methoden des Loggers verwenden:
    main_logger.info("Dies ist eine weitere Info-Nachricht direkt vom Logger.")


    # Beispiel für Logging in einem anderen Teil des Codes:
    def irgendeine_funktion(logger_param):
        log_message(logger_param, logging.INFO, "Nachricht aus irgendeine_funktion")
        # oder direkt:
        # logger_param.info("Direkte Nachricht aus irgendeine_funktion")


    irgendeine_funktion(main_logger)
```

Imagine un programa que recupera datos de una API. Puede utilizar el registro para registrar el estado de las solicitudes y los posibles errores.

Ein fiktives Beispiel:

```python
import requests
import json
import logging.config

logging.config.fileConfig("logging_config.ini")
logger = logging.getLogger("__name__")


def fetch_data(url: str):
    try:
        response = requests.get(url)
        response.raise_for_status()
        logger.info("Daten erfolgreich von %s abgerufen", url)
        data = response.json()

        # mostrar el contenido
        print(json.dumps(data, indent=4, ensure_ascii=False))
        logger.debug(f"Empfangende Daten: {json.dumps(data)}")

        return data

    except requests.exceptions.RequestException as e:
        logger.error("Fehler beim Abrufen der Daten: %s", e)
        return {}

fetch_data("https://httpbin.org/get")
```

## Logging

En Python, puede rotar los archivos de registro usando el módulo **logging.handlers**, especialmente el RotatingFileHandler.Este manejador le permite rotar archivos de log en un tamaño específico** y mantener un número máximo de archivos de copia de seguridad. Esto le permite asegurarse de que el espacio en disco no sea cargado innecesariamente por archivos de registros.

Aquí hay un ejemplo sencillo de cómo utilizar el RotatingFileHandler:

```python
from logging.handlers import RotatingFileHandler

# Logger erstellen
logger = logging.getLogger('MeinLogger')
logger.setLevel(logging.DEBUG)

# RotatingFileHandler erstellen
handler = RotatingFileHandler('mein_log.log', maxBytes=10000, backupCount=5, delay=True)

# Formatter erstellen und dem Handler hinzufügen
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Handler zum Logger hinzufügen
logger.addHandler(handler)

# Testnachrichten loggen
for i in range(1000):
    logger.debug('Dies ist eine Debug-Nachricht %d', i)
```
