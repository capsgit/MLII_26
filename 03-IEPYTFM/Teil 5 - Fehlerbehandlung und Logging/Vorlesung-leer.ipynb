### Einfache Fehlerbehandlung
Der einfachste Weg Fehler abzufangen, ist ein einfacher Try:/Except-Block. In dem Folgenden Beispiel wird die Eingabe einer Ganzzahl erwartet. Gibt der Benutzer einen Buchstaben ein, würde das zu einem Fehler führen, und das Programm wird beendet:<br><br>

```python
while True:
    eingabe = int(input("Bitte gib eine Zahl ein: "))
    print(eingabe)
```

Mit Hilfe eines Try:/Except:-Blocks kann der Programmierer den Fehler abfangen, und somit die Unterbrechung des Programms verhindern:
Einfaches Beispiel, aber nicht nach PEP8-Standard, aber dazu kommen wir später.

```python
from sys import exit

while True:
    eingabe = input("Bitte gib eine Zahl ein (e für exit): ")
    if eingabe == "e":
        exit(0)

    try:
        eingabe = int(eingabe)
        print(eingabe ** 2)

    except ValueError:
        print("Bitte gib eine Zahl ein.")
```

### Mehrere Fehler abfangen
Je nach Fehler lassen sich verschiedene Fehlermeldungen aufrufen, bzw. der Programmierer kann unterscheiden, was bei welcher Art Fehlermeldung passieren soll. Das größte Problem stellen immer logische Fehler dar. Diese Fehler können wir nicht abfangen, und diese sind in der Regel schwer zu finden.<br><br>

```python
while True:
    zahl = input("Bitte gib eine Zahl ein (e für exit): ")
    teiler = int("bitte geben Sie einen Teiler ein")

    if zahl == "e":
        exit(0)

    try:
        zahl = int(zahl)
        teiler = int(teiler)

        ergebnis = zahl / teiler
        print(f"Das Ergebnis von {zahl} / {teiler} ist {ergebnis}.")

    except ValueError:
        print("Bitte nur Zahlen eingeben!")
        exit(1)

    except ZeroDivisionError:
        print("Der Teiler darf nicht 0 sein!")
        exit(2)
```


Beispiel **Division durch 0**. Divisionen durch 0 können natürlich nicht durchgeführt werden. Das Programm würde mit einer Fehlermeldung aussteigen.


Im folgenden recht realen Beispiel werden 3 verschiedene Fehlermeldungen abgefangen:

```python
while True:

    try:
        zahl = input("Bitte geben Sie eine Zahl ein (e für Ende): ")
        teiler = input("Bitte geben Sie einen Teiler ein: ")

    except KeyboardInterrupt:
        print("Abgebrochen!")
        exit(0)

    if zahl == "e":
        exit(0)

    try:
        zahl = int(zahl)
        teiler = int(teiler)

        ergebnis = zahl / teiler
        print(f"Das Ergebnis von {zahl} / {teiler} ist {ergebnis}.")

    except ValueError:
        print("Bitte nur Zahlen eingeben!")
        exit(1)

    except ZeroDivisionError:
        print("Der Teiler darf nicht 0 sein!")
        exit(2)

    except KeyboardInterrupt:
        print("Abgebrochen!")
        exit(0)
```

#### Fehlermeldung ausgeben und behandeln

```python
while True:

    try:
        zahl = input("Bitte geben Sie eine Zahl ein (e für Ende): ")
        teiler = input("Bitte geben Sie einen Teiler ein: ")

    except KeyboardInterrupt:
        print("Abgebrochen!")
        exit(0)

    if zahl == "e":
        exit(0)

    try:
        zahl = int(zahl)
        teiler = int(teiler)

        ergebnis = zahl / teiler
        print(f"Das Ergebnis von {zahl} / {teiler} ist {ergebnis}.")

    except Exception as e:
        print("Es ist ein Fehler aufgetreten")
        print(e)
        continue
```


### Eigene Fehlermeldungen erstellen
Unter Umständen möchten Sie auf ein eigenes Ereignis reagieren, auf das es sonst keine Fehlerbehandlung gibt.<br>
Im folgenden Beispiel möchten Sie ein eMailformular programmieren. Wenn keine gültige Mailadresse eingetragen wurde, wird ein Fehler ausgelöst. 
Als Erstes erstellen Sie eine Klasse, die nach der Fehlermeldung bezeichnet wird. Diese Klasse muss keinen Inhalt enthalten. (Mehr zum Thema Klassen im Thema OOP)


```python
class InvalidMailError(Exception):
    pass

def mailform(adresse, betreff, inhalt):

    if not "@" in adresse:
        raise InvalidMailError("Sie haben keine gültige eMailadresse eingegeben!")  # raise löst einen Fehler aus

    print("Eine eMail mit dem folgenden Inhalt wurde versendet:!")
    print("Empfänger: ", adresse)
    print("Betreff: ", betreff)
    print("Inhalt: ", inhalt)

mailform("dirk@wbs.de", "Test", "Hallo Welt!")
```

Bei dem Aufruf der Funktion erscheint der gewünschte Fehler: "**InvalidMail**", den wir **jetzt** abfangen können.

```python
class InvalidMailError(Exception):
    pass

def mailform(adresse, betreff, inhalt):

    if not "@" in adresse:
        raise InvalidMailError("Sie haben keine gültige eMailadresse eingegeben!")  # raise löst einen Fehler aus

    print("Eine eMail mit dem folgenden Inhalt wurde versendet:!")
    print("Empfänger: ", adresse)
    print("Betreff: ", betreff)
    print("Inhalt: ", inhalt)

try:
    mailform("dirkwbs.de", "Test", "Hallo Welt!")

except InvalidMailError:
    print("Es ist ein Fehler aufgetreten!")
    exit(1)
```

#### Exkurs Regex (Reguläre Ausdrücke)

```python
import re

EMAIL_REGEX = re.compile(
    r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
)

class InvalidMailError(Exception):
    pass


def mailform(adresse, betreff, inhalt):
    if not EMAIL_REGEX.match(adresse):
        raise InvalidMailError("Sie haben keine gültige eMailadresse eingegeben!")  # raise löst einen Fehler aus

    print("Eine eMail mit dem folgenden Inhalt wurde versendet:!")
    print("Empfänger: ", adresse)
    print("Betreff: ", betreff)
    print("Inhalt: ", inhalt)


try:
    mailform("dirk@wbs.de", "Test", "Hallo Welt!")
except InvalidMailError as e:
    print(e)
```

## Error-Handling Hirarchie

Die Fehler-Hierarchie in Python basiert auf der Klasse BaseException, von der alle anderen Ausnahme- und Fehlerklassen abgeleitet sind. Hier sind einige wichtige Punkte zu dieser Hierarchie:

- **BaseException**: An der Spitze der Hierarchie steht die Klasse BaseException. Sie ist die Basisklasse für alle anderen Ausnahmeklassen. Normalerweise sollten Sie Ausnahmen von Exception ableiten, nicht direkt von BaseException.

- **Exception**: Diese Klasse ist ein direkter Nachkomme von BaseException und die Superklasse für die meisten standardmäßigen Ausnahme-Klassen. Sie sollten Exception erben, wenn Sie eigene benutzerdefinierte Ausnahmeklassen erstellen.

- **Standard Fehlerklassen**: Unter Exception gibt es viele vordefinierte Fehlerklassen, die verschiedene Arten von Fehlern repräsentieren:
- **ArithmeticError**: Basisklasse für Fehler bei arithmetischen Operationen wie **OverflowError**, **ZeroDivisionError**.
- **BufferError**: Fehler beim Zugriff auf Buffer-bezogene Operationen.
- **LookupError**: Basisklasse für Fehler bei Lookup-Operationen, wie IndexError (falscher Index in Sequenzen) und KeyError (fehlender Schlüssel in einem Wörterbuch).
- **EnvironmentError**: Basisklasse für Fehler, die außerhalb des Python-Interpreters auftreten (in Python 3 durch **OSError** ersetzt).
- **OSError**: Fehler im Zusammenhang mit dem Betriebssystem, wie Datei nicht gefunden (**FileNotFoundError**) oder Zugriffsverletzung (**PermissionError**).

- **Eigene Ausnahmeklassen**: Sie können Ihre eigenen Ausnahmeklassen erstellen, indem Sie von Exception oder einer ihrer Unterklassen erben. Dies ist nützlich, um spezifische Fehlerbedingungen in Ihrer Anwendung darzustellen.

**Hierarchie und Vererbung**: Das Prinzip der Vererbung in der Ausnahmehierarchie bedeutet, dass Sie mit einem except-Block eine Gruppe von verwandten Fehlern abfangen können, indem Sie die Basisklasse behandeln. Zum Beispiel würde except Exception: alle Nicht-System-bezogenen Fehler abfangen.

**System-Exit-Ausnahmen**: **SystemExit**, **KeyboardInterrupt** und **GeneratorExit** sind spezielle Ausnahmen, die von **BaseException** und nicht von Exception abgeleitet sind. Diese sind für System-bezogene Ereignisse gedacht und sollten normalerweise nicht in einem typischen Fehlerbehandlungscode gefangen werden.

Die genaue Kenntnis und Verwendung dieser Hierarchie ermöglicht es Ihnen, robustere und fehlertolerante Python-Anwendungen zu erstellen, indem Sie gezielt auf spezifische Fehlerarten reagieren und diese entsprechend behandeln

# Logging

**Logging** ist in Python eine standardisierte Methode, um Informationen über den Ablauf eines Programms aufzuzeichnen. Es ist eine wichtige Technik für die Fehlersuche und Überwachung des Verhaltens von Anwendungen. Python bietet mit dem Modul "**logging**" eine flexible Logging-Funktionalität.

## Grundlagen des Python logging-Moduls

1. **Importieren** des Moduls: Um das Logging-System zu verwenden, importieren Sie zunächst das logging-Modul.
2. **Logging-Levels**: Das Modul definiert verschiedene Standard-Logging-Levels, wie DEBUG, INFO, WARNING, ERROR und CRITICAL, die die Schwere oder Bedeutung der Log-Nachrichten angeben.
3. **Logger konfigurieren**: Sie können den Logger konfigurieren, um festzulegen, welche Nachrichten aufgezeichnet werden und wohin sie gesendet werden (z.B. Konsole, Datei).

Ein einfaches Beispiel:

```python
import logging

# Grundkonfiguration des Loggings (Standard ist WARNING und höher)
logging.basicConfig(level=logging.DEBUG)

# Verschiedene Log-Nachrichten
logging.debug("Debug-Information")
logging.info("Allgemeine Info")
logging.warning("Warnung")
logging.error("Ein Fehler ist aufgetreten")
logging.critical("Kritischer Fehler")
```

In diesem Beispiel:

basicConfig konfiguriert das Logging-System auf der Wurzelebene mit dem angegebenen Level. 
**level=logging.DEBUG** bewirkt, dass alle Nachrichten geloggt werden.
Verschiedene Nachrichten werden mit unterschiedlichen Schweregraden geloggt, bzw in diesen Fall ausgegeben.

Für das nächste Beispiel erstellen Sie sich bitte eine Datei **test_log.py** und fügen dort den folgenden Inhalt ein:

```python
import os
import logging
from pathlib import Path

# os.chdir(Path(__file__).parent)

# Erstellen unseres loggers

logger = logging.getLogger()  # root logger

logger.setLevel(logging.DEBUG)

formatter = logging.Formatter('%(name)s - %(levelname)s - %(asctime)s - %(message)s')  # Formatierung

# Filehandler:
file_handler = logging.FileHandler('app.log')
file_handler.setLevel(logging.DEBUG)
file_handler.setFormatter(formatter)

# Streamhandler (wird gebraucht, damit die Ausgabe nicht nur in der Datei, sondern auch im Terminal angezeigt wird)
stream_handler = logging.StreamHandler()
stream_handler.setLevel(logging.WARNING)
stream_handler.setFormatter(formatter)

# Die beiden Handler werden dem Logger hinzugefügt
logger.addHandler(file_handler)
logger.addHandler(stream_handler)


logger.debug("Das ist eine Debug-Information")
logger.info("Das ist eine information!")
logger.warning("Das ist eine Warnung")
logger.error("Ein Fehler ist aufgetreten! Biepub! Alarm! Alle in ihre Rettungskapseln")
logger.critical("Ein Kritischer Fehler ist aufgetreten, der Weg zu den Rettungskapseln lohnt sich auch nicht mehr!")
```

Nach der Ausführung des Programms wird eine Datei "**app.log**" erstellt, schauen Sie sich den Inhalt an.

In diesem Beispiel:
- **filename='app.log'** leitet die Log-Nachrichten in die Datei app.log um.
- **filemode='w'** setzt den Schreibmodus der Datei. 'w' überschreibt die Datei bei jedem Start des Programms.
- **format='%(name)s - %(levelname)s - %(message)s'** definiert das Format der Log-Nachrichten.
- Nur Nachrichten mit dem Level INFO und höher werden geloggt.

Hinweise zum Logging
Verwenden Sie das Logging anstelle von print()-Statements, um ein professionelles Fehlerbehandlungs- und Informationsprotokollsystem zu erhalten.
Richten Sie für komplexe Anwendungen mehrere Logger mit unterschiedlichen Levels und Ausgabezielen ein.
Logging kann auch für die Ausgabe in Konsolen, das Senden von E-Mails, das Schreiben in Datenbanken usw. konfiguriert werden.

Das logging-Modul ist äußerst mächtig und anpassbar und eine wesentliche Komponente für jede ernsthafte Python-Entwicklung.

## Logging für Fortgeschrittene

Das Python-logging-Modul kann mit einer Konfigurationsdatei, typischerweise im **INI-Format**, konfiguriert werden. Diese Methode ermöglicht es, die Logger-Konfiguration von Ihrem Code zu trennen und sie bei Bedarf ohne Änderungen am Quellcode anzupassen. Hier ist ein Beispiel, wie Sie das logging-Modul mit einer .ini-Konfigurationsdatei verwenden können, und ein praxisnahes Beispiel für ein Python-Programm, das Logging verwendet.

### Schritt 1: Erstellen Sie eine INI-Konfigurationsdatei für Logging

Erstellen Sie eine Datei namens logging_config.ini mit dem folgenden Inhalt:

```python
import os
import logging
from pathlib import Path
import sys  # Hinzugefügt für sys.stdout beim StreamHandler


# --- Konfigurationsfunktion ---
def setup_logging(log_file_name="app.log",
                  logger_level=logging.DEBUG,
                  file_handler_level=logging.DEBUG,
                  stream_handler_level=logging.INFO,
                  log_format="%(asctime)s - %(levelname)s - %(name)s - %(message)s"):
    """
    Konfiguriert das Logging-System.

    Args:
        log_file_name (str): Der Name der Log-Datei.
        logger_level (int): Das minimale Level für den Logger selbst (z.B. logging.DEBUG).
        file_handler_level (int): Das minimale Level für den FileHandler.
        stream_handler_level (int): Das minimale Level für den StreamHandler (Konsole).
        log_format (str): Das Formatierungs-String für die Log-Einträge.

    Returns:
        logging.Logger: Das konfigurierte Logger-Objekt.
    """
    # Setzt das Arbeitsverzeichnis auf den Ordner, in dem sich diese Datei befindet
    # Vorsicht: Dies ändert das globale Arbeitsverzeichnis. Sicherstellen, dass dies gewünscht ist.
    try:
        os.chdir(Path(__file__).parent)
    except Exception as e:
        print(f"Warnung: Konnte Arbeitsverzeichnis nicht ändern: {e}", file=sys.stderr)

    # Verwende einen spezifischen Logger statt des Root-Loggers für bessere Modularität
    # __name__ verwendet den Modulnamen (z.B. 'test_log'), was oft nützlich ist.
    logger = logging.getLogger(__name__)
    logger.setLevel(logger_level)

    # Verhindere das Hinzufügen von Handlern, falls die Funktion mehrmals aufgerufen wird
    if logger.hasHandlers():
        logger.handlers.clear()

    # Formatierung definieren
    formatter = logging.Formatter(log_format)

    # Konfiguration des Filehandlers
    try:
        file_handler = logging.FileHandler(log_file_name, encoding='utf-8')  # Encoding hinzugefügt
        file_handler.setLevel(file_handler_level)
        file_handler.setFormatter(formatter)
        logger.addHandler(file_handler)
    except Exception as e:
        print(f"Fehler beim Konfigurieren des FileHandlers ({log_file_name}): {e}", file=sys.stderr)
        # Optional: Hier könnte man entscheiden, ob das Programm ohne File-Logging weiterlaufen soll

    # Konfiguration des Streamhandlers (Konsole)
    try:
        stream_handler = logging.StreamHandler(sys.stdout)  # Explizit stdout verwenden
        stream_handler.setLevel(stream_handler_level)
        stream_handler.setFormatter(formatter)
        logger.addHandler(stream_handler)
    except Exception as e:
        print(f"Fehler beim Konfigurieren des StreamHandlers: {e}", file=sys.stderr)
        # Optional: Hier könnte man entscheiden, ob das Programm ohne Konsolen-Logging weiterlaufen soll

    # Wichtig: Propagation verhindern, wenn man NICHT möchte, dass auch der Root-Logger (falls konfiguriert) die Meldungen erhält.
    # logger.propagate = False # Oft nützlich in größeren Anwendungen, hier erstmal auskommentiert

    return logger


# --- Logging-Funktion ---
def log_message(logger: logging.Logger, level: int, message: str):
    """
    Loggt eine Nachricht mit dem angegebenen Level über den übergebenen Logger.

    Args:
        logger (logging.Logger): Das Logger-Objekt, das zum Loggen verwendet werden soll.
        level (int): Das Logging-Level (z.B. logging.DEBUG, logging.INFO, logging.WARNING, logging.ERROR, logging.CRITICAL).
        message (str): Die zu loggende Nachricht.
    """
    if logger:
        logger.log(level, message)
    else:
        # Fallback, falls kein Logger übergeben wurde (sollte nicht passieren bei korrekter Nutzung)
        print(f"FEHLER: Kein Logger zum Loggen der Nachricht vorhanden: {logging.getLevelName(level)} - {message}",
              file=sys.stderr)


# --- Hauptteil / Beispielverwendung ---
if __name__ == "__main__":
    # 1. Logging einmalig konfigurieren und Logger-Objekt erhalten
    #    Hier werden die Standardwerte aus der Funktion verwendet (app.log, DEBUG, DEBUG, INFO)
    main_logger = setup_logging()

    # Alternative Konfiguration (Beispiel):
    # main_logger = setup_logging(log_file_name="mein_anderes_log.txt",
    #                             logger_level=logging.INFO,
    #                             file_handler_level=logging.WARNING,
    #                             stream_handler_level=logging.INFO)

    # 2. Die log_message Funktion verwenden, um Nachrichten zu loggen
    log_message(main_logger, logging.DEBUG, "Das ist eine Debug-Information (via Funktion)")
    log_message(main_logger, logging.INFO, "Das ist eine Information! (via Funktion)")
    log_message(main_logger, logging.WARNING, "Das ist eine Warnung (via Funktion)")
    log_message(main_logger, logging.ERROR, "Ein Fehler ist aufgetreten! (via Funktion)")
    log_message(main_logger, logging.CRITICAL, "Ein Kritischer Fehler ist aufgetreten (via Funktion)")

    # Man kann natürlich auch weiterhin direkt die Methoden des Loggers verwenden:
    main_logger.info("Dies ist eine weitere Info-Nachricht direkt vom Logger.")


    # Beispiel für Logging in einem anderen Teil des Codes:
    def irgendeine_funktion(logger_param):
        log_message(logger_param, logging.INFO, "Nachricht aus irgendeine_funktion")
        # oder direkt:
        # logger_param.info("Direkte Nachricht aus irgendeine_funktion")


    irgendeine_funktion(main_logger)
```

### Praxisbeispiel
Stellen Sie sich ein Programm vor, das Daten von einer API abruft. Sie könnten das Logging verwenden, um den Status der Anfragen und mögliche Fehler zu protokollieren.

Ein fiktives Beispiel:

```python
import requests
import json
import logging.config

logging.config.fileConfig("logging_config.ini")
logger = logging.getLogger("__name__")


def fetch_data(url: str):
    try:
        response = requests.get(url)
        response.raise_for_status()
        logger.info("Daten erfolgreich von %s abgerufen", url)
        data = response.json()

        # mostrar el contenido
        print(json.dumps(data, indent=4, ensure_ascii=False))
        logger.debug(f"Empfangende Daten: {json.dumps(data)}")

        return data

    except requests.exceptions.RequestException as e:
        logger.error("Fehler beim Abrufen der Daten: %s", e)
        return {}

fetch_data("https://httpbin.org/get")
```

## Logging

In Python können Sie die Logdateien rotieren, indem Sie das **logging.handlers-Modul** verwenden, speziell den RotatingFileHandler.Dieser Handler ermöglicht es Ihnen, Logdateien nach einer bestimmten Größe zu **rotieren** und eine maximale Anzahl von Backup-Dateien zu behalten.
Auf diese Weise kann man sicher gehen, dass der Speicherplatz nicht unnötig von Logdateien belastet wird.

Hier ist ein einfaches Beispiel, wie Sie den RotatingFileHandler einsetzen können:

```python
from logging.handlers import RotatingFileHandler

# Logger erstellen
logger = logging.getLogger('MeinLogger')
logger.setLevel(logging.DEBUG)

# RotatingFileHandler erstellen
handler = RotatingFileHandler('mein_log.log', maxBytes=10000, backupCount=5, delay=True)

# Formatter erstellen und dem Handler hinzufügen
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Handler zum Logger hinzufügen
logger.addHandler(handler)

# Testnachrichten loggen
for i in range(1000):
    logger.debug('Dies ist eine Debug-Nachricht %d', i)
```
