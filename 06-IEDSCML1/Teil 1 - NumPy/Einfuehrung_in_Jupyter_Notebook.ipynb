## Was ist Jupyter Notebook?

Jupyter Notebook ist eine interaktive Entwicklungsumgebung im Browser.

Man kann Code ausführen, Text schreiben und Ergebnisse direkt unter dem Code sehen.


## Was ist ein Kernel?

Der Kernel ist der Ausführungsprozess im Hintergrund.
Er führt Code aus und speichert Variablen im Speicher.


```python
x = 10
y = 5
x + y
```

## Abhängigkeit von Zellen

In Jupyter hängt der Zustand vom Kernel ab.
Variablen, Imports und Funktionen existieren nur,
wenn die entsprechende Zelle vorher ausgeführt wurde.

Das Notebook ist kein lineares Skript.
Die Ausführungsreihenfolge bestimmt den Zustand.


### Beispiel mit Import

Wenn diese Zelle nicht ausgeführt wurde,
kennt der Kernel numpy nicht.


```python
import numpy as np
```

```python
x = np.array([1, 2, 3, 4])
x
```

Wenn man den Kernel neu startet und direkt die obere Zelle ausführt,
entsteht ein Fehler, weil numpy noch nicht importiert wurde.

Typischer Fehler:
NameError: name 'np' is not defined


### Wichtige Erkenntnis

Ein Notebook speichert keine Logik der Reihenfolge.
Es speichert nur Zellen.

Saubere Arbeitsweise:
Kernel neu starten → Run All

Nur wenn das ohne Fehler läuft,
ist das Notebook konsistent.
