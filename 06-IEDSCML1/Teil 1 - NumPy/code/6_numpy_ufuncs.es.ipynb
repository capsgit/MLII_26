# Numpy uFuncs - Universelle Funktionen

En las matrices de Numpy, las funciones universales de Numply se pueden realizar de forma elemental operaciones a alta velocidad. La razón es que las operaciones son vectorizadas, es decir, que se realizan en paralelo y especializada por la CPU (SIMD, Single Instruction, Multiple Data).

## Simple arithmetische Operatoren
Los operadores aritméticas están implementados en Numpy como uFunctions y operan en todos los elementos.

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([1, 1])
b = np.array([2, 3])
print("inicio   a:", a)
print("inicio   b:", b)
print("-----------------------------")

# Basic operations
c = np.add(a, b)
print("adicion:", c)
print("-----------------------------")

# oder mit Operatoren
c = a + b
d = a * b
e = a / b
print("adicion 2.0:", c)
print("multiplicacion:", d)
print("division:", e)

a = np.arange(1, 10).reshape((3,3))
b = np.arange(1, 10).reshape((3,3))
print("-----------------------------")

print("a: \n", a)
print("b: \n", b)
print("-----------------------------")

result = a + b
print("resultado: \n", result)
print("-----------------------------")
```

## Skalarmultiplikation

Cada entrada del vector se multiplica por el escalador (en este caso 5).

```python
z = np.array([1, 2, 3])
y = np.array([[1, 2, 3], [3, 4, 5]])
# Multiplizieren aller Werte von z und y mit dem Skalar 5

z_resultado_por_5 = z * 5
print("*5: \n", z ,z_resultado_por_5)
print(" ")

z_resultado_dividido_5 = z / 5
print("/5: \n", z, z_resultado_dividido_5)
print("-----------------------------")

y_resultado_por_5 = y * 5
print("*5: \n", y, "\n", "-------", "\n", y_resultado_por_5)
print(" ")
y_resultado_dividido_5 = y / 5
print("/5: \n", y, "\n", "-------", "\n", y_resultado_dividido_5)
```

## Matrixmultiplikation

### Regla de multiplicación de matriz

La multiplicación de la matriz se define cuando el **número de columnas de la primera matriz** es igual al ** número de filas de la segunda matriz**.

#### Mathematische Regel:
Si $ A $ es una matriz de la dimensión $ m \times n $ y $ B $ una matriz del tamaño $ n \Times p $, entonces el producto $ C = A \cdot B  dólares es una matriz de la dimension $m \times P $.

#### Berechnung eines Eintrags:
La entrada $ c_{ij} $ en la línea $i$-ten y $j$-tente columna de la matriz $ C $ se calcula como: $c_{j} = \sum_{k=1}^n a_{ik} \cdot b_{ kj}  USD El **Skalarprodukt** es la línea de $I$-ésimo de  dólares con la columna $J$-ter de  USDB$.

---

### Beispiel:
Sei $ A $ eine $2 \times 3$-Matrix:
$$
A = \begin{pmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{pmatrix}
$$

y $ B $ una matriz de $3 \times 2$:

$$
B = \begin{pmatrix} 7 & 8 \\ 9 & 10 \\ 11 & 12 \end{pmatrix}
$$


Entonces el producto $ C = A \cdot B $ es una matriz de $2 \times 2$: $C = \begin{pmatrix} (1\c dot7 + 2\cdocdot9 + 3\cDOt11) & (1\ctot8 + 2 \ cdot10 + 3 \endot12) \\ (4\cdort7+ 5\cdót9+ 6\cdot11) & (4\coddot8+ 5 \ddot10 + 6\ctdot12), \end{pmatrix} $

$$
C = \begin{pmatrix} 58 & 64 \\ 139 & 154 \end{pmatrix}
$$

---

### Hinweise:
1. La multiplicación de matriz es **no commutativo**: $ A \cdot B \neq B  \ctot A $ (normalmente).
2. La multiplicación de matriz es **asociativa**: $ (A \cdot B) \ cdot C = A \ccdot (B \ctot C) $.
3. La multiplicación de matriz es **distributiva**: $ A \cdot (B + C) = A  \ctot B + A •cdoto C $.

$ A = \left( \begin{matrix} 2 & 3 & 4 \\ 4 & 5 & 1 \end{matrix} \right) $
$ B = \left( \begin{matrix} 1 & 2 \\ 3 & 4 \\ 2 & 4 \end{matrix} \right) $
$ C = \left( \begin{matrix} 19 & 32 \\ 21 & 32 \end{matrix} \right) $

```python
A = np.array([[2, 3, 4], [4, 5, 1]])
B = np.array([[1, 2], [3, 4], [2, 4]])
C = A @ B
print(A)
print("--")
print(B)
print("--")
print("C: \n", C)
print("-------")
D = B @ A
print("D (also c al revez): \n", D)
```

## Arithmetische Vergleichsfunktionen

Las funciones de comparación aritmética en NumPy permiten realizar comparaciones elementales entre arrays o entre un array y un escalar. Los ejemplos son más grandes** \((>)\), **pequeño igual** {(\leq)\)\, o **igual** ·(==)\ que devuelven cada uno un arrecife booleano. Además de las funciones como el póster P0 o el p1 también se puede utilizar directamente la ortografía de los operadores**, por ejemplo, el p2 o el P3 , ya que Numpy sobrecarga y ejecuta estos operadores de forma equivalente.

```python
A = np.array([[2, 6], [4, 8]])
B = np.array([[1, 9], [3, 16]])

A < B
```

```python
A = np.array([[2, 6], [4, 8]])
B = np.array([[1, 9], [3, 16]])

A >= B, A != B
```
