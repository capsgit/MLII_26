# Broadcasting
El broadcasting en NumPy permite operaciones aritméticas entre matrices de diferentes formas adaptando automáticamente las matrices más pequeñas a las más grandes. Una dimensión de 1 se "estirará" para permitir la operación. Esto ahorra tiempo y memoria de Numpy sin crear copias innecesarias de los datos.

![title](broadcasting_1.png)

```python
"""import numpy as np"""
```

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([2])

result = a * b

x = a * 2
print("a * 2 = ", x)
print("a * b = ",result)
```

## 1) Agregar matrices con diferentes shape
Los arrays se pueden agregar entre sí si tienen a) el mismo shape o b) es uno de los shapes 1.

## 2) Skalar-Multiplikation
Escalar a Array-Broadcasting: El escalador se amplía a toda la forma del array.

$a = 3$
