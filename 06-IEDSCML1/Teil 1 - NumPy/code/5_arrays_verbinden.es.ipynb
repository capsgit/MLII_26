## Concatenate de matrices monodimensionales y multidimensionales
El método P0 permite conectar matrices de una y de varias dimensiones a un array. Aquí es importante entender el término de eje (axis). Axis 0 designa las filas (rows), mientras que axis 1 designa las columnas.

**Generell se puede decir que el shape del nuevo array sólo cambia en el eje en el que se unen los arrays**.

**Sólo se pueden contagiar los mismos shapes (igual dimensión) de matrices en el eje de no-concatenación**


- axis=0 (Zeilenweise Konkatenation) → Spaltenzahl muss gleich sein.       
- axis=1 (Spaltenweise Konkatenation) → Zeilenzahl muss gleich sein. 

![title](numpy-arrays-have-axes.webp)

## Concatenate von zwei zweidimensionalen Arrays an Axis 0
Cuando realizamos concatenate en matrices con axis=0, los arrays se agrupan "entrelazados," lo que significa que el shape de las columnas (columnas) sigue siendo el mismo, pero cambia de línea. Por lo tanto, los arrays se "apilan" verticalmente. Para ello, el número de columnas en ambas matrices debe coincidir. El shap en el eje 0 ha cambiado, el arnés de columna sigue siendo igual.

![title](explanation_numpy-concatenate-axis-0.webp)

```python
import numpy as np

arr1 = np.array([[1, 1, 1], [1, 1, 1]])
arr2 = np.array([[9, 9, 9], [9, 9, 9], [9, 9, 9]])
arr = np.concatenate((arr1, arr2), axis=0)
print(arr)
```

## Concatenate von zwei zweidimensionalen Arrays an Axis 1
Cuando realizamos concatenate con axis=1, los arrays se unen "alternadamente," lo que da lugar a que el shape de las columnas (columnas) cambie, pero el array de fila permanece igual.

```python
arr1 = np.array([[1, 1, 1], [1, 1, 1]])
arr2 = np.array([[9, 9, 9], [9, 9, 9]])

arr3_y = np.concatenate((arr1, arr2), axis=0)
arr3_x = np.concatenate((arr1, arr2), axis=1)

print("Y / axis 0: ", '\n', arr3_y)
print("base Y: ", arr3_y.base)
print("-----------------------------")
print("X / axis 1: ", '\n', arr3_x)
print("base X: ", arr3_x.base)


# alternativa (vstack)
x_st = np.hstack([arr1, arr2])

# alternativa (vstack)
y_st = np.vstack((arr1, arr2))
print("alternativas: ")
print("-----------------------------")
print("vertical / y_st / axis 0: ", '\n', y_st)
print("-----------------------------")
print("horizontal / x_st / axis 1: ", '\n',x_st)
print("-----------------------------")
```

## Combinar vector y matriz vía vstack en un array multidimensional
Para conectar los arrays con el ABSP0, deben tener las mismas dimensiones. Por lo tanto, no se puede conectar un array de una o varias dimensiones. Para ello, se utiliza el método ABS P1. vstack significa apilamiento vertical.

En el caso de vstack, sólo debe existir que ambas matrices tengan el mismo número de columnas. Por supuesto, también se pueden conectar dos matrices multidimensionales con vstacke.

```python
arr1 = np.array([1, 1, 1])
arr2 = np.array([[9, 9, 9], [9, 9, 9]])

print(np.vstack((arr1, arr2)))
```
