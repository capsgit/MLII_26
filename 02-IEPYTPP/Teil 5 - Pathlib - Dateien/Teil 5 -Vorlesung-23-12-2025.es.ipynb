## Dateien

Es gibt verschiedene Modi, mit denen Dateien geöffnet werden können:<br /><br />
**r** - Die Datei ist ausschließlich zum **Lesen** bestimmt. Existiert diese Datei nicht, wird eine Fehlermeldung generiert <br />
**w** - Die Datei wird zum **Schreiben** geöffnet, bzw. Sollte diese nicht existieren, wird sie angelegt<br />
**a** - Die Datei wird zum **Anhängen** neuer Daten geöffnet. Die alten Daten werden **nicht** überschrieben, sondern es wird am Ende weiter geschrieben.<br />
**rb** - Die Datei wird zum Lesen im Binärmodus geöffnet <br />
**wb** - Die Datei wird zum Schreiben im Binärmodus geöffnet <br /><br />

Al leer archivos, es posible utilizar un bloque Try/Except:

```python
# -*- coding: utf-8 -*-
try:
    datei = open("meinedatei.txt", mode="a", encoding="utf-8")  # Datei wird geöffnet
    text = input("Bitte geben Sie einen klugen Satz ein: ") + "\n"  # ohne "\n" gibt es keinen Zeilenumbruch
    datei.write(text)  # Datei wird geschrieben
    datei.close()  # Datei wird geschlossen

except Exception as e:
    print("ALARM! Es ist ein Fehler aufgetreten:")
    print(e)
```

### La instrucción with (Manager de contextos)

La instrucción with facilita enormemente el manejo de los archivos. No se abre ni cierra. El archivo se abre automáticamente, se edita, etc. y se cierra. •P0 • Este ejemplo muestra cómo crear un archivo, escribir texto en él y leer el contenido.

```python
with open("test.txt", "w", encoding="utf-8") as f:
    f.write("Hallo Welt!\n")
    f.write("Das ist eine weitere Zeile.\n")
```

#### Abrir archivo con el administrador de contexto:

```python
# -*- coding: utf-8 -*-

with open("test.txt", "r", encoding="utf-8") as f:
    print(f.read())
```

#### Archivo de lectura y procesamiento en línea:

Aquí se lee un archivo en línea y cada línea se le asigna un número.

```python
# -*- coding: utf-8 -*-

with open("test.txt", "r", encoding="utf-8") as f:
    for nummer, zeile in enumerate(f, start=1):
        print(nummer, zeile.strip())
```

#### Daten in eine Excel-Datei schreiben

Creamos un registro sencillo y lo escribimos en un nuevo archivo de Excel llamado " P1" con Siempre.

```python
# -*- coding: utf-8 -*-
import pandas as pd

# Beispiel-Daten erstellen (z.B. ein Dictionary)
daten = {
    'Name': ['Anna', 'Bernd', 'Clara'],
    'Alter': [28, 34, 29],
    'Stadt': ['Berlin', 'Hamburg', 'München']
}

# Datenframe erstellen
df = pd.DataFrame(daten)

# Datenframe speichern:
dateiname = "meine_daten.xlsx"

df.to_excel(dateiname, index=False)
```

#### Daten aus einer Excel-Datei lesen

Ahora vamos a volver a leer los datos del archivo de Excel creado previamente (ajustado a P0).

```python
# -*- coding: utf-8 -*-
import pandas as pd

dateiname = "meine_daten.xlsx"

df_gelesen = pd.read_excel(dateiname, header=0)

print(df_gelesen)

# print(df_gelesen["Alter"]) Nur das Alter ausgeben
```

#### Escribir varias hojas en un archivo de Excel

Con la entrada P0 se pueden almacenar varios marcos de datos en un solo archivo utilizando diferentes hojas (la entrada P1 ).

```python
# -*- coding: utf-8 -*-
import pandas as pd

# Beispiel-Daten erstellen (z.B. ein Dictionary)
daten = {
    'Name': ['Anna', 'Bernd', 'Clara'],
    'Alter': [28, 34, 29],
    'Stadt': ['Berlin', 'Hamburg', 'München']
}

daten2 = {
    "Producto": ["Apfel", "Banane", "Kiwi"],
    "Precio": [1.20, 0.50, 2.75],
    "Estado": ["Nuevo", "Usado", "Nuevo"]
}

# Datenframe erstellen
df = pd.DataFrame(daten)
df2 = pd.DataFrame(daten2)

dateiname = "Hojas_varias.xlsx"

with pd.ExcelWriter(dateiname, engine="openpyxl") as writer:
    df2.to_excel(writer, sheet_name="Productos", index=False)
    df.to_excel(writer, sheet_name="Personas", index=False)
```

#### Ein bestimmtes Tabellenblatt aus einer Excel-Datei laden

Se puede cargar de forma específica una hoja de cálculo indicando el argumento "P1" con la referencia "P0" específica.

```python
# -*- coding: utf-8 -*-
import pandas as pd

df_geladen = pd.read_excel("Hojas_varias.xlsx", sheet_name="Personas", engine="openpyxl")

#print(df_geladen)

for sheet, df in df_geladen.items():
    print(f"Blatt: {sheet}")
    print(df, "\n")
```

#### Cargar todas las hojas de un archivo de Excel

Se puede cargar todas las hojas al mismo tiempo con el P0. Se devuelve un diccionario en el que los nombres de las hojas son las llaves.

```python
# -*- coding: utf-8 -*-
import pandas as pd

df_geladen = pd.read_excel("Hojas_varias.xlsx", sheet_name=None, engine="openpyxl")

#print(df_geladen)

for sheet, df in df_geladen.items():
    print(f"Blatt: {sheet}")
    print(df, "\n")
```
