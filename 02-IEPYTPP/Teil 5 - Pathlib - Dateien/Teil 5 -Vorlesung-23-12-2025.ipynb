## Dateien

Es gibt verschiedene Modi, mit denen Dateien geöffnet werden können:<br /><br />
**r** - Die Datei ist ausschließlich zum **Lesen** bestimmt. Existiert diese Datei nicht, wird eine Fehlermeldung generiert <br />
**w** - Die Datei wird zum **Schreiben** geöffnet, bzw. Sollte diese nicht existieren, wird sie angelegt<br />
**a** - Die Datei wird zum **Anhängen** neuer Daten geöffnet. Die alten Daten werden **nicht** überschrieben, sondern es wird am Ende weiter geschrieben.<br />
**rb** - Die Datei wird zum Lesen im Binärmodus geöffnet <br />
**wb** - Die Datei wird zum Schreiben im Binärmodus geöffnet <br /><br />

### Lesen:

Am einfachsten öffnet man eine Datei mit dem folgenden Befehl:

```python
text = open("meinedatei.txt", mode="r", encoding="utf-8")  # Datei wird geöffnet
print(text.read())  # Datei wird gelesen
text.close()  # Datei wird geschlossen
```

Beim Lesen von Dateien bietet sich das an, einen Try/Except-Block zu verwenden:

```python
# -*- coding: utf-8 -*-
try:
    datei = open("meinedatei.txt", mode="a", encoding="utf-8")  # Datei wird geöffnet
    text = input("Bitte geben Sie einen klugen Satz ein: ") + "\n"  # ohne "\n" gibt es keinen Zeilenumbruch
    datei.write(text)  # Datei wird geschrieben
    datei.close()  # Datei wird geschlossen

except Exception as e:
    print("ALARM! Es ist ein Fehler aufgetreten:")
    print(e)
```

Folgende Methoden können auf Dateien angewandt werden, da diese Instanzen der Klasse file sind:


|**Methode**        | **Bedeutung**
|:---------------|:-----------------------------------------------------------------------------------------|
|**close()** | Die Datei wird **geschlossen und gespeichert**.                                          |
|**flush()** | Die Datei wird nicht geschlossen, aber **gespeichert**.                                  |
|**read()**  | Die Datei wird **gelesen** und der Inhalt als **String** zurückgegeben.                  |
|**readline()** | Die **nächste Zeille** wird gelesen und als **String** zurückgegeben.                    |
|**readlines** | Die Datei wird **zeilenweise gelesen** und in einer **Liste aus Strings** zurückgegeben. |
|**seek(position)** | Der Cursor wird auf die entsprechend angegebene Position gesetzt                         |
|**tell()** | Es werden alle Cursorpositionen ausgegeben                                               |
|**write(var)** | Die in **var** enthaltene Zeichenkette wird geschrieben                                  |



```python
# -*- coding: utf-8 -*-
text = open("meinedatei.txt", mode="r", encoding="utf-8")  # Datei wird geöffnet
print(text.readline(), end="")
print(text.readline(), end="")
print(text.readline(), end="")
text.close()  # Datei wird geschlossen
```

```python
# -*- coding: utf-8 -*-
text = open("meinedatei.txt", mode="r", encoding="utf-8")  # Datei wird geöffnet
print(text.readlines())

text.close()  # Datei wird geschlossen
```

<br />In der Tabelle lesen Sie den Begriff "**Cursor**". Dieser ist sozusagen die Position, an der gelesen, oder geschrieben wird.<br /><br />

Außerdem gibt es da noch folgende Attribute:

|**Attribut**        |**Bedeutung**
|:---------------|:---------------------------------|
|**closed** | ist **True** wenn die Datei geschlossen ist|
|**mode** | Zeigt an, ob die Datei zum lesen geöffnet ist (**r**) oder zum Schreiben (**w**)|
<br />

**Noch ein Hinweis:** <br />
Betriebssysteme mögen es nicht sehr, wenn Dateien geöffnet und nicht wieder geschlossen werden. Daher immer daran denken, die Dateien, die geöffnet wurden, auch wieder zu schließen. Je nachdem, welchen Betriebssystem sie nutzen, kann die Anzahl der geöffneten Dateien auch begrenzt sein.<br />
Ein **guter Stil** ist es auch, die Datei **öffnen**, **lesen**, **schließen**. Danach **bearbeiten**, **öffnen**, **speichern**, **schließen**. Dann ist die Datei auch dann geschlossen, sollte etwas bei der Bearbeitung schief gehen, und das Programm abstürzen.<br />
Beim Nutzen der **with**-Anweisung entfällt das Öffnen und schließen, aber dazu **später** mehr.

```python
# -*- coding: utf-8 -*-
text = open("meinedatei.txt", mode="r", encoding="utf-8")  # Datei wird geöffnet
print(text.readlines())

text.close()  # Datei wird geschlossen

print(text.closed)  # Ist die Datei geschlossen?
```

### Die with-Anweisung  (Kontextmanager)

Die with-Anweisung erleichtert dem Umgang mit Dateien enorm. Das Öffnen und Schließen entfällt. Die Datei wird automatisch geöffnet, bearbeitet etc. und geschlossen.
<br />
Dieses Beispiel zeigt, wie man eine Datei erstellt, Text hineinschreibt und dann den Inhalt liest.

```python
with open("test.txt", "w", encoding="utf-8") as f:
    f.write("Hallo Welt!\n")
    f.write("Das ist eine weitere Zeile.\n")
```

#### Datei mit Kontextmanager öffnen:

```python
# -*- coding: utf-8 -*-

with open("test.txt", "r", encoding="utf-8") as f:
    print(f.read())
```

#### Datei zeilenweise Lesen und Verarbeiten:

Hier wird eine Datei zeilenweise gelesen und jede Zeile wird mit einer Nummer versehen.

```python
# -*- coding: utf-8 -*-

with open("test.txt", "r", encoding="utf-8") as f:
    for nummer, zeile in enumerate(f, start=1):
        print(nummer, zeile.strip())
```

#### Daten in eine Excel-Datei schreiben

Wir erstellen einen einfachen Datensatz und schreiben ihn mit `pandas` in eine neue Excel-Datei namens `meine_excel_datei.xlsx`.

```python
# -*- coding: utf-8 -*-
import pandas as pd

# Beispiel-Daten erstellen (z.B. ein Dictionary)
daten = {
    'Name': ['Anna', 'Bernd', 'Clara'],
    'Alter': [28, 34, 29],
    'Stadt': ['Berlin', 'Hamburg', 'München']
}

# Datenframe erstellen
df = pd.DataFrame(daten)

# Datenframe speichern:
dateiname = "meine_daten.xlsx"

df.to_excel(dateiname, index=False)
```

#### Daten aus einer Excel-Datei lesen

Nun lesen wir die Daten aus der zuvor erstellten Excel-Datei (`meine_excel_datei.xlsx`) wieder ein.

```python
# -*- coding: utf-8 -*-
import pandas as pd

dateiname = "meine_daten.xlsx"

df_gelesen = pd.read_excel(dateiname, header=0)

print(df_gelesen)

# print(df_gelesen["Alter"]) Nur das Alter ausgeben
```

#### Mehrere Tabellenblätter in eine Excel-Datei schreiben

Mit `pandas.ExcelWriter` können mehrere DataFrames in einer Datei gespeichert werden, indem verschiedene Blätter (`sheets`) verwendet werden.

```python
# -*- coding: utf-8 -*-
import pandas as pd

# Beispiel-Daten erstellen (z.B. ein Dictionary)
daten = {
    'Name': ['Anna', 'Bernd', 'Clara'],
    'Alter': [28, 34, 29],
    'Stadt': ['Berlin', 'Hamburg', 'München']
}

daten2 = {
    "Producto": ["Apfel", "Banane", "Kiwi"],
    "Precio": [1.20, 0.50, 2.75],
    "Estado": ["Nuevo", "Usado", "Nuevo"]
}

# Datenframe erstellen
df = pd.DataFrame(daten)
df2 = pd.DataFrame(daten2)

dateiname = "Hojas_varias.xlsx"

with pd.ExcelWriter(dateiname, engine="openpyxl") as writer:
    df2.to_excel(writer, sheet_name="Productos", index=False)
    df.to_excel(writer, sheet_name="Personas", index=False)
```

#### Ein bestimmtes Tabellenblatt aus einer Excel-Datei laden

Mit `pd.read_excel` kann man gezielt ein bestimmtes Tabellenblatt laden, indem man das Argument `sheet_name` angibt.

```python
# -*- coding: utf-8 -*-
import pandas as pd

df_geladen = pd.read_excel("Hojas_varias.xlsx", sheet_name="Personas", engine="openpyxl")

#print(df_geladen)

for sheet, df in df_geladen.items():
    print(f"Blatt: {sheet}")
    print(df, "\n")
```

#### Alle Tabellenblätter einer Excel-Datei laden

Mit `sheet_name=None` kann man alle Blätter gleichzeitig laden. Es wird ein Dictionary zurückgegeben, in dem die Blattnamen die Schlüssel sind.

```python
# -*- coding: utf-8 -*-
import pandas as pd

df_geladen = pd.read_excel("Hojas_varias.xlsx", sheet_name=None, engine="openpyxl")

#print(df_geladen)

for sheet, df in df_geladen.items():
    print(f"Blatt: {sheet}")
    print(df, "\n")
```
