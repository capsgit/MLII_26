En los casos más raros, los programas simplemente se ejecutan de arriba a abajo. Es necesario tomar decisiones que modifiquen o controlen el funcionamiento de un programa durante la duración de un mismo. Además, deben realizarse evaluaciones, seleccionarse y evaluarse.

Solche Dinge werden mit Schleifen und Verzweigungen realisiert. <br>

La forma más simple del bucle comprueba si una declaración fue o es incorrecta:

```python
aussage = False
if aussage == True:
    print("Du hast recht!")
if aussage == False:
    print("Du hast unrecht!")
```

Ahora hay varias maneras de revisar esta declaración:

```python
aussage = True
if aussage:
    print("Du hast recht!")
if not aussage:
    print("Du hast unrecht!")
```

Die If-Anweisungen werden in diesem Beispiel von oben nach unten ausgewertet, je nachdem welchen Wert **x** besitzt, wird die "Abzweigung" genommen. <br>

```python
x = 10
if x != 10:
    print(" x ist nicht gleich 10")
else:
    print("x ist gleich 10")
```

### Mehrfache Verzweigung:

```python
x = 4
print("x hat aktuell den Wert:", x)
if x < 5:
    print("x ist kleiner als 5")
if x == 5:
    print("x entspricht haargenau 5")
if x > 5:
    print("x ist größer als 5")

print("Ende!")
```

Otra posibilidad de ramificaciones es el uso de **elif** y **else**:

```python
x = 6
print("x hat aktuell den Wert:", x)
if x < 5:
    print("x ist kleiner als 5")
elif x == 5:
    print("x entspricht haargenau 5")
else:
    print("x ist größer als 5")

print("Ende!")
```


**Nota:** Ni **elif** ni **else** se utilizan correctamente, ya que las ramificaciones **if-** se procesan sucesivamente.

## Schleifen:
El bucle **for** y el bucle **while** son una herramienta importante para el desarrollo del programa, por ejemplo, para realizar acciones múltiples o para realizar evaluaciones multifásicas.

* Se utiliza un bucle **for** para ejecutar un paso del programa a una secuencia conocida de valores al mismo tiempo.
* Se utilizará un bucle **while** cuando, por ejemplo, una entrada del usuario dé lugar a la ejecución de un paso del programa (en su caso varias veces), o no.

Un bucle **for** también se llama **lazo de cuenta**, un bucle **while** **-lazo controlado por condiciones**.

A continuación se muestra un ejemplo de bucle **for**:

```python
for i in 3, 7, 10, 11, 13, 15:
    erg = i * i
    print(i, "*", i, "=", erg)
```

### itarable oder Iteration
Una secuencia de objetos también se puede realizar con un bucle **for**, se llama objetos **iterables** o **iterables**. Estos **itarables** nos encontrarán una y otra vez en Python. Un ejemplo sería una lista ****. Este proceso se llama iterar.

```python
a = 1
b = 2
tiere = [a, b, "Maus", "Huhn", "Pferd", "Nashorn"]

for haustiere in tiere:
    print(haustiere)
```
