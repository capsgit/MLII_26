In den seltesten Fällen laufen Programme einfach von oben nach unten ab. Es müssen während der Laufzeit eines Programmes Entscheidungen getroffen werden, die den Ablauf eines Programms ändern oder steuern.
Weiterhin müssen Bewertungen durchgeführt, Auswahlen getroffen und diese bewertet werden.


Solche Dinge werden mit Schleifen und Verzweigungen realisiert. <br>

Die einfachste Form der Schleife überprüft ob eine Aussage war oder falsch ist:

```python
aussage = False  # True oder False müssen mir GROSSEN Anfangsbuchstaben geschrieben werden!

if aussage == True:
    print("Du hast recht!")

if aussage == False:
    print("Du hast unrecht!")
```

Jetzt ist es so, dass es verschiedene Möglichkeiten gibt, diese Aussage zu überprüfen:

```python
aussage = True  # True oder False müssen mir GROSSEN Anfangsbuchstaben geschrieben werden!

if aussage:
    print("Du hast recht!")

if not aussage:
    print("Du hast unrecht!")
```

Die If-Anweisungen werden in diesem Beispiel von oben nach unten ausgewertet, je nachdem welchen Wert **x** besitzt, wird die "Abzweigung" genommen. <br>

```python
zahl = 11

# entweder
if zahl != 10:
    print("Zahl ist nicht 10")

# oder
else:
    print("Zahl ist 10")
```

## 🔍 Prüfmethoden für Inhalte (`True` oder `False`):

| Methode           | Beschreibung                                       | Beispiel                        |
|-------------------|----------------------------------------------------|---------------------------------|
| `str.isdigit()`   | Nur Ziffern (0–9)                                   | `"123".isdigit()` → `True`     |
| `str.isnumeric()` | Ziffern inkl. z. B. Brüche (z. B. ½)                | `"Ⅻ".isnumeric()` → `True`     |
| `str.isdecimal()` | Nur dezimale Zahlen                                 | `"123".isdecimal()` → `True`   |
| `str.isalpha()`   | Nur Buchstaben (a–z, A–Z, keine Zahlen)             | `"Hallo".isalpha()` → `True`   |
| `str.isalnum()`   | Nur Buchstaben **und/oder** Zahlen (keine Leerz.)   | `"Test123".isalnum()` → `True` |
| `str.isspace()`   | Nur Leerzeichen (Space, Tab, Zeilenumbruch etc.)    | `" \t\n".isspace()` → `True` |
| `str.islower()`   | Nur Kleinbuchstaben (wenn Buchstaben vorhanden)     | `"abc".islower()` → `True`     |
| `str.isupper()`   | Nur Großbuchstaben (wenn Buchstaben vorhanden)      | `"ABC".isupper()` → `True`     |
| `str.istitle()`   | Titel-Schreibweise (z. B. `"Das Ist Ein Titel"`)    | `"Das Ist Gut".istitle()` → `True` |

## Schleifen:
Schleifen sind ein wichtiges Werkzeug für den Programmablauf. Durch diese können beispielsweise Aktionen mehrfach durchgeführt oder Bewertungen mehrstufig durchgeführt werden.
Es muss zwischen 2 Arten von Schleifen unterschieden werden. Die **for**-Schleife und die **while**-Schleife.

* Eine **for**-Schleife wird verwendet, wenn zu immer der gleichen Zeit ein Programmschritt auf eine bekannte Folge von Werten ausgeführt werden soll.
* Eine **while**-Schleife wird verwendet, wenn sich erst durch z.B. eine Eingabe des Benutzers ergibt, ob ein Programmschritt (ggf. mehrfach) ausgeführt werden soll, oder nicht.

Eine **for**-Schleife wird auch **Zählschleife** genannt, eine **while**-Schleife **bedingungsgesteuerte Schleife**.


Nachfolgend ein Beispiel für eine **for**-Schleife:

```python
for i in 3, 6, 9:
    print(i)
```

```python
erg = 0

for i in 3, 6, 9:
    # innerhalb der Schleife
    erg = i + i
    print(erg)

# außerhalb der Schleife
```

### itarable oder Iteration
Eine Abfolge von Objekten kann auch mit einer **for**-Schleife durchlaufen werden können, nennt man **iterierbare Objekte** oder **iterables**. Solche **itarables** werden uns in Python immer wieder begegnen. Ein Beispiel wäre eine **Liste**.
Dieser Vorgang wird iterieren genannt.

```python
for i in ["Auto", "Fahrrad", "Motorrad", "eBike"]:
    print(i)
```

### Schleifen abbrechen mit "break"
Mit **break** haben wir die Möglichkeit eine Schleife vorzeitig zu beenden. Wir können beispielsweise so lange würfeln,
bis wir eine **6** haben, danach dürfen wir einen Spielzug machen.

```python
zahlen = 3, 5, 2, 5, 6, 3, 1, 4

for i in zahlen:
    print(f"Sie haben eine {i} gewürfelt")
    if i == 6:
        print("Bitte machen Sie Ihren Spielzug!")
        break

    else:
        print("Bitte nochmal würfeln!")
```

**Wichtig:** Gerade in umfangreicheren Schleifen und Kontrollstrukturen muss exakt die **Einrückung** der Schleifen und Verzweigungen eingehalten werden, da sonst ein falsches Ergebnis erreicht wird.

### continue
Im Gegensatz zu break wird die Schleife nicht gänzlich verlassen, sondern nur der aktuelle **Schleifendurchlauf** wird unterbrochen. Die Schleife macht beim nächsten Element weiter.

```python
# Ziel: Zahlen von 1 - 10 ausgeben, mit Ausnahme der 5
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

### Unser Passwortgenerator mit break:

```python
# Ihr Code hier!# -*- coding: utf-8 -*-
import string
import random
import pyperclip as pc  # muss vorher installiert werden (pip install pyperclip)
import pyfiglet

# Zeichenpool erstellen:
zeichenpool = string.ascii_lowercase
grossbuchstaben = string.ascii_uppercase
zeichen = string.punctuation
ziffern = string.digits

print(pyfiglet.figlet_format("Passwortgenerator", font="future", width=100))
print("Ihr Passwort enthält bisher nur Kleinbuchstaben...")

auswahl = input("Möchten Sie auch Großbuchstaben verwenden (j/n)? ").lower()
if auswahl == "j":
    zeichenpool += grossbuchstaben  # heißt: zeichenpool = zeichenpool + grossbuchstaben

auswahl = input("Möchten Sie auch Sonderzeichen verwenden (j/n)? ").lower()
if auswahl == "j":
    zeichenpool += zeichen

auswahl = input("Möchten Sie auch Ziffern verwenden (j/n)? ").lower()
if auswahl == "j":
    zeichenpool += ziffern

laenge = int(input("Wie viele Stellen soll ihr Passwort lang sein? "))
while len(zeichenpool) < laenge:
    zeichenpool += zeichenpool

# Der Benutzer bekommt nacheinander 5 Passwörter angezeigt, und kann sich eins aussuchen
count = 0

for i in range(5):
    # Zufällige Zeichen aus dem Zeichenpool in eine Liste speichern und in einem String konvertieren:
    passwort  = "".join(random.sample(zeichenpool, laenge))

    print(passwort)  # Ausgabe PW
    pc.copy(passwort)  # PW wird in Zwischenablage kopiert
    count += 1
    print(f"Das Passwort ({count} von 5) wurde in die Zwischenablage kopiert!")  # yay!
    # print("Das Passwort (" + str(count), "von 5) wurde in die Zwischenablage kopiert!")  # yay!
    auswahl = input("Möchten Sie ein neues Passwort generieren? (j/n)? ").lower()
    if auswahl == "n":
        break

if count == 5:
    print("Maximale Anzahl der Passwörter wurde generiert!")
```

#### Else in for-Schleifen
Der else-Teil einer for-Schleife wird nur ausgeführt, wenn die Schleife vollständig durchläuft, also ohne dass ein break passiert.

Mit break: else wird übersprungen.

Ohne break: else wird ausgeführt.

Dadurch eignet sich for … else perfekt für Fälle wie:
„Suche etwas – und falls ich es nicht finde, mache am Ende etwas anderes.“

**Beispiel 1**: Suche mit for ... else: (ein klassischer Anwendungsfall)
Wir durchsuchen eine Liste nach einem bestimmten Element.
Wenn wir es finden → sofort abbrechen.
Wenn wir die gesamte Liste durchlaufen und nichts gefunden haben → else ausführen.

```python

namen = ["Anna", "Bernd", "Claudia", "Dirk"]
suche = "Hans"

for name in namen:
    if name == suche:
        print("Gefunden!", name)
        break

else:
    # Wird ausgeführt, wenn Schleife NICHT durch break abgebrochen wird.
    print("Name nicht gefunden!")

# Programmende
```

Wichtig:
Der else-Block wird nur ausgeführt, wenn kein break ausgeführt wurde.

**Beispiel 2**: Prüfen, ob eine Zahl Primzahl ist

Auch sehr beliebt: Eine Schleife versucht, einen Teiler zu finden.
Wird einer gefunden → break.
Wird kein Teiler gefunden → else bedeutet: Die Zahl ist prim.

```python
zahl = 29

for i in range(2, int(zahl ** 0.5) + 1):
    if zahl % i == 0:
        print(f"{zahl} ist keine Primzahl (teilbar durch {i})")
        break
else:
    print(f"{zahl} ist eine Primzahl")
```

**Merke**:else an einer for- oder while-Schleife gehört logisch zu der Aussage:

„Falls wir nicht vorher aus der Schleife aussteigen, mache das hier …“

### Die ***while***-Schleife
Die ***while***-Schleife dient wie vorher erwähnt der ***bedingungsgesteuerten*** Schleifenbildung.
In dem folgenden Beispiel lassen wir zufallsgesteuerte Werte so lange addieren, bis der Wert 50 erreicht.

```python
zahl = 0

while zahl > 0:
    print(zahl)
    zahl -= 1
```

Überprüfung des Werts innerhalb der Schleife:

```python
# Ziel: Alle geraden Zahlen bis 50 ausgeben

zahl = 1

while zahl <= 50:
    zahl += 1
    if zahl %2 == 0:
        print(zahl, end=', ')
```

### Unser Spiel mit einer while-Schleife

```python
import random


leben = 5
punkte = 0

while leben > 0:
    random.seed()
    zahl = random.randint(1, 10)

    print(zahl)  # Testausgabe für Debugging

    eingabe = input("Bitte geben Sie eine Zahl zwischen 1 und 10 an: ")

    if zahl != int(eingabe):
        leben -= 1
        print("Leider falsch, Sie haben noch", leben, "Leben")

    else:
        print("Gratulation richtig geraten")
        punkte += 1

print("Spielende! Sie haben insgesamt", punkte, "Punkt(e)")
```

### Do while - Schleife (repeat until)

in vielen Programmiersprachen gibt es eine **do while - Schleife**. Bei Python ist dem **nicht** der Fall. Man könnte eine entsprechende Schleife höchsten auf die folgende Weise simulieren:

```python
import random
random.seed()

zahl = random.randint(1,10)
eingabe = 0
versuch = 0

while True:
    versuch += 1
    eingabe = int(input("Bitte geben Sie eine Zahl zwischen und 1 und 10 ein:"))

    if eingabe == zahl:
        print("Richtig geraten!")
        break

    if eingabe != zahl and versuch >= 5:
        print(f"Leider falsch!")
        break
```

### Match Case (Switch Case)
Mit **match** und **case** haben wir seit Python 3.10 eine sehr komfortable Möglichkeit bekommen eine Eingabe zu bewerten:

```python

lang = input("Welche Programmiersprache wollen Sie lernen? ").lower()

match lang:
    case "javascript":
        print("Sie können Web-Entwickler werden!")

    case "python":
        print("Sie können alles werden!")

    case "php":
        print("Sie können Backend-Entwickler werden!")

    case _:
        print("Die Programmiersprache spielt keine Rolle, sondern, nur die Fähigkeit Probleme zu lösen!")
```
