## Schleifen:
El bucle **for** y el bucle **while** son una herramienta importante para el desarrollo del programa, por ejemplo, para realizar acciones múltiples o para realizar evaluaciones multifásicas.

* Se utiliza un bucle **for** para ejecutar un paso del programa a una secuencia conocida de valores al mismo tiempo.
* Se utilizará un bucle **while** cuando, por ejemplo, una entrada del usuario dé lugar a la ejecución de un paso del programa (en su caso varias veces), o no.

Un bucle **for** también se llama **lazo de cuenta**, un bucle **while** **-lazo controlado por condiciones**.

A continuación se muestra un ejemplo de bucle **for**:

```python
for i in 3, 6, 9:
    print(i)
```

```python
erg = 0

for i in 3, 6, 9:
    # innerhalb der Schleife
    erg = i + i
    print(erg)

# außerhalb der Schleife
```

### Cancelar bucle con "break"
Con **break** tenemos la posibilidad de terminar un bucle antes de tiempo. Por ejemplo, podemos tirar dados hasta que tengamos un **6**, después podemos hacer un movimiento.

```python
zahlen = 3, 5, 2, 5, 6, 3, 1, 4

for i in zahlen:
    print(f"Sie haben eine {i} gewürfelt")
    if i == 6:
        print("Bitte machen Sie Ihren Spielzug!")
        break

    else:
        print("Bitte nochmal würfeln!")
```

**Importante:** Precisamente en ciclos y estructuras de control más amplios, se debe respetar exactamente la **retención** de los rectificados y ramificaciones, ya que de lo contrario se conseguirá un resultado incorrecto.

### continue
A diferencia de break, el bucle no se abandona completamente, sino que sólo se interrumpe el actual **paso del bucle**. El bucle continúa en el siguiente elemento.

```python
# Ziel: Zahlen von 1 - 10 ausgeben, mit Ausnahme der 5
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

### Nuestro generador de contraseñas con break:

```python
# Ihr Code hier!# -*- coding: utf-8 -*-
import string
import random
import pyperclip as pc  # muss vorher installiert werden (pip install pyperclip)
import pyfiglet

# Zeichenpool erstellen:
zeichenpool = string.ascii_lowercase
grossbuchstaben = string.ascii_uppercase
zeichen = string.punctuation
ziffern = string.digits

print(pyfiglet.figlet_format("Passwortgenerator", font="future", width=100))
print("Ihr Passwort enthält bisher nur Kleinbuchstaben...")

auswahl = input("Möchten Sie auch Großbuchstaben verwenden (j/n)? ").lower()
if auswahl == "j":
    zeichenpool += grossbuchstaben  # heißt: zeichenpool = zeichenpool + grossbuchstaben

auswahl = input("Möchten Sie auch Sonderzeichen verwenden (j/n)? ").lower()
if auswahl == "j":
    zeichenpool += zeichen

auswahl = input("Möchten Sie auch Ziffern verwenden (j/n)? ").lower()
if auswahl == "j":
    zeichenpool += ziffern

laenge = int(input("Wie viele Stellen soll ihr Passwort lang sein? "))
while len(zeichenpool) < laenge:
    zeichenpool += zeichenpool

# Der Benutzer bekommt nacheinander 5 Passwörter angezeigt, und kann sich eins aussuchen
count = 0

for i in range(5):
    # Zufällige Zeichen aus dem Zeichenpool in eine Liste speichern und in einem String konvertieren:
    passwort  = "".join(random.sample(zeichenpool, laenge))

    print(passwort)  # Ausgabe PW
    pc.copy(passwort)  # PW wird in Zwischenablage kopiert
    count += 1
    print(f"Das Passwort ({count} von 5) wurde in die Zwischenablage kopiert!")  # yay!
    # print("Das Passwort (" + str(count), "von 5) wurde in die Zwischenablage kopiert!")  # yay!
    auswahl = input("Möchten Sie ein neues Passwort generieren? (j/n)? ").lower()
    if auswahl == "n":
        break

if count == 5:
    print("Maximale Anzahl der Passwörter wurde generiert!")
```

#### Else in for-Schleifen
La parte else de un bucle for sólo se ejecuta cuando el bucle pasa por completo, así que sin que pase un break.

Con break: else se salta.

Sin break: else se ejecuta.

Esto hace que for ... else sea perfecto para casos como: búsqueda un poco • y si no lo encuentro, al final haga algo diferente.

**Ejemplo 1**: Búsqueda con for ... else: (un caso de aplicación clásico) Buscamos una lista de un elemento específico. Si lo encontramos → Cancelar inmediatamente. Si hemos pasado por toda la lista y no hemos encontrado nada → else ejecutar.

```python

namen = ["Anna", "Bernd", "Claudia", "Dirk"]
suche = "Hans"

for name in namen:
    if name == suche:
        print("Gefunden!", name)
        break

else:
    # Wird ausgeführt, wenn Schleife NICHT durch break abgebrochen wird.
    print("Name nicht gefunden!")

# Programmende
```

Importante: El bloque else sólo se ejecuta si no se ejecuta ningún break.

** Ejemplo 2**: Verificar si un número es número primo

También muy popular: Un bucle intenta encontrar un separador. Se encuentra uno → break. No se encuentra ningún separador → else significa: El número es prim.

```python
zahl = 29

for i in range(2, int(zahl ** 0.5) + 1):
    if zahl % i == 0:
        print(f"{zahl} ist keine Primzahl (teilbar durch {i})")
        break
else:
    print(f"{zahl} ist eine Primzahl")
```

### El bucle ***while***
El bucle ***while*** sirve, como se mencionó anteriormente, la formación de rectificado *** controlada por condiciones. En el ejemplo siguiente, sumaremos valores controlados aleatoriamente hasta que el valor alcance 50.

```python
zahl = 0

while zahl > 0:
    print(zahl)
    zahl -= 1
```

### Nuestro juego con un bucle while

```python
import random


leben = 5
punkte = 0

while leben > 0:
    random.seed()
    zahl = random.randint(1, 10)

    print(zahl)  # Testausgabe für Debugging

    eingabe = input("Bitte geben Sie eine Zahl zwischen 1 und 10 an: ")

    if zahl != int(eingabe):
        leben -= 1
        print("Leider falsch, Sie haben noch", leben, "Leben")

    else:
        print("Gratulation richtig geraten")
        punkte += 1

print("Spielende! Sie haben insgesamt", punkte, "Punkt(e)")
```

### Do while - Schleife (repeat until)

en muchos lenguajes de programación hay un bucle **do while**. En Python el ** no es el caso. Se podría simular un bucle correspondiente más alto de la manera siguiente:

```python
import random
random.seed()

zahl = random.randint(1,10)
eingabe = 0
versuch = 0

while True:
    versuch += 1
    eingabe = int(input("Bitte geben Sie eine Zahl zwischen und 1 und 10 ein:"))

    if eingabe == zahl:
        print("Richtig geraten!")
        break

    if eingabe != zahl and versuch >= 5:
        print(f"Leider falsch!")
        break
```

### Match Case (Switch Case)
Con **match** y **case** hemos tenido una manera muy cómoda de evaluar una entrada desde Python 3.10:

```python

lang = input("Welche Programmiersprache wollen Sie lernen? ").lower()

match lang:
    case "javascript":
        print("Sie können Web-Entwickler werden!")

    case "python":
        print("Sie können alles werden!")

    case "php":
        print("Sie können Backend-Entwickler werden!")

    case _:
        print("Die Programmiersprache spielt keine Rolle, sondern, nur die Fähigkeit Probleme zu lösen!")
```
