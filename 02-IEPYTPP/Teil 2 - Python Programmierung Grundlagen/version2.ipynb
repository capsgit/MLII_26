## Aufgabe 11:

Anhalteweg:

Es gibt folgende Faustformeln zur Berechnung von Anhaltewegen:

Reaktionsweg (in Metern) = (Geschwindigkeit (in km/h) geteilt durch 10) mal 3

Bremsweg (in Metern) = (Geschwindigkeit (in km/h) geteilt durch 10) mal (Geschwindigkeit (in km/h) geteilt durch 10)

Anhalteweg (in Metern) = Reaktionsweg plus Bremsweg

Eingabe: km/h (via input-funktion)

Especifique la ruta de reacción, la distancia de frenado y la distancia parada correspondiente a km/h. Nota: 124.4 es un valor de entrada válido.

```python
geschwindigkeit = float(input("Gebe eine Geschwindigkeit (in km/h) ein: "))

reaktionsweg = (geschwindigkeit / 10) * 3
bremsweg = (geschwindigkeit / 10) ** 2
anhalteweg = reaktionsweg + bremsweg

print(f"Geschwindigkeit: {geschwindigkeit:.2f} km/h")
print()
print(f"Reaktionsweg: {reaktionsweg:.2f} m")
print(f"Bremsweg: {bremsweg:.2f} m")
print(f"Anhalteweg: {anhalteweg:.2f} m")
print("---------------------------")

```

## Aufgabe 12:

Calcule los minutos indicados en días, horas y minutos para utilizar floordiv y modulo.


MINUTES_DAY = 24 * 60 # 1440


¿Cuántos días son 3434 minutos y cuántas horas y minutos quedan?

Beispiel

minuten = 3434 

son 2 días, 9 horas y 14 minutos

```python
total_minuten = int(input("geben Sie die Anzahl von Minuten um in Tagen berechnen zu werden: "))

tage = total_minuten // 1440
minuten_ubrig = total_minuten % 1440

stunden = minuten_ubrig // 60
minuten = minuten_ubrig % 60

print(f"{total_minuten} Minuten entsprechen:")
print(f"{tage} Tage, {stunden} Stunden und {minuten} Minuten")
print("---------------------------")
```

## Aufgabe 13

El pulso óptimo en los deportes de resistencia depende de la edad. Se puede determinar con la fórmula Puls = 165 - 0.75 * edad.

Así, el pulso óptimo para 18 años es 151 Escribe un programa que permite el siguiente diálogo:
