### Einführung in Algorithmen und Problemlösungen

Algorithmen sind Schritt-für-Schritt-Anweisungen zur Lösung eines bestimmten Problems. Sie beschreiben präzise, wie Daten verarbeitet und Operationen ausgeführt werden
müssen, um ein gewünschtes Ergebnis zu erzielen. Algorithmen finden in allen Bereichen der Informatik Anwendung, von der Softwareentwicklung bis hin zur künstlichen
Intelligenz.

Problemlösung ist der Prozess der Identifizierung und Implementierung einer Lösung für ein Problem. Dieser Prozess beinhaltet:

Problemdefinition: Das Problem klar und präzise verstehen, einschließlich der Eingabedaten, der gewünschten Ausgabe und der relevanten Einschränkungen.
Entwurf einer Lösung: Einen Algorithmus entwickeln, der das Problem löst.
Implementierung der Lösung: Den Algorithmus in einer Programmiersprache oder anderen Notation kodieren.
Testen und Evaluierung: Die Lösung auf Korrektheit und Effizienz testen.

#### Grundlegende Algorithmen

Es gibt verschiedene Arten von Algorithmen, die für verschiedene Arten von Problemen eingesetzt werden können. Zu den grundlegenden Algorithmen gehören:

Suchalgorithmen: Finden ein bestimmtes Element in einer Menge von Daten.
Sortieralgorithmen: Ordnen Daten in einer bestimmten Reihenfolge.
Graph-Algorithmen: Bearbeiten Beziehungen zwischen Objekten in einem Graphen.
Rekursion: Zerlegen eines Problems in kleinere Teilprobleme derselben Art.

#### Problemlösungsstrategien

Neben der Auswahl des richtigen Algorithmus gibt es verschiedene Strategien, die zur Lösung von Problemen eingesetzt werden können:

Zerlege und herrsche: Zerlegen Sie ein großes Problem in kleinere, leichter zu lösende Teilprobleme.
Greedy-Algorithmen: Finden Sie bei jedem Schritt die beste lokale Lösung, um sukzessive zur Gesamtlösung zu gelangen.
Dynamische Programmierung: Speichern Sie Zwischenergebnisse, um wiederholte Berechnungen zu vermeiden.

#### Vorteile von Algorithmen

Die Verwendung von Algorithmen bietet verschiedene Vorteile:

Effizienz: Algorithmen ermöglichen es, Probleme effizient und mit minimalem Ressourcenverbrauch zu lösen.
Genauigkeit: Algorithmen liefern präzise und reproduzierbare Ergebnisse.
Verallgemeinerbarkeit: Algorithmen können auf verschiedene ähnliche Probleme angewendet werden.

#### Fazit

Algorithmen und Problemlösungskompetenzen sind grundlegende Fähigkeiten in der Informatik und anderen Bereichen. Die Beherrschung dieser Fähigkeiten ermöglicht es Ihnen, komplexe Probleme effizient und präzise zu lösen.

## Datentypen:


In diesen einfachen Fall von eben handelt es sich um den Datentyp "**String**", was nichts anderes bedeutet, als eine Zeichenkette.
Natürlich gibt es weitere Datentypen, auch **primitive Datentypen** genannt, wie zum Beispiel:

* **String** (Zeichenketten) Beispiel: "Eierkuchen", "Eierkuchen ist lecker!", oder auch Zahlen, die eine einfache Zeichenkette darstellen.
* **Integer** (Ganzzahlen) Beispiele: 1, 2 ,15 , 10000 usw. Auch -2, -15, -1000 (negative Ganzzahlen)
* **Float** (Kommazahlen) Beispiele: 1.3, 5.9, 500.456 natürlich auch -20.2, -68.23 usw. auch 4E3 (4 mal 10 hoch 3)
* **Bool** (Wahrheitswerte) Beispiele: True, False

Die Unterscheidung der Datentypen ist wichtig, da die verschiedenen Datentypen unterschiedlich behandelt werden. So ist es zum Beispiel nicht möglich 2 **Strings**  zu multiplizieren, ein **Float** und **Integer** aber schon, wobei ein **Float** dabei rauskommt.
<br><br><br>


**Beispiel**: 2 Zahlen addieren (probieren Sie ruhig einige verschiedene Zahlen aus:)
<br>

```python
2 + 2
```

<br><br>...oder Multiplizieren:<br><br>

```python
2 * 4
```

<br><br>
Natürlich gibt es einige weitere mathematische Operatoren, die ohne das Laden zusätzlicher Module zur Verfügung stehen:

* **Division**: **/** Beispiel: 6/2 Ergebnis: 3
* **Subtraktion**: **-** Beispiel: 6-4
* **Modulo** (Rest einer Division): **%** Beispiel: 7%2 Ergebnis: 1
* **Ganzzahlendivision** (Division **ohne** Rest: **//** Beispiel: 7//2 Ergebnis: 3
* **Exponentation** (Hochrechnung): **\*\*** Beispiel: 2**3 Ergebnis: 8

# Vergleichsoperatoren

Vergleichsoperatoren werden dafür verwendet Werte zu vergleichen. Die Werte können entweder Zahlen, Strings oder ähnliches sein.


**==** - gleich <br>
**!=** - ungleich <br>
**>**  - größer <br>
**<**  - kleiner <br>
**>=** - größer oder gleich <br>
**<=** - kleiner oder gleich <br>


```python
print(3 >= 2)

print(3 >= 3 < 4 != 10)


```
