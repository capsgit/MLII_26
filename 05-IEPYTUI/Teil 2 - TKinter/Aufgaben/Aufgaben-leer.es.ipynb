## Aufgabe 1 Grundlegendes Tkinter-Fenster:

Crear una simple ventana de Tkinter con el título "Bienvenido a Tkunter" y un tamaño fijo de 300x200 píxeles.

```python
import tkinter as tk

def main():
    root = tk.Tk()
    root.title("Hello TKInter´s World")
    root.geometry("300x200")

    root.mainloop()

if __name__ == "__main__":
    main()
```

## Tarea 2 botón y etiqueta:

Añade un botón y una etiqueta a tu ventana de Tkinter. Si pulsas el botón, el texto de la etiqueta cambiará a "Button ha sido pulsado."

```python
import tkinter as tk
from tkinter import ttk

def main():
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")
    root.geometry("300x300")

    # Label
    label = ttk.Label(root, text="Ich bin ein Label!")
    label.pack(pady=10)

    # Funktion für den Button
    def button_clicked():
        label.config(text="Button wurde gedrückt")

    # Button
    button = ttk.Button(root, text="Ich bin ein Button", command=button_clicked)
    button.pack(pady=10)

    root.mainloop()

if __name__ == "__main__":
    main()
```

## Tarea 3 Campo de entrada y transmisión de datos:

Cree un programa Tkinter con un campo de entrada (Entry) y un botón. Si pulsa el botón, el contenido del cuadro de entrada se mostrará en la consola.

```python
import tkinter as tk
from tkinter import ttk

def main():
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")
    root.geometry("300x300")


    # Eingabefeld erstellen:
    eingabe = ttk.Entry(root)
    eingabe.pack(pady=5)

    def button_clicked():
        input_text = eingabe.get()
        print(input_text)

    # Button
    button = ttk.Button(root, text="Klick mich", command=button_clicked)
    button.pack(pady=10)



    root.mainloop()

if __name__ == "__main__":
    main()
```
