## Aufgabe 1 Grundlegendes Tkinter-Fenster:

Erstellen Sie ein einfaches Tkinter-Fenster mit dem Titel "Willkommen bei Tkinter" und einer festen Größe von 300x200 Pixeln.

```python
import tkinter as tk

def main():
    root = tk.Tk()
    root.title("Hello TKInter´s World")
    root.geometry("300x200")

    root.mainloop()

if __name__ == "__main__":
    main()
```

## Aufgabe 2 Button und Label:

Fügen Sie Ihrem Tkinter-Fenster einen Button und ein Label hinzu. Wenn der Button gedrückt wird, soll sich der Text des Labels in "Button wurde gedrückt" ändern.

```python
import tkinter as tk
from tkinter import ttk

def main():
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")
    root.geometry("300x300")

    # Label
    label = ttk.Label(root, text="Ich bin ein Label!")
    label.pack(pady=10)

    # Funktion für den Button
    def button_clicked():
        label.config(text="Button wurde gedrückt")

    # Button
    button = ttk.Button(root, text="Ich bin ein Button", command=button_clicked)
    button.pack(pady=10)

    root.mainloop()

if __name__ == "__main__":
    main()
```

## Aufgabe 3 Eingabefeld und Datenübertragung:

Erstellen Sie ein Tkinter-Programm mit einem Eingabefeld (Entry) und einem Button. Wenn der Button gedrückt wird, soll der Inhalt des Eingabefeldes in der Konsole ausgegeben werden.

```python
import tkinter as tk
from tkinter import ttk

def main():
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")
    root.geometry("300x300")


    # Eingabefeld erstellen:
    eingabe = ttk.Entry(root)
    eingabe.pack(pady=5)

    def button_clicked():
        input_text = eingabe.get()
        print(input_text)

    # Button
    button = ttk.Button(root, text="Klick mich", command=button_clicked)
    button.pack(pady=10)



    root.mainloop()

if __name__ == "__main__":
    main()
```
