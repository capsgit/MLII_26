Aquí está el ejemplo de una simple construcción de ventana de Tkinter:

```python
import tkinter as tk

def main():
    # <App /> => ventana principal
    root = tk.Tk()
    root.title("Hello TKInter´s World")

    # Widget -> Label erstellen
    label = tk.Label(root, text="Hello World")

    # Widget anzeigen
    label.pack()


    # Main loop
    root.mainloop()

if __name__ == "__main__":
    main()
```

## Agregar widgets.

En Tkinter, los widgets son los componentes de la interfaz gráfica de usuario. Incluyen elementos como botones, etiquetas, campos de texto y muchos más. Les mostraré cómo usar algunos wigets básicos en Tkint:

- **Button**: Ein einfacher Klick-Button.
- **Entry**: Un campo de texto para entradas de usuario.
- **Label**: Zur Anzeige von Text oder Bildern.
- **Checkbutton**: Una casilla de verificación para opciones que se pueden activar o desactivar.

Aquí está un ejemplo que utiliza estos widgets en una ventana de Tkinter:

```python
import tkinter as tk
from tkinter import ttk

def main():
    # Hauptfenster erstellen:
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")  # Fenster-Überschrift
    root.geometry("300x300")  # Fenstergröße

    # Label-Widget erstellen:
    label = tk.Label(root, text="Ich bin ein Label!")
    label.pack()  # Das Label-Widget wird gepackt (pack() ist ein einfacher Layout-Manager)

    # Eingabefeld erstellen:
    eingabe = tk.Entry(root)
    eingabe.pack()

    # Password-Feld erstellen:
    eingabepw = tk.Entry(root, show="*")
    eingabepw.pack()

    # Button erstellen:
    button = tk.Button(root, text="Ausführen", command=lambda: label.config(text="Hallo Welt!"))
    button.pack()

    # Checkbox (Haken) erstellen:
    check_var = tk.BooleanVar()
    checkbutton = tk.Checkbutton(root, text="Bitte auswählen", variable=check_var)
    checkbutton.pack()


    # Hauptschleife (Mainloop):
    root.mainloop()

if __name__ == "__main__":
    main()
```


Estos widgets son la base para la mayoría de las aplicaciones GUI. Se pueden combinar y personalizar para crear interfaces de usuario complejas. A continuación, podemos dedicarnos al manejo de eventos o a la gestión de diseños.

## Event Handling

Manejo de eventos es una parte esencial de la programación GUI. En Tkinter, los eventos se utilizan para responder a acciones de usuario como clics, teclas o movimientos del ratón. Aquí hay algunos ejemplos básicos de manejo de eventos en Tk Inter:

1. **Button-Click Event**: Ejecutar una función cuando se hace clic en un botón.
2. **Encontrar un evento Key-Press**: Responder a las pulsaciones de teclas.
3. **Encontrar un evento del ratón**: Responder a las acciones del ratón como clics o movimientos.

Aquí hay un ejemplo de cómo usar estos eventos en Tkinter:

```python
import tkinter as tk
from tkinter import ttk
from sys import exit

def on_button_click():
    print("Button wurde gedrückt!")


def on_key_press(event):
    print("Eingabe:", event.char)


def on_mouse_click(event):
    print("Linksklick bei:", event.x, event.y)

def on_mouse_click_right(event):
    print("Rechtsklick bei:", event.x, event.y)

def on_mouse_click_middle(event):
    print("Mittelklick bei:", event.x, event.y)

def on_mouse_move(event):
    print("Mausbewegung:", event.x, event.y)


def main():
    # Hauptfenster erstellen:
    root = tk.Tk()
    root.title("TKInter Events")  # Fenster-Überschrift
    root.geometry("300x300")  # Fenstergröße

    # Key Press Event
    root.bind("<KeyPress>", on_key_press)

    # Mausklick-Event (links)
    root.bind("<Button-1>", on_mouse_click)

    # Mausklick-Event (rechts)
    root.bind("<Button-3>", on_mouse_click_right)

    # Mausklick-Event (mitte)
    root.bind("<Button-2>", on_mouse_click_middle)

    # Mausbewegung
    root.bind("<Motion>", on_mouse_move)

    # Button erstellen:
    button = tk.Button(root, text="Klick mich!", command=on_button_click)  # Angabe der Funktion OHNE Klammern!
    button.pack()


    # Hauptschleife (Mainloop):
    root.mainloop()

if __name__ == "__main__":
    main()
```

In diesem Beispiel:

- El botón ejecuta la función on_button_click cuando se hace clic.
- El evento Key-Press es manejado por la función on_key_press que se llama en cada pulsación de tecla en la ventana.
- El evento Mouse-Click es manejado por la función on_mouse_click que se llama cada vez que se hace clic en el botón izquierdo del ratón en la ventana.

- `<Button-4>` : Mausrad nach oben (Linux)
- `<Button-5>` : Mausrad nach unten (Linux)
- `<Double-Button-1>` : Doppelklick linke Maustaste

- `<Shift-Button-1>` : Linksklick mit gedrückter Shift-Taste
- `<Control-Button-1>` : Linksklick mit gedrückter Strg-Taste
- `<Alt-Button-3>` : Rechtsklick mit gedrückter Alt-Taste

## Mausbewegungen
- `<Motion>` : Mausbewegung über dem Widget
- `<Enter>` : Maus betritt Widget
- `<Leave>` : Maus verlässt Widget
- `<MouseWheel>` : Mausrad (Windows/Mac, Linux nutzt Button-4/5)
- `<B1-Motion>` : Mausbewegung mit gedrückter linker Maustaste (Drag)
- `<B2-Motion>` : Mausbewegung mit gedrückter mittlerer Maustaste
- `<B3-Motion>` : Mausbewegung mit gedrückter rechter Maustaste
- `<Shift-B1-Motion>` : Drag mit linker Maustaste + Shift

## Tastatur
- `<Key>` : Beliebige Taste gedrückt
- `<KeyPress>` : Synonym zu `<Key>`
- `<KeyRelease>` : Taste losgelassen
- `<Return>` : Enter-Taste

## Fokus / Fenster
- `<FocusIn>` : Widget bekommt Fokus
- `<FocusOut>` : Widget verliert Fokus
- `<Activate>` : Fenster wird aktiv
- `<Deactivate>` : Fenster wird inaktiv
- `<Map>` : Widget wird sichtbar (gemappt)
- `<Unmap>` : Widget wird unsichtbar

- `<Expose>` : Widget muss neu gezeichnet werden
- `<Visibility>` : Sichtbarkeit des Widgets ändert sich

## System / Sonstiges
- `<Destroy>` : Widget wird zerstört

- `<Gravity>` : Schwerkraft-Ereignis (selten, X11)

- `<SelectionRequest>` : Externe Anfrage auf Clipboard-Inhalt
- `<SelectionNotify>` : Rückmeldung zu Clipboard-Anfrage

#### Ein paar Beispiele:

```python
import tkinter as tk

def log_event(event):
    """Allgemeine Event-Logger-Funktion"""
    print(f"Event: {event.type} | Widget: {event.widget} | Keysym: {event.keysym if hasattr(event, 'keysym') else ''} | Num: {getattr(event, 'num', '')} | State: {event.state} | x={event.x}, y={event.y}")

root = tk.Tk()
root.title("Tkinter Event Tester")
root.geometry("400x300")

label = tk.Label(root, text="Drücke Tasten oder benutze die Maus\n(alles wird im Terminal ausgegeben)",
                 font=("Arial", 12), justify="center")
label.pack(expand=True, fill="both")

# Alle relevanten Eventtypen binden
events = [
    "<Key>", "<KeyRelease>",
    "<Button-1>", "<Button-2>", "<Button-3>",
    "<Double-Button-1>", "<Triple-Button-1>",
    "<ButtonRelease-1>", "<B1-Motion>", "<B2-Motion>", "<B3-Motion>",
    "<Motion>", "<Enter>", "<Leave>", "<MouseWheel>",
    "<FocusIn>", "<FocusOut>",
    "<Configure>", "<Destroy>",
]

for ev in events:
    root.bind_all(ev, log_event)

root.mainloop()
```

#### Erweiterte TKinter Widgets

```python
import tkinter as tk
from tkinter import ttk

def update_label_from_list(listbox, label):
    selection = listbox.curselection()
    if selection:
        label.config(text=listbox.get(selection))

def main():
    # Hauptfenster erstellen:
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")

    root.geometry("300x400") # Breite mal Höhe

    label = tk.Label(root, text="Bitte auswählen:")
    label.pack()

    # List Box
    listbox = tk.Listbox(root)
    listbox.pack()
    for i in ["Auswahl 1", "Auswahl 2", "Auswahl 3", "Auswahl 4"]:
        listbox.insert(tk.END, i)  # Jeder Eintrag wird ans ENDE eingefügt.


    listbox.bind("<<ListboxSelect>>", lambda event: update_label_from_list(listbox, label))


    # Canvas (Zeichenbereich) erstellen:
    canvas = tk.Canvas(root, bg="white", height=100, width=200)
    canvas.pack(padx=20, pady=20)
    canvas.create_oval(50, 25, 150, 75, fill="blue")

    # Schieberegler erstellen:
    # scale = ttk.Scale(root, from_=0, to=255, orient=tk.HORIZONTAL) # ttk
    # scale = tk.Scale(root, from_=0, to=255, orient="vertical", resolution=10) # 10er Schritte, vertikal
    scale = tk.Scale(root, from_=0, to=255, orient="horizontal", resolution=10) # tk
    scale.pack()

    root.mainloop()


if __name__ == "__main__":
    main()
```

In diesem Beispiel:

## Tkinter Label con StringVar

- Ein Tkinter-Fenster wird erstellt.

- Se crea una etiqueta cuyo texto está vinculado al objeto StringVar.


#### Un reloj digital con StringVar:


```python
import time


def update_time():
    global time_var, root, label
    current_time = time.strftime("%H:%M:%S")
    time_var.set(current_time)

    root.after(1000, update_time)


def main():
    global time_var, root, label
    root = tk.Tk()
    root.title("Digital-Uhr")
    root.geometry("300x100")

    time_var = tk.StringVar()

    label = tk.Label(root, textvariable=time_var, font=("Comic Sans MS", 48))
    label.pack()

    update_time()

    root.mainloop()


if __name__ == "__main__":
    main()
```

Ejemplo: La misma aplicación pero con Nested Functions, ahorrando variables globales.

```python
import tkinter as tk
import time


def main():
    def update_time():
        # global time_var, root, label
        current_time = time.strftime("%H:%M:%S")
        time_var.set(current_time)

        root.after(1000, update_time)

    # global time_var, root, label
    root = tk.Tk()
    root.title("Digital-Uhr")
    root.geometry("300x100")

    time_var = tk.StringVar()

    label = tk.Label(root, textvariable=time_var, font=("Comic Sans MS", 48))
    label.pack()

    update_time()

    root.mainloop()


if __name__ == "__main__":
    main()
```

Was passiert hier?

- La función update_time actualiza el valor de StringVar time_var con la hora actual.
- La etiqueta está vinculada a time_var y muestra la hora actual.
- La función root.after(1000, update_time) garantiza que la actualización_tim se llame una vez por segundo, actualizando la hora cada segundo.
- Durante este proceso, la ventana Tkinter sigue siendo interactiva y responde a las acciones del usuario.

## Layout-Management in Tkinter

La gestión de diseños en Tkinter se refiere a cómo se organizan los widgets dentro de la ventana. Tkunter ofrece varios métodos de gestión de maquetaciones: **pack**, **grid** y **place**.

- **Pack**: Ordena widgets verticales u horizontales. Es fácil de usar, pero ofrece menos control que grid.

- **Grid**: Coloca los widgets en un formato similar a la tabla. Ofrece un control más fino sobre la colocación de los  Widgets, donde cada celda de la tabla puede contener un wiget.

- **Place**: Si permite especificar widgets en posiciones exactas, ofrece el control más alto, pero es menos flexible en cuanto a la adaptación a diferentes tamaños de ventana.

```python

def main():
    root = tk.Tk()
    root.title("Layoutmangement")
    root.geometry("300x100")

    frame_pack = tk.Frame(root, borderwidth=2, relief="solid")
    frame_pack.pack(side="left", fill="both", expand=True)

    label1 = tk.Label(frame_pack, text="Label 1 (Pack)")
    label1.pack()

    label2 = tk.Label(frame_pack, text="Label 2 (Pack)")
    label2.pack()

    frame_grid = tk.Frame(root, borderwidth=2, relief="solid")
    frame_grid.pack(side="right", fill="both", expand=True)

    label3 = tk.Label(frame_grid, text="Label 3 (Grid)")
    label3.grid(row=0, column=0)  # Reihe Spalte

    label4 = tk.Label(frame_grid, text="Label 4 (Grid)")
    label4.grid(row=1, column=0)

    label5 = tk.Label(frame_grid, text="Label 5 (Grid)")
    label5.grid(row=0, column=1)  # Reihe Spalte

    label6 = tk.Label(frame_grid, text="Label 6 (Grid)")
    label6.grid(row=1, column=1)



    root.mainloop()


if __name__ == "__main__":
    main()
```

#### Verwendung von **Pack**:

- **frame_pack** = tk.Frame(root, borderwidth=2, relief="solid"): Crea un marco (frame) que sirve de contenedor para otros widgets. El marco tiene un ancho de borde de 2 y un estilo relieve que resalta visualmente el marco.
- **frame_pack. pack(side="left," fill="both," expand=True)**: Coloca el frame_pack en la ventana principal. **side="Left"** significa que el marco se coloca a la izquierda. **fill="both** permite que el cuadro se expanda tanto horizontal como verticalmente. **expand=Tru** permite al marco tomar espacio adicional en la pantalla.
- **label1 = tk.Label(frame_pack, text="Label 1 (Pack)")**: Erstellt ein Label-Widget im frame_pack.
- **label1.pack()**: Coloca el widget de etiqueta dentro del frame_pack. Por defecto, pack ordenará el  Widget verticalmente.

#### Verwendung von Grid:

- **frame_grid = tk.Frame(root, borderwidth=2, relief="solid")**: Crea un marco más que un contenedor para widgets de diseño de widd.
- **frame_grid.pack(side="right," fill="both," expand=True)**: Coloca el frame_grid en el lado derecho de la ventana principal.
- **label3 = tk.Label(frame_grid, text="Label 3 (Grid)")**: Erstellt ein Label-Widget im frame_grid.
- **label3.grid(row=0, columna=0)**: Coloca el widget label3 en la primera fila **(row =0)** y la primera columna **(column=0) ** del frame_grid.
- **label4 = tk.Label(frame_grid, text="Label 4 (Grid)")**: Erstellt ein weiteres Label-Widget.
- **label4.grid(row=1, columna=1)**: Coloca el widget label4 en la segunda fila **(row=)** y la segunda columna **(column=1) **.



En Tkinter hay varios estilos de relieve que cambian la apariencia de los bordes de los widgets, como marcos, etiquetas, etc. La propiedad de relieve determina el estilo del marco 3D que se dibuja alrededor del wiget. Aquí están los estilos disponibles:

**flat**: Ningún relieve, el borde aparece plano. **raised**: Un borde alto que parece que el widget está flotando por encima de la superficie. **sunken**: un borde profundo que es lo opuesto al raised y que parece como si el wadget se hubiera metido en la superficie **groove**: Es un borde enterrado o desgastado que parece una lente o un canal. **ridge**: es un borde elevado que se parece a una grada o a un borde y es lo contrario de grooce.

Por ejemplo, en su código, puede cambiar la propiedad relief de un marco para obtener diferentes efectos visuales

```python
def main():
    root = tk.Tk()
    root.title("Layoutmangement")
    root.geometry("300x100")

    frame_pack = tk.Frame(root, borderwidth=2, relief="solid")
    frame_pack.pack(side="left", fill="both", expand=True)

    label1 = tk.Label(frame_pack, text="Label 1 (Pack)")
    label1.pack()

    label2 = tk.Label(frame_pack, text="Label 2 (Pack)")
    label2.pack()

    frame_grid = tk.Frame(root, borderwidth=2, relief="solid")
    frame_grid.pack(side="right", fill="both", expand=True)

    # Cell-Optionen festlegen:
    cell_opts = dict(borderwidth=1, relief="solid", padx=6, pady=3)
    # Ende

    label3 = tk.Label(frame_grid, text="Label 3 (Grid)", **cell_opts)
    label3.grid(row=0, column=0, sticky="nsew")

    label4 = tk.Label(frame_grid, text="Label 4 (Grid)", **cell_opts)
    label4.grid(row=1, column=0, sticky="nsew")

    label5 = tk.Label(frame_grid, text="Label 5 (Grid)", **cell_opts)
    label5.grid(row=0, column=1, sticky="nsew")

    label6 = tk.Label(frame_grid, text="Label 6 (Grid)", **cell_opts)
    label6.grid(row=1, column=1, sticky="nsew")

    # Größe und Ausdehnung konfigurieren:
    frame_grid.grid_columnconfigure(0, weight=1)
    frame_grid.grid_columnconfigure(1, weight=1)
    frame_grid.grid_rowconfigure(0, weight=1)
    frame_grid.grid_rowconfigure(1, weight=1)

    root.mainloop()


if __name__ == "__main__":
    main()
```

Nota: Mientras que pack es un método más simple que ordena widgets sucesivamente, grid ofrece un control más fino permitiendo un diseño similar a la tabla.

## Diálogos y mensajes

Tkinter ofrece varias formas de mostrar diálogos y mensajes:

- **Messagebox**: Mensajes pop-up simples para información, advertencia, error, etc.
- **Filedialog**: Para abrir y guardar archivos.
- **Simpledialog**: Para entradas de usuario simples.

Ejemplo de cuadro de mensajes y diálogo de archivos:

```python
import tkinter as tk
from tkinter import messagebox, filedialog, simpledialog


def zeige_messagebox():
    messagebox.showinfo("Info", "Hallo Welt!")

def oeffne_filedialog():
    dateiname = filedialog.askopenfilename(initialdir="/", title="Bitte wählen Sie eine Datei aus",
                                           filetypes=(("Textdateien", "*.txt"), ("Alle Dateien", "*.*")))

    if dateiname:
        label_dateiname.config(text=dateiname)

    else:
        label_dateiname.config(text="Nichts ausgewählt")

def speichern_filedialog():
    dateiname = filedialog.asksaveasfilename(initialdir="/",
                                             title="Bitte wählen Sie einen Dateinamen aus",
                                             filetypes=(("Textdateien", "*.txt"),
                                                        ("Markdown Dateien", "*.md"),
                                                        ("Alle Dateien", "*.*")),
                                             confirmoverwrite=True # Frage, ob Datei überschrieben werden soll
                                             )

    if dateiname:
        label_dateiname.config(text=dateiname)
    else:
        label_dateiname.config(text="Nichts ausgewählt")

def frage_nach_eingabe():
    # antwort = None
    antwort = simpledialog.askstring("Eingabe", "Bitte geben Sie etwas ein:")
    if antwort:
        messagebox.showinfo("Ihre Antwort",antwort)

def frage_ja_nein():
    antwort = messagebox.askyesno("Ja/Nein", "Sind Sie sicher?")

    if antwort:
        messagebox.showinfo("Ja", "Danke!")
    else:
        messagebox.showinfo("Nein", "Okay.")

def frage_ja_nein_abbrechen():
    antwort = messagebox.askyesnocancel("Speichern/Überschreiben", "Sind Sie sicher?")
    # Am Beispiel von Dateien:
    if antwort:
        # Vorhandener Dateiname wird verwendet
        messagebox.showinfo("Ja", "Ja wurde ausgewählt, Danke!")

    elif antwort is False:
        # Neuer Dateiname darf ausgewählt werden
        messagebox.showinfo("Nein", "Nein wurde ausgewählt. Okaaaay")

    else:
        # Keine Aktion, zurück
        messagebox.showinfo("Abbruch", "Dialog abgebrochen.")

def zeige_warnung():
    messagebox.showwarning("Warnung", "Bitte beachten Sie, \ndass diese Warnung eine Warnung ist!")


def main():
    global label_dateiname
    root = tk.Tk()
    root.title("Dialoge und Nachrichten")
    root.geometry("300x300")

    btn_messagebox = tk.Button(root, text="Zeige Messagebox", command=zeige_messagebox)
    btn_messagebox.pack()

    btn_filedialog = tk.Button(root, text="Öffne Filedialog", command=oeffne_filedialog)
    btn_filedialog.pack()

    btn_savefiledialog = tk.Button(root, text="Speichern Filedialog", command=speichern_filedialog)
    btn_savefiledialog.pack()

    btn_simpledialog = tk.Button(root, text="Öffne SimpleDialog", command=frage_nach_eingabe)
    btn_simpledialog.pack()

    btn_ja_nein = tk.Button(root, text="Frage Ja/Nein", command=frage_ja_nein)
    btn_ja_nein.pack()

    btn_ja_nein_abbrechen = tk.Button(root, text="Frage Ja/nein/Abbrechen", command=frage_ja_nein_abbrechen)
    btn_ja_nein_abbrechen.pack()

    btn_warning = tk.Button(root, text="Zeige Warnung", command=zeige_warnung)
    btn_warning.pack()

    label_dateiname = tk.Label(root, text="Keine Datei ausgewählt")
    label_dateiname.pack()

    root.mainloop()


if __name__ == "__main__":
    main()
```

In diesem Beispiel:

- **oeffne_filedialog**: Cuando se selecciona un archivo, la función actualiza el nombre del archivo label_ con la ruta del archivo seleccionado.
- **Pregunta_Después_Entrar**: Esta función abre un diálogo **simple** en el que el usuario puede introducir una cadena. La respuesta introducida se muestra en una caja de mensajes.
- El widget **label_archivo** sirve para mostrar información como la ruta del último archivo seleccionado. Esto le permite dar al usuario una respuesta visual sobre qué acciones se han realizado.

TK tiene un poco de polvo, pero aún así tiene su encanto. Un poco de mejora ofrece ttk https://docs.python.org/3/library/tkinter.ttk.html.

Por ejemplo, es elegante y colorido con ttkbootstrap (https://ttkbotstrap.readthedocs.io/en/latest/) o customtkinter (http://customtkiinter.tomshimansky.com/).

```python
import tkinter as tk
from tkinter import ttk
import tkinter.ttk as tk_progress  # Für tk Progressbar

def button_click():
    print("Button wurde geklickt!")

# Hauptfenster erstellen
root = tk.Tk()
root.title("tk vs ttk Widgets Vergleich")
root.geometry("600x500")

# Überschrift
tk.Label(root, text="Vergleich von tk und ttk Widgets",
         font=("Arial", 16)).pack(pady=10)

# Frame für tk-Widgets
tk_frame = tk.LabelFrame(root, text="tk Widgets", padx=10, pady=10)
tk_frame.pack(side=tk.LEFT, padx=10, pady=10, fill=tk.BOTH, expand=True)

# Frame für ttk-Widgets
ttk_frame = tk.LabelFrame(root, text="ttk Widgets", padx=10, pady=10)
ttk_frame.pack(side=tk.RIGHT, padx=10, pady=10, fill=tk.BOTH, expand=True)

# tk-Widgets
tk.Label(tk_frame, text="tk Label", font=("Arial", 12)).pack(pady=5)
tk.Button(tk_frame, text="tk Button", command=button_click).pack(pady=5)
tk.Entry(tk_frame, width=20).pack(pady=5)
tk.Checkbutton(tk_frame, text="tk Checkbox").pack(pady=5)

# ttk-Widgets
ttk.Label(ttk_frame, text="ttk Label", font=("Arial", 12)).pack(pady=5)
ttk.Button(ttk_frame, text="ttk Button", command=button_click).pack(pady=5)
ttk.Entry(ttk_frame, width=20).pack(pady=5)
ttk.Checkbutton(ttk_frame, text="ttk Checkbox").pack(pady=5)

# tk Schieberegler (Scale)
tk.Label(tk_frame, text="tk Schieberegler").pack(pady=5)
tk_slider = tk.Scale(tk_frame, from_=0, to=100,
                     orient=tk.HORIZONTAL, length=200)
tk_slider.pack(pady=5)

# ttk Schieberegler
ttk.Label(ttk_frame, text="ttk Schieberegler").pack(pady=5)
ttk_slider = ttk.Scale(ttk_frame, from_=0, to=100,
                       orient=tk.HORIZONTAL, length=200)
ttk_slider.pack(pady=5)

# tk Fortschrittsbalken
tk.Label(tk_frame, text="tk Fortschrittsbalken").pack(pady=5)
tk_progress_bar = tk_progress.Progressbar(tk_frame, length=200,
                                mode='determinate', maximum=100)
tk_progress_bar.pack(pady=5)
tk_progress_bar['value'] = 50  # Setze Fortschritt auf 50%

# ttk Fortschrittsbalken
ttk.Label(ttk_frame, text="ttk Fortschrittsbalken").pack(pady=5)
ttk_progress = ttk.Progressbar(ttk_frame, length=200,
                                mode='determinate', maximum=100)
ttk_progress.pack(pady=5)
ttk_progress['value'] = 50  # Setze Fortschritt auf 50%

root.mainloop()
```
