<br>

- **Anpassbarkeit und Erweiterbarkeit**: Tkinter lässt sich durch verschiedene Mittel anpassen und erweitern, beispielsweise durch Integration mit anderen Bibliotheken oder durch das Erstellen eigener Widgets.

<br>

- Ideal für didaktische Zwecke: Aufgrund seiner Einfachheit und breiten Verfügbarkeit wird Tkinter oft in Bildungsumgebungen eingesetzt, um die Grundlagen der Programmierung und GUI-Entwicklung zu vermitteln.



Tkinter mag zwar nicht die modernsten oder fortschrittlichsten GUI-Funktionen bieten, wie sie in anderen Bibliotheken wie **PyQt** oder **Kivy** zu finden sind, aber für viele Anwendungsfälle, insbesondere für Einsteiger, Bildungszwecke und kleinere Projekte, ist es eine ausgezeichnete Wahl.

Hier ist das Beispiel für einen einfachen Tkinter-Fensteraufbau:

````python
import tkinter as tk

def main():
    # <App /> => ventana principal
    root = tk.Tk()
    root.title("Hello TKInter´s World")

    # Widget -> Label erstellen
    label = tk.Label(root, text="Hello World")

    # Widget anzeigen
    label.pack()


    # Main loop
    root.mainloop()

if __name__ == "__main__":
    main()
````

## Widgets hinzufügen. 

In Tkinter sind Widgets die Bausteine für die grafische Benutzeroberfläche. Sie umfassen Elemente wie Buttons, Labels, Textfelder und viele mehr. Ich werde Ihnen zeigen, wie man einige grundlegende Widgets in Tkinter verwendet:

- **Button**: Ein einfacher Klick-Button.
- **Entry**: Ein Textfeld für Benutzereingaben.
- **Label**: Zur Anzeige von Text oder Bildern.
- **Checkbutton**: Eine Checkbox für Optionen, die aktiviert oder deaktiviert werden können.

Hier ist ein Beispiel, das diese Widgets in einem Tkinter-Fenster verwendet:

````python
import tkinter as tk
from tkinter import ttk

def main():
    # Hauptfenster erstellen:
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")  # Fenster-Überschrift
    root.geometry("300x300")  # Fenstergröße

    # Label-Widget erstellen:
    label = tk.Label(root, text="Ich bin ein Label!")
    label.pack()  # Das Label-Widget wird gepackt (pack() ist ein einfacher Layout-Manager)

    # Eingabefeld erstellen:
    eingabe = tk.Entry(root)
    eingabe.pack()

    # Password-Feld erstellen:
    eingabepw = tk.Entry(root, show="*")
    eingabepw.pack()

    # Button erstellen:
    button = tk.Button(root, text="Ausführen", command=lambda: label.config(text="Hallo Welt!"))
    button.pack()

    # Checkbox (Haken) erstellen:
    check_var = tk.BooleanVar()
    checkbutton = tk.Checkbutton(root, text="Bitte auswählen", variable=check_var)
    checkbutton.pack()


    # Hauptschleife (Mainloop):
    root.mainloop()

if __name__ == "__main__":
    main()
````


Diese Widgets sind die Grundlage für die meisten GUI-Anwendungen. Sie können kombiniert und angepasst werden, um komplexe Benutzeroberflächen zu erstellen. Als nächstes können wir uns dem Event-Handling oder dem Layout-Management widmen. 

## Event Handling

Event-Handling ist ein wesentlicher Bestandteil der GUI-Programmierung. In Tkinter werden Ereignisse verwendet, um auf Benutzeraktionen wie Klicks, Tastendrucke oder Mausbewegungen zu reagieren. Hier sind einige grundlegende Beispiele für Event-Handling in Tkinter:

1. **Button-Click Event**: Ausführen einer Funktion, wenn ein Button geklickt wird.
2. **Binden eines Key-Press Events**: Reaktion auf Tastendrücke.
3. **Binden eines Mouse-Events**: Reaktion auf Mausaktionen wie Klicks oder Bewegungen.

Hier ist ein Beispiel, das zeigt, wie man diese Events in Tkinter verwendet:


````python
import tkinter as tk
from tkinter import ttk
from sys import exit

def on_button_click():
    print("Button wurde gedrückt!")

def on_key_press(event):
    print("Eingabe:", event.char)

def on_mouse_click(event):
    print("Linksklick bei:", event.x, event.y)

def on_mouse_click_right(event):
    print("Rechtsklick bei:", event.x, event.y)

def on_mouse_click_middle(event):
    print("Mittelklick bei:", event.x, event.y)

def on_mouse_move(event):
    print("Mausbewegung:", event.x, event.y)


def main():
    # Hauptfenster erstellen:
    root = tk.Tk()
    root.title("TKInter Events")  # Fenster-Überschrift
    root.geometry("300x300")  # Fenstergröße

    # Key Press Event
    root.bind("<KeyPress>", on_key_press)

    # Mausklick-Event (links)
    root.bind("<Button-1>", on_mouse_click)

    # Mausklick-Event (rechts)
    root.bind("<Button-3>", on_mouse_click_right)

    # Mausklick-Event (mitte)
    root.bind("<Button-2>", on_mouse_click_middle)

    # Mausbewegung
    root.bind("<Motion>", on_mouse_move)

    # Button erstellen:
    button = tk.Button(root, text="Klick mich!", command=on_button_click)  # Angabe der Funktion OHNE Klammern!
    button.pack()


    # Hauptschleife (Mainloop):
    root.mainloop()

if __name__ == "__main__":
    main()
````

In diesem Beispiel:

- Der Button führt die on_button_click Funktion aus, wenn er geklickt wird.
- Das Key-Press Event wird durch die on_key_press Funktion gehandhabt, die bei jedem Tastendruck im Fenster aufgerufen wird.
- Das Mouse-Click Event wird durch die on_mouse_click Funktion gehandhabt, die bei jedem Klick der linken Maustaste im Fenster aufgerufen wird.

# Übersicht Tkinter-Events (Signale)

## Mausklicks
- `<Button-1>` : Linke Maustaste gedrückt
- `<Button-2>` : Mittlere Maustaste gedrückt (oder Scrollrad-Klick)
- `<Button-3>` : Rechte Maustaste gedrückt
- `<Button-4>` : Mausrad nach oben (Linux)
- `<Button-5>` : Mausrad nach unten (Linux)
- `<Double-Button-1>` : Doppelklick linke Maustaste
- `<Triple-Button-1>` : Dreifachklick linke Maustaste
- `<ButtonRelease-1>` : Linke Maustaste losgelassen
- `<Shift-Button-1>` : Linksklick mit gedrückter Shift-Taste
- `<Control-Button-1>` : Linksklick mit gedrückter Strg-Taste
- `<Alt-Button-3>` : Rechtsklick mit gedrückter Alt-Taste

## Mausbewegungen
- `<Motion>` : Mausbewegung über dem Widget
- `<Enter>` : Maus betritt Widget
- `<Leave>` : Maus verlässt Widget
- `<MouseWheel>` : Mausrad (Windows/Mac, Linux nutzt Button-4/5)
- `<B1-Motion>` : Mausbewegung mit gedrückter linker Maustaste (Drag)
- `<B2-Motion>` : Mausbewegung mit gedrückter mittlerer Maustaste
- `<B3-Motion>` : Mausbewegung mit gedrückter rechter Maustaste
- `<Shift-B1-Motion>` : Drag mit linker Maustaste + Shift

## Tastatur
- `<Key>` : Beliebige Taste gedrückt
- `<KeyPress>` : Synonym zu `<Key>`
- `<KeyRelease>` : Taste losgelassen
- `<Return>` : Enter-Taste
- `<Escape>` : Escape-Taste
- `<Tab>` : Tab-Taste
- `<BackSpace>` : Rückschritt
- `<Delete>` : Entfernen-Taste
- `<Shift_L>` / `<Shift_R>` : Linke/rechte Shift-Taste
- `<Control_L>` / `<Control_R>` : Linke/rechte Strg-Taste
- `<Alt_L>` / `<Alt_R>` : Linke/rechte Alt-Taste
- `<F1>` … `<F12>` : Funktionstasten
- `<space>` : Leertaste
- `<a>`, `<b>`, … `<z>` : Zeichen-Tasten
- `<Control-c>` : Strg+C (Copy)
- `<Control-v>` : Strg+V (Paste)
- `<Control-Alt-Delete>` : (Windows-Spezial, oft vom OS abgefangen)
- `<Shift-Return>` : Shift + Enter
- `<Control-Shift-Escape>` : (Taskmanager unter Windows, selten abfangbar)
- `<KP_Enter>` : Enter auf dem Nummernblock
- `<KP_Add>`, `<KP_Subtract>`, `<KP_Multiply>`, `<KP_Divide>` : Rechenzeichen auf dem Nummernblock
- `<KP_0>` … `<KP_9>` : Nummernblock-Ziffern

## Fokus / Fenster
- `<FocusIn>` : Widget bekommt Fokus
- `<FocusOut>` : Widget verliert Fokus
- `<Activate>` : Fenster wird aktiv
- `<Deactivate>` : Fenster wird inaktiv
- `<Map>` : Widget wird sichtbar (gemappt)
- `<Unmap>` : Widget wird unsichtbar
- `<Configure>` : Größe oder Position des Widgets ändert sich
- `<Expose>` : Widget muss neu gezeichnet werden
- `<Visibility>` : Sichtbarkeit des Widgets ändert sich
- `<Alt-F4>` : Versuch, Fenster zu schließen (meist vom OS behandelt)

## System / Sonstiges
- `<Destroy>` : Widget wird zerstört
- `<Circulate>` : Fensterreihenfolge ändert sich (selten genutzt)
- `<Gravity>` : Schwerkraft-Ereignis (selten, X11)
- `<Reparent>` : Widget wird in ein anderes Eltern-Widget verschoben
- `<Property>` : Fenster-Property wurde geändert
- `<Colormap>` : Farbzuordnung hat sich geändert
- `<ClientMessage>` : Nachricht von einem Client (X11)
- `<Keymap>` : Tastaturbelegung hat sich geändert
- `<MapRequest>` : (X11: anderes Fenster möchte sichtbar werden)
- `<SelectionClear>` : Auswahl (Clipboard) wurde geleert
- `<SelectionRequest>` : Externe Anfrage auf Clipboard-Inhalt
- `<SelectionNotify>` : Rückmeldung zu Clipboard-Anfrage

## Zeitgesteuert
- `.after(ms, callback)` : Signal nach Zeitablauf (kein echtes Event, aber üblich in Tkinter)
- `.after_cancel(id)` : Abbrechen eines geplanten `.after`-Events


#### Ein paar Beispiele:

````python
import tkinter as tk

def log_event(event):
    """Allgemeine Event-Logger-Funktion"""
    print(f"Event: {event.type} | Widget: {event.widget} | Keysym: {event.keysym if hasattr(event, 'keysym') else ''} | Num: {getattr(event, 'num', '')} | State: {event.state} | x={event.x}, y={event.y}")

root = tk.Tk()
root.title("Tkinter Event Tester")
root.geometry("400x300")

label = tk.Label(root, text="Drücke Tasten oder benutze die Maus\n(alles wird im Terminal ausgegeben)",
                 font=("Arial", 12), justify="center")
label.pack(expand=True, fill="both")

# Alle relevanten Eventtypen binden
events = [
    "<Key>", "<KeyRelease>",
    "<Button-1>", "<Button-2>", "<Button-3>",
    "<Double-Button-1>", "<Triple-Button-1>",
    "<ButtonRelease-1>", "<B1-Motion>", "<B2-Motion>", "<B3-Motion>",
    "<Motion>", "<Enter>", "<Leave>", "<MouseWheel>",
    "<FocusIn>", "<FocusOut>",
    "<Configure>", "<Destroy>",
]

for ev in events:
    root.bind_all(ev, log_event)

root.mainloop()
````

#### Erweiterte TKinter Widgets

````python
import tkinter as tk
from tkinter import ttk

def update_label_from_list(listbox, label):
    selection = listbox.curselection()
    if selection:
        label.config(text=listbox.get(selection))

def main():
    # Hauptfenster erstellen:
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")

    root.geometry("300x400") # Breite mal Höhe

    label = tk.Label(root, text="Bitte auswählen:")
    label.pack()

    # List Box
    listbox = tk.Listbox(root)
    listbox.pack()
    for i in ["Auswahl 1", "Auswahl 2", "Auswahl 3", "Auswahl 4"]:
        listbox.insert(tk.END, i)  # Jeder Eintrag wird ans ENDE eingefügt.


    listbox.bind("<<ListboxSelect>>", lambda event: update_label_from_list(listbox, label))


    # Canvas (Zeichenbereich) erstellen:
    canvas = tk.Canvas(root, bg="white", height=100, width=200)
    canvas.pack(padx=20, pady=20)
    canvas.create_oval(50, 25, 150, 75, fill="blue")

    # Schieberegler erstellen:
    # scale = ttk.Scale(root, from_=0, to=255, orient=tk.HORIZONTAL) # ttk
    # scale = tk.Scale(root, from_=0, to=255, orient="vertical", resolution=10) # 10er Schritte, vertikal
    scale = tk.Scale(root, from_=0, to=255, orient="horizontal", resolution=10) # tk
    scale.pack()

    root.mainloop()


if __name__ == "__main__":
    main()
````

In diesem Beispiel:

- **ListBox**: Zeigt eine Liste von Items an. Bei Auswahl eines Items wird das Label entsprechend aktualisiert.
- **Canvas**: Zeichnet ein einfaches blaues Oval. Sie können im Canvas diverse Formen und Grafiken zeichnen.
- **Scale**: Ein horizontaler Schieberegler, der Werte von 0 bis 100 auswählen kann.

#### Ein wenig Magie:

Hier wird mit wenig Aufwand ein roter Ball durch den Bildschirm bewegt:

````python
import tkinter as tk
from tkinter import ttk

def bewege_ball():
    global canvas, ball, root
    canvas.move(ball, 5, 0)
    root.after(50, bewege_ball)  # nach 50 Milisekunden wieder aufrufen

def main():

    # Variablen Global setzen:
    global canvas, ball, root
    # Hauptfenster erstellen:
    root = tk.Tk()
    root.title("Ein Ball")

    root.geometry("300x400") # Breite mal Höhe

    label = tk.Label(root, text="Guckst Du:")
    label.pack()
    # Canvas (Zeichenbereich) erstellen:
    canvas = tk.Canvas(root, bg="white", height=400, width=300)
    canvas.pack(padx=20, pady=20)

    ball = canvas.create_oval(10, 100, 50, 140, fill="red")

    bewege_ball()  # Funktion aufrufen

    root.mainloop()


if __name__ == "__main__":
    main()
````

- Ein roter Ball (ein Oval) wird auf dem Canvas gezeichnet.
- Die Funktion **bewege_ball** verschiebt den Ball horizontal über das Canvas.
- **root.after(50, bewege_ball)** ruft die Funktion bewege_ball alle 50 Millisekunden auf, was eine einfache Animation erzeugt.

- Ein roter Ball (ein Oval) wird auf dem Canvas gezeichnet.
- Die Funktion **bewege_ball** verschiebt den Ball horizontal über das Canvas.
- **root.after(50, bewege_ball)** ruft die Funktion bewege_ball alle 50 Millisekunden auf, was eine einfache Animation erzeugt.

**bewege_ball()**: Diese Funktion bewegt den Ball auf dem Canvas. Die move-Methode des Canvas-Objekts wird verwendet, um den Ball um 5 Pixel nach rechts zu bewegen (0 Pixel in der vertikalen Richtung). 
Dann wird die after-Methode des root-Fensters verwendet, um die bewege_ball-Funktion nach einer Verzögerung von 50 Millisekunden erneut aufzurufen. Dadurch entsteht eine Animation, bei der der Ball kontinuierlich nach rechts bewegt wird.

**main()**: Diese Funktion erstellt das Hauptfenster und das Canvas, zeichnet den Ball auf dem Canvas und startet die Animation. Zuerst wird ein Tkinter-Hauptfenster erstellt und der Titel festgelegt. Dann wird ein Canvas-Widget erstellt und auf dem Hauptfenster platziert. M

it der create_oval-Methode des Canvas-Objekts wird ein Ball gezeichnet, der durch ein Oval repräsentiert wird. Der Ball hat die Koordinaten (10, 100) für den linken oberen Punkt des umschreibenden Rechtecks und (50, 140) für den rechten unteren Punkt des umschreibenden Rechtecks. Der Ball wird mit roter Farbe gefüllt. 

Schließlich wird die **bewege_ball**-Funktion aufgerufen, um die Animation zu starten.


````python
import random

def animiere_text():
    # Zufällige Farbe generieren
    zufaellige_farbe = "#{:02x}{:02x}{:02x}".format(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    # Textfarbe aktualisieren
    canvas.itemconfig(text_item, fill=zufaellige_farbe)

    # Animation fortsetzen
    root.after(500, animiere_text)

def main():
    global root, canvas, text_item
    root = tk.Tk()
    root.title("WBS Animation in Tkinter")

    canvas = tk.Canvas(root, width=400, height=300)
    canvas.pack()

    # Text "WBS" auf dem Canvas erstellen
    text_item = canvas.create_text(200, 150, text="WBS", font=("Arial", 40, "bold"))

    # Animation starten
    animiere_text()

    root.mainloop()

if __name__ == "__main__":
    main()
````

- Die Buchstaben "WBS" werden in einer großen, fettgedruckten Arial-Schrift in der Mitte des Canvas angezeigt.
- Die Funktion animiere_text ändert die Farbe des Textes alle 500 Millisekunden zu einer zufälligen Farbe.
- Die root.after Methode wird verwendet, um die Animation kontinuierlich laufen zu lassen.

Der Ausdruck "**{:02x}{:02x}{:02x}**" ist eine Formatierungsspezifikation in Python. Hier wird jeder {}-Platzhalter durch einen Wert ersetzt, der im .format()-Teil angegeben ist. Der Teil **:02x** in jedem Platzhalter gibt an, wie dieser Wert formatiert werden soll:
**02**: Dies bedeutet, dass der Wert als zweistellige Zahl dargestellt werden soll. Falls der Wert eine einstellige Zahl ist, wird er mit einer führenden Null aufgefüllt.

**x**: Dies gibt an, dass der Wert als Hexadezimalzahl dargestellt werden soll.

Generierung von Zufallszahlen: random.randint(0, 255) generiert eine zufällige ganze Zahl zwischen 0 und 255. Dieser Wertebereich entspricht dem Bereich der einzelnen Farbkomponenten in einem RGB-Farbmodell, wobei 0 kein Beitrag der Komponente und 255 der volle Beitrag ist.

RGB-Farbwerte: Die drei Aufrufe von random.randint(0, 255) generieren zufällige Werte für die Rot-, Grün- und Blau-Komponenten der Farbe.

Zusammenstellung der Farbe: Die generierten Rot-, Grün- und Blau-Werte werden zu einem Hexadezimalstring zusammengesetzt, der eine Farbe im Format #RRGGBB repräsentiert. Das # am Anfang ist Teil des üblichen Formats für Farben im Hexadezimalformat, das in vielen Grafiksystemen verwendet wird, einschließlich Tkinter.

## Tkinter Label mit StringVar

- Ein Tkinter-Fenster wird erstellt.
- Ein StringVar-Objekt wird initialisiert.
- Ein Label wird erstellt, dessen Text an das StringVar-Objekt gebunden ist.


#### Eine digitale Uhr mit StringVar:



````python
import time


def update_time():
    global time_var, root, label
    current_time = time.strftime("%H:%M:%S")
    time_var.set(current_time)

    root.after(1000, update_time)


def main():
    global time_var, root, label
    root = tk.Tk()
    root.title("Digital-Uhr")
    root.geometry("300x100")

    time_var = tk.StringVar()

    label = tk.Label(root, textvariable=time_var, font=("Comic Sans MS", 48))
    label.pack()

    update_time()

    root.mainloop()


if __name__ == "__main__":
    main()
````

Beispiel: Selbe Anwendung aber mit Nested Functions, dadurch spart man sich globale Variablen.

````python
import tkinter as tk
import time


def main():
    def update_time():
        # global time_var, root, label
        current_time = time.strftime("%H:%M:%S")
        time_var.set(current_time)

        root.after(1000, update_time)

    # global time_var, root, label
    root = tk.Tk()
    root.title("Digital-Uhr")
    root.geometry("300x100")

    time_var = tk.StringVar()

    label = tk.Label(root, textvariable=time_var, font=("Comic Sans MS", 48))
    label.pack()

    update_time()

    root.mainloop()


if __name__ == "__main__":
    main()
````

Was passiert hier?

- Die Funktion update_time aktualisiert den Wert der StringVar time_var mit der aktuellen Uhrzeit.
- Das Label ist an time_var gebunden und zeigt daher die aktuelle Zeit an.
- Die Funktion root.after(1000, update_time) sorgt dafür, dass update_time einmal pro Sekunde aufgerufen wird, wodurch die Uhrzeit jede Sekunde aktualisiert wird.
- Während dieses Prozesses bleibt das Tkinter-Fenster interaktiv und reagiert auf Benutzeraktionen.

Dieses Beispiel zeigt, wie man eine dynamisch aktualisierte Anzeige wie eine Uhr in Tkinter realisieren kann, wobei die Oberfläche reaktionsfähig und interaktiv bleibt.

## Layout-Management in Tkinter

Das Layout-Management in Tkinter bezieht sich darauf, wie Widgets innerhalb des Fensters angeordnet werden. Tkinter bietet mehrere Methoden für das Layout-Management: **pack**, **grid** und **place**.

- **Pack**: Ordnet Widgets vertikal oder horizontal an. Es ist einfach zu verwenden, bietet aber weniger Kontrolle als grid.

- **Grid**: Platziert Widgets in einem tabellenähnlichen Format. Es bietet eine feinere Kontrolle über die Platzierung von Widgets, wobei jede Zelle in der Tabelle ein Widget enthalten kann.

- **Place**: Ermöglicht es, Widgets an genauen Positionen anzugeben, bietet die höchste Kontrolle, ist aber am wenigsten flexibel in Bezug auf die Anpassung an verschiedene Fenstergrößen.

````python

def main():
    root = tk.Tk()
    root.title("Layoutmangement")
    root.geometry("300x100")

    frame_pack = tk.Frame(root, borderwidth=2, relief="solid")
    frame_pack.pack(side="left", fill="both", expand=True)

    label1 = tk.Label(frame_pack, text="Label 1 (Pack)")
    label1.pack()

    label2 = tk.Label(frame_pack, text="Label 2 (Pack)")
    label2.pack()

    frame_grid = tk.Frame(root, borderwidth=2, relief="solid")
    frame_grid.pack(side="right", fill="both", expand=True)

    label3 = tk.Label(frame_grid, text="Label 3 (Grid)")
    label3.grid(row=0, column=0)  # Reihe Spalte

    label4 = tk.Label(frame_grid, text="Label 4 (Grid)")
    label4.grid(row=1, column=0)

    label5 = tk.Label(frame_grid, text="Label 5 (Grid)")
    label5.grid(row=0, column=1)  # Reihe Spalte

    label6 = tk.Label(frame_grid, text="Label 6 (Grid)")
    label6.grid(row=1, column=1)



    root.mainloop()


if __name__ == "__main__":
    main()
````

#### Verwendung von **Pack**:

- **frame_pack** = tk.Frame(root, borderwidth=2, relief="solid"): Erstellt einen Rahmen (Frame), der als Container für andere Widgets dient. Der Rahmen hat eine Randbreite (borderwidth) von 2 und einen relief-Stil, der den Rahmen optisch hervorhebt.
- **frame_pack.pack(side="left", fill="both", expand=True)**: Platziert den frame_pack im Hauptfenster. **side="left"** bedeutet, dass der Rahmen auf der linken Seite positioniert wird. **fill="both"** lässt den Rahmen sowohl horizontal als auch vertikal expandieren. **expand=True** erlaubt dem Rahmen, zusätzlichen Raum im Fenster einzunehmen.
- **label1 = tk.Label(frame_pack, text="Label 1 (Pack)")**: Erstellt ein Label-Widget im frame_pack.
- **label1.pack()**: Platziert das Label-Widget innerhalb des frame_pack. Standardmäßig wird pack das Widget vertikal anordnen.

#### Verwendung von Grid:

- **frame_grid = tk.Frame(root, borderwidth=2, relief="solid")**: Erstellt einen weiteren Rahmen als Container für Grid-Layout-Widgets.
- **frame_grid.pack(side="right", fill="both", expand=True)**: Platziert den frame_grid auf der rechten Seite des Hauptfensters.
- **label3 = tk.Label(frame_grid, text="Label 3 (Grid)")**: Erstellt ein Label-Widget im frame_grid.
- **label3.grid(row=0, column=0)**: Platziert das label3-Widget in der ersten Zeile **(row=0)** und der ersten Spalte **(column=0)** des frame_grid.
- **label4 = tk.Label(frame_grid, text="Label 4 (Grid)")**: Erstellt ein weiteres Label-Widget.
- **label4.grid(row=1, column=1)**: Platziert das label4-Widget in der zweiten Zeile **(row=1)** und der zweiten Spalte **(column=1)**.



In Tkinter gibt es verschiedene relief-Stile, die das Aussehen der Ränder von Widgets wie Frames, Labels usw. verändern. Die relief-Eigenschaft bestimmt den Stil der 3D-Rahmen, die um das Widget herum gezeichnet werden. Hier sind die verfügbaren Stile:

**flat**: Kein Relief, der Rand erscheint flach.
**raised**: Ein erhabener Rand, der so aussieht, als ob das Widget über der Oberfläche schwebt.
**sunken**: Ein vertiefter Rand, der das Gegenteil von raised ist und so aussieht, als ob das Widget in die Oberfläche eingelassen wäre.
**groove**: Ein eingegrabener oder gekehlter Rand, der wie eine Rille oder ein Kanal aussieht.
**ridge**: Ein erhabener Rand, der einem Grat oder einer Kante ähnelt und das Gegenteil von groove ist.

````python
def main():
    root = tk.Tk()
    root.title("Layoutmangement")
    root.geometry("300x100")

    frame_pack = tk.Frame(root, borderwidth=2, relief="solid")
    frame_pack.pack(side="left", fill="both", expand=True)

    label1 = tk.Label(frame_pack, text="Label 1 (Pack)")
    label1.pack()

    label2 = tk.Label(frame_pack, text="Label 2 (Pack)")
    label2.pack()

    frame_grid = tk.Frame(root, borderwidth=2, relief="solid")
    frame_grid.pack(side="right", fill="both", expand=True)

    # Cell-Optionen festlegen:
    cell_opts = dict(borderwidth=1, relief="solid", padx=6, pady=3)
    # Ende

    label3 = tk.Label(frame_grid, text="Label 3 (Grid)", **cell_opts)
    label3.grid(row=0, column=0, sticky="nsew")

    label4 = tk.Label(frame_grid, text="Label 4 (Grid)", **cell_opts)
    label4.grid(row=1, column=0, sticky="nsew")

    label5 = tk.Label(frame_grid, text="Label 5 (Grid)", **cell_opts)
    label5.grid(row=0, column=1, sticky="nsew")

    label6 = tk.Label(frame_grid, text="Label 6 (Grid)", **cell_opts)
    label6.grid(row=1, column=1, sticky="nsew")

    # Größe und Ausdehnung konfigurieren:
    frame_grid.grid_columnconfigure(0, weight=1)
    frame_grid.grid_columnconfigure(1, weight=1)
    frame_grid.grid_rowconfigure(0, weight=1)
    frame_grid.grid_rowconfigure(1, weight=1)

    root.mainloop()


if __name__ == "__main__":
    main()
````

````python
def daten_uebertragen(input_name, input_rufnummer, output_name, output_rufnummer):
    # mit get() wird der aktuelle inhalt gelesen:
    name = input_name.get()
    nummer = input_rufnummer.get()

    # Bevor ich schreiben kann, muss ich die Ausgabe-Felder aktivieren:
    output_name.config(state='normal')
    output_rufnummer.config(state='normal')

    # Werte übertragen:
    output_name.delete(0, tk.END)  # Feld leeren
    # output_name.insert(0, name)  # Daten einfügen (am Anfang anfangen)
    output_name.insert(tk.END, name)  # Daten einfügen (am Ende anfangen)

    output_rufnummer.delete(0, tk.END)
    # output_rufnummer.insert(0, nummer)
    output_rufnummer.insert(tk.END, nummer)

    # und wieder deaktivieren:
    output_name.config(state='readonly')
    output_rufnummer.config(state='readonly')

def main():

    # Erstelle das Hauptfenster
    root = tk.Tk()
    root.title("Daten Transfer Beispiel")

    # Erstelle Rahmen für besseres Layout
    input_frame = tk.LabelFrame(root, text="Eingabe", padx=10, pady=10)
    input_frame.grid(row=0, column=0, padx=10, pady=5, sticky="nsew")

    output_frame = tk.LabelFrame(root, text="Ausgabe", padx=10, pady=10)
    output_frame.grid(row=1, column=0, padx=10, pady=5, sticky="nsew")

    # Eingabefelder
    tk.Label(input_frame, text="Name:").grid(row=0, column=0, sticky="w")
    name_eingabe = tk.Entry(input_frame)
    name_eingabe.grid(row=0, column=1, padx=5, pady=2)

    tk.Label(input_frame, text="Rufnummer:").grid(row=1, column=0, sticky="w")
    rufnummer_eingabe = tk.Entry(input_frame)
    rufnummer_eingabe.grid(row=1, column=1, padx=5, pady=2)

    # Übertragen Button
    button = tk.Button(input_frame, text="Daten übertragen",
              command=lambda: daten_uebertragen(name_eingabe,
                                                rufnummer_eingabe,
                                                name_ausgabe,
                                                rufnummer_ausgabe))
    button.grid(row=2, column=0, columnspan=2, pady=5)

    # Ausgabefelder
    tk.Label(output_frame, text="Name:").grid(row=0, column=0, sticky="w")
    name_ausgabe = tk.Entry(output_frame, state='readonly')
    name_ausgabe.grid(row=0, column=1, padx=5, pady=2)

    tk.Label(output_frame, text="Rufnummer:").grid(row=1, column=0, sticky="w")
    rufnummer_ausgabe = tk.Entry(output_frame, state='readonly')
    rufnummer_ausgabe.grid(row=1, column=1, padx=5, pady=2)


    # Starte die Anwendung
    root.mainloop()

if __name__ == "__main__":
    main()
````

## Dialoge und Nachrichten

Tkinter bietet verschiedene Möglichkeiten, Dialoge und Nachrichten anzuzeigen:

- **Messagebox**: Einfache Pop-up-Nachrichten für Information, Warnung, Fehler, usw.
- **Filedialog**: Zum Öffnen und Speichern von Dateien.
- **Simpledialog**: Für einfache Benutzereingaben.

Beispiel für Messagebox und Filedialog:

````python
import tkinter as tk
from tkinter import messagebox, filedialog, simpledialog


def zeige_messagebox():
    messagebox.showinfo("Info", "Hallo Welt!")

def oeffne_filedialog():
    dateiname = filedialog.askopenfilename(initialdir="/", title="Bitte wählen Sie eine Datei aus",
                                           filetypes=(("Textdateien", "*.txt"), ("Alle Dateien", "*.*")))

    if dateiname:
        label_dateiname.config(text=dateiname)

    else:
        label_dateiname.config(text="Nichts ausgewählt")

def speichern_filedialog():
    dateiname = filedialog.asksaveasfilename(initialdir="/",
                                             title="Bitte wählen Sie einen Dateinamen aus",
                                             filetypes=(("Textdateien", "*.txt"),
                                                        ("Markdown Dateien", "*.md"),
                                                        ("Alle Dateien", "*.*")),
                                             confirmoverwrite=True # Frage, ob Datei überschrieben werden soll
                                             )

    if dateiname:
        label_dateiname.config(text=dateiname)
    else:
        label_dateiname.config(text="Nichts ausgewählt")

def frage_nach_eingabe():
    # antwort = None
    antwort = simpledialog.askstring("Eingabe", "Bitte geben Sie etwas ein:")
    if antwort:
        messagebox.showinfo("Ihre Antwort",antwort)

def frage_ja_nein():
    antwort = messagebox.askyesno("Ja/Nein", "Sind Sie sicher?")

    if antwort:
        messagebox.showinfo("Ja", "Danke!")
    else:
        messagebox.showinfo("Nein", "Okay.")

def frage_ja_nein_abbrechen():
    antwort = messagebox.askyesnocancel("Speichern/Überschreiben", "Sind Sie sicher?")
    # Am Beispiel von Dateien:
    if antwort:
        # Vorhandener Dateiname wird verwendet
        messagebox.showinfo("Ja", "Ja wurde ausgewählt, Danke!")

    elif antwort is False:
        # Neuer Dateiname darf ausgewählt werden
        messagebox.showinfo("Nein", "Nein wurde ausgewählt. Okaaaay")

    else:
        # Keine Aktion, zurück
        messagebox.showinfo("Abbruch", "Dialog abgebrochen.")

def zeige_warnung():
    messagebox.showwarning("Warnung", "Bitte beachten Sie, \ndass diese Warnung eine Warnung ist!")


def main():
    global label_dateiname
    root = tk.Tk()
    root.title("Dialoge und Nachrichten")
    root.geometry("300x300")

    btn_messagebox = tk.Button(root, text="Zeige Messagebox", command=zeige_messagebox)
    btn_messagebox.pack()

    btn_filedialog = tk.Button(root, text="Öffne Filedialog", command=oeffne_filedialog)
    btn_filedialog.pack()

    btn_savefiledialog = tk.Button(root, text="Speichern Filedialog", command=speichern_filedialog)
    btn_savefiledialog.pack()

    btn_simpledialog = tk.Button(root, text="Öffne SimpleDialog", command=frage_nach_eingabe)
    btn_simpledialog.pack()

    btn_ja_nein = tk.Button(root, text="Frage Ja/Nein", command=frage_ja_nein)
    btn_ja_nein.pack()

    btn_ja_nein_abbrechen = tk.Button(root, text="Frage Ja/nein/Abbrechen", command=frage_ja_nein_abbrechen)
    btn_ja_nein_abbrechen.pack()

    btn_warning = tk.Button(root, text="Zeige Warnung", command=zeige_warnung)
    btn_warning.pack()

    label_dateiname = tk.Label(root, text="Keine Datei ausgewählt")
    label_dateiname.pack()

    root.mainloop()


if __name__ == "__main__":
    main()
````

In diesem Beispiel:

- **oeffne_filedialog**: Wenn eine Datei ausgewählt wird, aktualisiert die Funktion das label_dateiname mit dem Pfad der ausgewählten Datei.
- **frage_nach_eingabe**: Diese Funktion öffnet einen **simpledialog**, in dem der Benutzer einen String eingeben kann. Die eingegebene Antwort wird dann in einer Messagebox angezeigt.
- Das **label_dateiname-Widget** dient dazu, Informationen wie den Pfad der zuletzt ausgewählten Datei anzuzeigen. Damit können Sie dem Benutzer ein visuelles Feedback geben, welche Aktionen ausgeführt wurden.

### Programm ordnungsgemäß beenden

````python
import sys
import tkinter as tk
from tkinter import messagebox


def programm_beenden():
    global root

    if messagebox.askyesno("Beenden", "Wollen Sie das Programm wirklich beenden?"):
        root.quit()
        root.destroy()


def main():
    # Hauptfenster erstellen:
    global root
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")  # Fenster-Überschrift
    root.geometry("300x100")  # Fenster Größe


    # Schließen durch "X" in der oberen rechten Ecke abfangen:
    root.protocol("WM_DELETE_WINDOW", programm_beenden)  # Bei schließen wird programm_beenden()-Funktion  aufgerufen

    # Hauptmenü erstellen
    hauptmenu = tk.Menu(root)
    root.config(menu=hauptmenu)

    # Datei-Menü
    datei_menu = tk.Menu(hauptmenu, tearoff=0)
    hauptmenu.add_cascade(label="Datei", menu=datei_menu)

    datei_menu.add_command(label="Beenden", command=programm_beenden)


    # Hauptschleife (Mainloop):
    root.mainloop()


if __name__ == "__main__":
    main()
````

````python
import sys
import tkinter as tk
from tkinter import messagebox


def programm_beenden():
    global root

    if messagebox.askyesno("Beenden", "Wollen Sie das Programm wirklich beenden?"):
        root.quit()
        root.destroy()


def main():
    # Hauptfenster erstellen:
    global root
    root = tk.Tk()
    root.title("Einfaches TKInter-Fenster")  # Fenster-Überschrift
    root.geometry("300x100")  # Fenster Größe


    # Schließen durch "X" in der oberen rechten Ecke abfangen:
    root.protocol("WM_DELETE_WINDOW", programm_beenden)  # Bei schließen wird programm_beenden()-Funktion  aufgerufen

    # Hauptmenü erstellen
    hauptmenu = tk.Menu(root)
    root.config(menu=hauptmenu)

    # Datei-Menü
    datei_menu = tk.Menu(hauptmenu, tearoff=0)
    hauptmenu.add_cascade(label="Datei", menu=datei_menu)

    datei_menu.add_command(label="Beenden", command=programm_beenden)


    # Hauptschleife (Mainloop):
    root.mainloop()


if __name__ == "__main__":
    main()
````

### tk vs. ttk (Themed TKinter)
TK ist ein wenig eingestaubt, hat aber trotzdem seinen Charme. Ein wenig Verbesserung bietet ttk https://docs.python.org/3/library/tkinter.ttk.html.

Richtig schick und bunt wird es beispielsweise mit ttkbootstrap (https://ttkbootstrap.readthedocs.io/en/latest/) oder customtkinter (https://customtkinter.tomschimansky.com/).

````python
import tkinter as tk
from tkinter import ttk
import tkinter.ttk as tk_progress  # Für tk Progressbar

def button_click():
    print("Button wurde geklickt!")

# Hauptfenster erstellen
root = tk.Tk()
root.title("tk vs ttk Widgets Vergleich")
root.geometry("600x500")

# Überschrift
tk.Label(root, text="Vergleich von tk und ttk Widgets",
         font=("Arial", 16)).pack(pady=10)

# Frame für tk-Widgets
tk_frame = tk.LabelFrame(root, text="tk Widgets", padx=10, pady=10)
tk_frame.pack(side=tk.LEFT, padx=10, pady=10, fill=tk.BOTH, expand=True)

# Frame für ttk-Widgets
ttk_frame = tk.LabelFrame(root, text="ttk Widgets", padx=10, pady=10)
ttk_frame.pack(side=tk.RIGHT, padx=10, pady=10, fill=tk.BOTH, expand=True)

# tk-Widgets
tk.Label(tk_frame, text="tk Label", font=("Arial", 12)).pack(pady=5)
tk.Button(tk_frame, text="tk Button", command=button_click).pack(pady=5)
tk.Entry(tk_frame, width=20).pack(pady=5)
tk.Checkbutton(tk_frame, text="tk Checkbox").pack(pady=5)

# ttk-Widgets
ttk.Label(ttk_frame, text="ttk Label", font=("Arial", 12)).pack(pady=5)
ttk.Button(ttk_frame, text="ttk Button", command=button_click).pack(pady=5)
ttk.Entry(ttk_frame, width=20).pack(pady=5)
ttk.Checkbutton(ttk_frame, text="ttk Checkbox").pack(pady=5)

# tk Schieberegler (Scale)
tk.Label(tk_frame, text="tk Schieberegler").pack(pady=5)
tk_slider = tk.Scale(tk_frame, from_=0, to=100,
                     orient=tk.HORIZONTAL, length=200)
tk_slider.pack(pady=5)

# ttk Schieberegler
ttk.Label(ttk_frame, text="ttk Schieberegler").pack(pady=5)
ttk_slider = ttk.Scale(ttk_frame, from_=0, to=100,
                       orient=tk.HORIZONTAL, length=200)
ttk_slider.pack(pady=5)

# tk Fortschrittsbalken
tk.Label(tk_frame, text="tk Fortschrittsbalken").pack(pady=5)
tk_progress_bar = tk_progress.Progressbar(tk_frame, length=200,
                                mode='determinate', maximum=100)
tk_progress_bar.pack(pady=5)
tk_progress_bar['value'] = 50  # Setze Fortschritt auf 50%

# ttk Fortschrittsbalken
ttk.Label(ttk_frame, text="ttk Fortschrittsbalken").pack(pady=5)
ttk_progress = ttk.Progressbar(ttk_frame, length=200,
                                mode='determinate', maximum=100)
ttk_progress.pack(pady=5)
ttk_progress['value'] = 50  # Setze Fortschritt auf 50%

root.mainloop()
````
