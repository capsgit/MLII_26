Den Ordner mit der UI (dem Userinterface) muss in das Stammverzeichnis des Projekts kopiert werden. Es gibt verschiedene Möglichkeiten mit dem Code weiterzuarbeiten. In diesen Abschnitt
fahren wir fort, indem wir den C++-Code in Python-Code umwandeln.
Dazu brauchen wir ein kleines Skript:

```python
from qtpy import uic

uic.compileUiDir("ui")
print("Ausgeführt...")# Ihr Code hier !!!
```

Nachfolgend das eigentliche Skript:

```python
from qtpy import QtWidgets
import sys
from ui.mainwindow import Ui_MainWindow


class MainWindow(QtWidgets.QMainWindow):
    def __init__(self):
        super().__init__(parent=None)
        self.ui = Ui_MainWindow()

        self.ui.setupUi(self)
        self.ui.pbgruss.clicked.connect(self.helloworld)  # Der Button wird mit der Funktion verknüpft


    def helloworld(self):
        self.ui.tlgruss.setText("Hallo Welt!")  # Ändert den Text des Textlabels


def main():
    app = QtWidgets.QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())

if __name__ == "__main__":
    main()
```

Mit dem folgenden Code arbeiten wir weiter... <br>

Jetzt versehen wir den Button (**pbgruss**) mit der Funktion, das er Text in das Label (**legruss**) schreiben soll:

```python
from qtpy import QtWidgets
import sys
from ui.mainwindow import Ui_MainWindow


class MainWindow(QtWidgets.QMainWindow):
    def __init__(self):
        super().__init__(parent=None)
        self.ui = Ui_MainWindow()

        self.ui.setupUi(self)


def main():
    app = QtWidgets.QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())

if __name__ == "__main__":
    main()

```
