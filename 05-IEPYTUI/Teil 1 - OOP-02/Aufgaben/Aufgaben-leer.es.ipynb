## Tarea 1 Clases y objetos:
Cree una clase de Python llamada "persona." Esta clase debería tener atributos de instancia para el **Nombre** y la **edad** de una persona. A continuación, cree dos objetos de persona y dé su información.

```python
class Person:
    def __init__(self, name: str, age: str):
        self.name = name
        self.age = age

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        if not name:
            raise ValueError("name no puede ser vacío")
        self.__name = name

    @property
    def age(self):
        return self.__age
    @age.setter
    def age(self, age):
        if not age:
            raise ValueError("age no puede ser vacío")
        self.__age = int(age)


p1 = Person("Max", "25")
p2 = Person("Peter", "83")

print(p1.name)
print(p2.name)
```

## Tarea 3: Atributos de clase y métodos de clase

Cree una clase "**cuenta bancaria**" con un atributo de clase " ** tipo de interés**" y un método de clase, "** tipos de interés_calculados**," que calcula el importe del interés por un saldo dado de la cuenta. A continuación, cree dos objetos de cuenta bancaria y acceda al método de cálculo de los intereses.

```python
class Bankkonto:
    zinssatz = 0.015

    def __init__(self, kontostand: float):
        self.kontostand = float(kontostand)

    @classmethod
    def berechne_zinsen(cls, kontostand: float) -> float:
        return float(kontostand) * cls.zinssatz

k1 = Bankkonto(18000)
k2 = Bankkonto(2500)

print(f"Konto 1 Zinsen: {k1.berechne_zinsen(k1.kontostand):.2f}, €")
print(f"Konto 2 Zinsen: {k2.berechne_zinsen(k2.kontostand):.2f}, €")
```
