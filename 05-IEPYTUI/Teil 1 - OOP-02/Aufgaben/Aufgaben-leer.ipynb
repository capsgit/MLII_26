## Aufgabe 1 Klassen und Objekte:
Erstellen Sie eine Python-Klasse namens "Person". Diese Klasse sollte Instanzattribute für den **Namen** und das **Alter** einer Person haben. Erstellen Sie dann zwei Person-Objekte und gib ihre Informationen aus.

```python
class Person:
    def __init__(self, name: str, age: str):
        self.name = name
        self.age = age

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        if not name:
            raise ValueError("name no puede ser vacío")
        self.__name = name

    @property
    def age(self):
        return self.__age
    @age.setter
    def age(self, age):
        if not age:
            raise ValueError("age no puede ser vacío")
        self.__age = int(age)


p1 = Person("Max", "25")
p2 = Person("Peter", "83")

print(p1.name)
print(p2.name)
```

## Aufgabe 3: Klassenattribute und Klassenmethoden

Erstellen Sie eine Klasse "**Bankkonto**" mit einem Klassenattribut "**zinssatz**" und einer Klassenmethode "**berechne_zinsen**", die den Zinsbetrag für ein gegebenes Kontoguthaben berechnet. Erstellen Sie dann zwei Bankkonto-Objekte und rufe die Methode zur Berechnung der Zinsen auf.

```python
class Bankkonto:
    zinssatz = 0.015

    def __init__(self, kontostand: float):
        self.kontostand = float(kontostand)

    @classmethod
    def berechne_zinsen(cls, kontostand: float) -> float:
        return float(kontostand) * cls.zinssatz

k1 = Bankkonto(18000)
k2 = Bankkonto(2500)

print(f"Konto 1 Zinsen: {k1.berechne_zinsen(k1.kontostand):.2f}, €")
print(f"Konto 2 Zinsen: {k2.berechne_zinsen(k2.kontostand):.2f}, €")
```
