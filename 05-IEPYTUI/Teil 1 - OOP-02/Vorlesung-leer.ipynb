# Datenkapselung

Unter Datenkapselung versteht man den Schutz der Daten eines Objekts vor ***direktem Zugriff***

Zugriff auf Objektdaten/- attribute soll nur über sogenannte ***getter*** und ***setter*** erfolgen.
Unter Geheimnisprinzip versteht man das "verstecken" interner Implementierungsdetails, damit sie von außen nicht sichtbar sind.

### Python kennt 3 Stufen von Datenabstraktion: public, protected, private
***public***
```name```
Attribut ohne führende Unterstriche.
Sie sind innerhalb einer Klasse und auch von außen les- und schreibbar.<br>

***protected***
```_name```
Attribut mit einem führenden Unterstrich.
Man kann zwar von außen lesend und schreibend zugreifen, der Entwickler macht aber klar das man diese Attribute so nicht
benutzen sollte; Protected-Attribute sind beim Importieren wichtig.<br>

***private***
```__name```
Attribut mit zwei führenden Unterstrichen.
Sie sind von außen weder sichtbar noch benutzbar.<br><br>

````python
from os import name


class Artikel:
    def __init__(self, nr, name, preis, bestand):
        self.__nr = nr  # private
        self.__name = name  # private
        self._preis = preis  # protected
        self.bestand = bestand  # public


apfel = Artikel(1, "Apfel", 0.30, 10)

print(apfel.bestand)  # funktioniert
print(apfel._preis)  # funktioniert
# print(apfel.__name)  # funktioniert nicht
# print(apfel.__nr)  # funktioniert nicht
````

## Arbeiten mit Getter und Setter

Nachdem objektorientierten Prinzip der Kapselung gilt, das alle Attribute vor unkontrolliertem Zugriff von außen geschützt werden sollen.
Das heißt, es werden Methoden gebraucht, die den Zugriff auf Attribute steuern können. Zum Beispiel zur Prüfung, ob die Zuweisung korrekt ist.
Dies sind die sogenannten Getter und Setter. Getter werden verwendet, um Werte abzufragen, Setter um Werte zuzuweisen.

````python
class Artikel:
    def __init__(self, artikelnummer, name, preis, bestand):
        self.__artikelnummer = artikelnummer  # private
        self.__name = name  # private
        self._preis = preis  # protected
        self.bestand = bestand  # public

    def get_nummer(self):
        return self.__artikelnummer

    def set_nummer(self,nummer):
        self.__artikelnummer = nummer


apfel = Artikel(123456, "Apfel", 0.30, 10)

print(apfel.get_nummer())
apfel.set_nummer(654321)
print(apfel.get_nummer())
````

Die bevorzugte Variante in Python ist allerdings das Verwenden von **Properties** (Eigenschaften).

````python
class Artikel:
    def __init__(self, artikelnummer, name, preis, bestand):
        self.__artikelnummer = artikelnummer
        self.__name = name
        self.__preis = preis
        self.__bestand = bestand

    @property  # Decorator
    def preis(self):
        return self.__preis * 1.19  # Preis wird natürlich aus einer Datenbank gezogen

    @preis.setter  # Decorator
    def preis(self, neuer_preis):
        if type(neuer_preis) != float:
            print("Ungültiger Preis!")
        else:
            self.__preis = neuer_preis
            # Neuer Preis wird dann in der Datenbank gespeichert

apfel = Artikel(123456, "Apfel", 0.30, 10)

apfel.preis = 0.40
print(apfel.preis)
````

Weiteres Beispiel:

````python
class Test:
    def __init__(self):
        self.__daten = {1, 2, 3}

    @property
    def daten(self):
        return self.__daten
````

Problematik: Ich kann die Daten-Auflistung nun von außen bearbeiten.

````python
t = Test()
print("Original: " + str(t.daten))

verweis = t.daten
verweis.add(6)
print("Verändert über den Verweis: " + str(verweis))

print("Nun auch verändert im Objekt: " + str(t.daten))
````

Das Beispiel zeigt, dass das Attribut nun über den Verweis (verweis) von außen manipuliert werden kann. <br>
Deshalb ist es wichtig, dass der Getter nicht das Attribut selbst, sondern eine Kopie zurückgibt:

````python
class Test:
    def __init__(self):
        self.__daten = {1, 2, 3}
    @property
    def daten(self):
        return self.__daten.copy()


t = Test()
print("Original: " + str(t.daten))

verweis = t.daten
verweis.add(6)
print("Die Kopie wurde verändert: " + str(verweis))

print("Im Objekt aber nicht: " + str(t.daten))
````

Eine weitere Möglichkeit, Properties zu erstellen, ist durch die `property`-Funktion

````python
class Frucht:
    def __init__(self, name, farbe):
        self.__name = name
        self.__farbe = farbe

    def get_name(self):
        """Getter-Methode"""
        return self.__name

    def set_name(self, name):
        """Setter-Methode"""
        self.__name = name

    # Beides durch property-Funktion verbinden:
    name = property(get_name, set_name)


apfel = Frucht("Apfel", "rot")

print(apfel.name)  # Zugriff über property
apfel.name = "Braeburn"  # Zugriff über property
print(apfel.name)  # Zugriff über property
````

## Überladen von Operatoren

Beispiel: Additionsoperator passt sich den Datentypen an

````python
z1 = 5
z2 = 3.14
print(z1 + z2)

w1 = "Hallo"
w2 = "Welt"
print(w1 +" "+ w2)
````

Diese Anpassung der Operatoren an den jeweiligen Datentyp ist eigentlich schon eine Überladung von Opereratoren. Je nachdem, welcher Datentyp benutzt wird, hat der Operator eine andere Funktionalität. <br>
**Überladen heißt also, dem Operator eine neue Bedeutung zu geben.**

Zunächst wollen wir eine Klasse Bruch erstellen, die einen mathematischen Bruch definiert. <br>
Diese Bruchklasse kann Zähler und Nenner verwalten.

Prototyp der Klasse ***Bruch***

````python
class Bruch:
    def __init__(self, zaehler, nenner):
        self.zaehler = zaehler
        self.nenner = nenner

    def ausgabe(self):
        print(self.zaehler / self.nenner)
````

**Addition von Brüchen**

````python
a = Bruch(1,2)
b = Bruch(3,4)

c= a + b       # esto no funciona por la "class"
c.ausgabe()
````

Das Beispiel zeigt, dass der Addition-Operator nicht für die Bruchklasse definiert ist.

Die Lösung des Problems ist eine Methode, die den Additionsoperator für die Bruchklasse definiert. 
Solche Methoden sind in Python bereits als built-in Methoden vorhanden und heißen **magische Methoden** (magic methods).
Eine der magischen Methoden ermöglicht es uns zum Beispiel, dass die Addition von 2 Bruch-Objekten durchgeführt werden kann.

Für den Addition-Operator nennt sich diese magische Methode **```__add__(self, instanz_2)```**.

Beispielanwendung für unsere Klasse Bruch

````python
class Bruch:
    def __init__(self, zaehler, nenner):
        self.zaehler = zaehler
        self.nenner = nenner

    def ausgabe(self):
        print(self.zaehler, "/", self.nenner)

    def __add__(self, other):
        neu_zaehler = self.zaehler * other.nenner + other.zaehler * self.nenner
        neu_nenner = self.nenner * other.nenner
        return Bruch(neu_zaehler, neu_nenner)

a = Bruch(1,2)
b = Bruch(3,4)

c= a + b       # esto nmo funciona por la "class"
c.ausgabe()
````

Alternative:

````python
class Bruch:
    def __init__(self, ü_zaehler, ü_nenner):
        self.zaehler = ü_zaehler
        self.nenner = ü_nenner

    def kuerzen(self):
        def ggT(a, b):
            while b != 0:
                a, b = b, a % b
            return a
        teiler = ggT(self.zaehler, self.nenner)
        self.zaehler = self.zaehler // teiler
        self.nenner = self.nenner // teiler

    def ausgabe(self):
        self.kuerzen()
        print(self.zaehler, "/", self.nenner)

    def __add__(self, other):
        neu_zaehler = self.zaehler * other.nenner + other.zaehler * self.nenner
        neu_nenner = self.nenner * other.nenner
        return Bruch(neu_zaehler, neu_nenner)

a = Bruch(1, 2)
b = Bruch(3, 4)

c = a + b
c.ausgabe()
````

# Vererbung/ Inheritence

Ein weiteres Konzept in der OOP ist die Vererbung.
Dabei erbt die Unterklasse alle Attribute und Methoden der Hauptklasse / Basisklasse und hat in der Regel noch weitere, eigene Attribute und Methoden.
Aus diesem Grund bezeichnet man auch die Hauptklassen / Basisklassen als ***allgemeine*** und die Unterklassen als ***spezialisierte*** Klassen.

## Einfache Vererbung

````python
class Tier:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe

    def schlafen(self, dauer):
        print(self.name, "duerme ahora", dauer, "minutos")

    def spiel(self, dauer, juego_name):
        print(self.name, "juega ahora", juego_name, dauer, "minutos")

    def sonido(self,):
        print("Au Au Au")

class Essl(Tier):
    def __init__(self, rasse, name, farbe, gewicht, augenfarbe):
        super().__init__(rasse, name, farbe)
        self.gewicht = gewicht
        self.augenfarbe = augenfarbe

t1 = Tier("collin", "Tony", "blau")
e1 = Essl("boyaco", "Burro", "grey", 10, "blue")

e1.schlafen(10)
e1.spiel(90, "futbol")
print(t1.name)
print(e1.name)
````

## Konstruktorverkettung
Eine Klasse initialisiert immer nur die eigenen Attribute.
Basis deklariert *variable1* und *variable2*, also ist der Konstruktor dieser Klasse dafür zuständig, diese Attribute auch zu beschreiben.
Sub deklariert *variable3* und damit ist Sub dafür zuständig, *variable3* zu beschreiben.

````python
class Basis:
    def __init__(self, variable1, variable2):
        self.variable1 = variable1
        self.variable2 = variable2

class Sub(Basis):
    def __init__(self, variable1, variable2, variable3):
        super().__init__(variable1, variable2)
        self.variable3 = variable3

# Objekt Basis
basis = Basis(1, "A")
print("Aus Basis:", basis.variable1, basis.variable2)

# geht nicht, weil Basis die Variable nicht enthält
# print(basis.variable3)

# Objekt der Klasse Sub
sub = Sub(2, "B", "C")
print("Aus Sub:", sub.variable1, sub.variable2, sub.variable3)
````

Dank der **Konstruktorverkettung** werden die übergebenen Werte für Sub an den Konstruktor von Basis weitergereicht und dort initialisiert. Der Konstruktor von Sub muss anschließend nur noch den Wert für das eigene Attribut zuweisen.

## Überladen von Methoden
#### Beispiel:
Anstelle von <i>bellen</i> und <i>miauen</i> soll es nur ****laut_machen**** geben.
Die Methode laut_machen soll aber bei Hund und Katze unterschiedlich benutzt(überladen) werden.

````python
class Tier:
    def __init__(self, rasse, name, farbe):
        self.rasse = rasse
        self.name = name
        self.farbe = farbe

    def schlafen(self, dauer):
        print(self.name, "schläft jetzt seit", dauer, "Minuten ")

    def spielen(self, dauer):
        print(self.name, "spielt jetzt", dauer, "Minuten ")

    def laut_machen(self):
        pass


#Klasse Katze erbt von Klasse Tier
class Katze(Tier):
    def __init__(self, rasse, name, farbe):
        super().__init__(rasse, name, farbe)

    def maus_fangen(self, anzahl = 1):
        print(self.name, "hat jetzt schon", anzahl, "Mäuse gefangen ")

    def laut_machen(self):
        print("miau")


# Klasse Hund erbt von Klasse Tier
class Hund(Tier):
    def __init__(self, rasse, name, farbe):
        super().__init__(rasse, name, farbe)

    def laut_machen(self):
        print("wuff")


# Objekt Katze
katerPaul = Katze("Mix", "Paul", "getigert")
hundAnrei = Hund("Germanischer Bärenhund", "Anrei", "kupfer")

# Testen der Unterklassen Katze und Hund

katerPaul.laut_machen()
hundAnrei.laut_machen()
````

#### Im folgenden Beipiel werden 3 Python Dateien erstellt: <br>
1. Die Basisklasse <i> Mitarbeiter.py </i>
2. Die Unterklasse <i> Angestellter.py </i>
3. Das Hauptprogramm <i> Main.py </i>

````python
# Mitarbeiter.py

class Mitarbeiter():
    # BasisKlasse für alle Angestellten Klassn

    anzahl = 0
    def __init__(self, name, gehalt):
        self.name = name
        self.gehalt = gehalt
        Mitarbeiter.anzahl += 1

    @staticmethod # statische Methode - ähnlich wie Klassenattribute ohne Objektverweis aufrufbar
    def zeige_anzahl():
        return "Gesamt Mitarbeiter %d" % Mitarbeiter.anzahl

    def __str__(self):
        return "Name: " + self.name + " \nGehalt: " + str(self.gehalt)

# ENDE der Klassendefinition

# Testobjekt
chef = Mitarbeiter("Markus Boss", 120000)

# Testlauf der Klasse
print(chef)
print(Mitarbeiter.zeige_anzahl())
````

````python
# Angestellter.py

#from Mitarbeiter import Mitarbeiter

class Angestellter(Mitarbeiter):
    # Angestellter erbt von Mitarbeiter

    def __init__(self, name, gehalt, projektnummer):
      # Aufruf des übergeordneten Konstruktors der Elternklasse mit den Parametern
      super().__init__(name, gehalt)
      self.projektnummer = projektnummer # neues Attribut für die Angestellten Klasse


    def __str__(self):
      return "Ausgabe Angestellte: \nName: " + self.name + " \nGehalt: " + str(self.gehalt) + "\nProjektnummer: " + str(self.projektnummer)

# ENDE der Klassendefinition

# Testobjekt
angestellt = Angestellter("Evelin Müller", 34000, 12345)

# Testlauf der Klasse
print(angestellt)
print(angestellt.zeige_anzahl())
````

````python
# Hauptprogramm
# Main.py

#from Angestellter import Angestellter, angestellt

# Objekte erstellen
angest1 = Angestellter("Zara Svennson", 20000, 3456)
angest2 = Angestellter("Manni Deutsch", 50000, 678)

# Ausführen verschiedener Methoden
print(angest1)
print(angest2)

angest2.gehalt = 44300
print(angest2)

print(angestellt.zeige_anzahl())
````

## Assoziation
Eine Assoziation ist eine „hat“-Beziehung zwischen zwei Klassen. Beispielsweise kann eine Klasse Abteilungsleiter, die ebenfalls von Mitarbeiter erbt, in Beziehung zu einer Klasse Büro stehen. Klassen müssen untereinander nicht zwangsweise in Beziehung zueinander stehen. Sollte aber in der Realität eine Beziehung wie „Ein Abteilungsleiter hat ein Büro“ oder „Zu einem Abteilungsleiter gehört ein Büro“ oder „Ein Büro wird von einem Abteilungsleiter besetzt“ etc vorkommen, würden wir diese Klassen auch im Programm miteinander verbinden.

````python
class Buero:
    def __init__(self, nr):
        self.zimmernummer = nr

    def __str__(self):
        return self.zimmernummer

class Mitarbeiter:
    def __init__(self, name, buero):
        self.name = name
        self.buero = buero

    def __str__(self):
        return f"{self.name}, {self.buero}"


buero1 = Buero("123")

hr_meier = Mitarbeiter("Herr Meier", buero1)

print(hr_meier)
````

## Mehrfachvererbung
Die Mehrfachvererbung unterscheidet sich von der einfachen Vererbung dadurch, dass eine Klasse direkt von mehreren Klassen erbt.
Dadurch kann ein Softwaresystem noch realitätsnäher umgesetzt werden.
Mehrfachvererbung ist deutlich komplexer; es müssen mehr Aspekte beachtet werden.
C# und Java unterstützen keine Mehrfachvererbung. Außer Python sind dies nur C++, CLOS, Eiffel und Perl.


Was erwarten Sie, ist die Ausgabe des Programms?

````python
class Basis1:
    def __init__(self):
        print("Konstruktor von Basis1")

class Basis2:
    def __init__(self):
        print("Konstruktor von Basis2")

class Mehrfach(Basis1, Basis2):
    def __init__(self):
        print("Konstruktor von Mehrfach")
        print("Basisklassenkonstruktor", end = " \n")
        super().__init__()

m = Mehrfach()

# b1 = Basis1()
# b2 = Basis2()
````

**Durch die Anwendung von ```super()```wird der Basisklassenkonstruktor von Basis_1 aufgerufen. Bei Mehrfachvererbung verweist ```super()``` immer auf die zuerst angegebene Basisklasse.**
<br>Möchte man auch auf die Methoden der anderen Basisklassen zugreifen. so muss das duch explizite Angabe des Klassennamens geschehen.

````python
class Basis1:
    def __init__(self):
        print("Konstruktor von Basis1")

class Basis2:
    def __init__(self):
        print("Konstruktor von Basis2")

class Mehrfach(Basis1, Basis2):
    def __init__(self):
        print("Konstruktor von Mehrfach")
        print("Basisklassenkonstruktor", end = " \n")
        Basis1.__init__(self)
        Basis2.__init__(self)

m = Mehrfach()
````

````python
class Basis1:
    def __init__(self, name):
        self.name = name
        print(f"[Basis1] name = {self.name}")


class Basis2:
    def __init__(self, version):
        self.version = version
        print(f"[Basis2] version = {self.version}")


class Mehrfach(Basis1, Basis2):
    def __init__(self, name, version, info):
        # Alle Argumente explizit weitergeben
        Basis1.__init__(self, name)
        Basis2.__init__(self, version)
        self.info = info
        print(f"[Mehrfach] info = {self.info}")


# Instanz erzeugen
m = Mehrfach("MeineBasisklasse", "1.2.3", "Dies ist die abgeleitete Klasse")

# Ausgabe der Attribute
print("\nAttribute der Instanz:")
print(f"name: {m.name}")
print(f"version: {m.version}")
print(f"info: {m.info}")
````

````python
# Erste Basisklasse: Definiert die Eigenschaften eines Landfahrzeugs
class Landfahrzeug:
    def __init__(self, motorleistung_ps):
        self.motorleistung_ps = motorleistung_ps
        print(f"[Landfahrzeug] Motorleistung: {self.motorleistung_ps} PS")

    def fahren(self):
        print(f"Das Fahrzeug fährt auf der Straße mit {self.motorleistung_ps} PS.")

# Zweite Basisklasse: Definiert die Eigenschaften eines Wasserfahrzeugs
class Wasserfahrzeug:
    def __init__(self, tiefgang_meter):
        self.tiefgang_meter = tiefgang_meter
        print(f"[Wasserfahrzeug] Tiefgang: {self.tiefgang_meter} m")

    def schwimmen(self):
        print(f"Das Fahrzeug schwimmt im Wasser mit einem Tiefgang von {self.tiefgang_meter} m.")

# Abgeleitete Klasse: Erbt von beiden Basisklassen
class Amphibienfahrzeug(Landfahrzeug, Wasserfahrzeug):
    def __init__(self, motorleistung_ps, tiefgang_meter, modell_name):
        # Initialisiert die erste Basisklasse
        Landfahrzeug.__init__(self, motorleistung_ps)
        # Initialisiert die zweite Basisklasse
        Wasserfahrzeug.__init__(self, tiefgang_meter)

        # Eigenes Attribut der abgeleiteten Klasse
        self.modell_name = modell_name
        print(f"[Amphibienfahrzeug] Modell: {self.modell_name}")

# --- Anwendung ---

# Eine Instanz des Amphibienfahrzeugs erstellen
amphicar = Amphibienfahrzeug(150, 1.2, "Sealegs 7.1")

print("\n--- Fähigkeiten und Attribute der Instanz ---")
# Zugriff auf Attribute beider Elternteile und der eigenen Klasse
print(f"Modell: {amphicar.modell_name}")
print(f"Leistung: {amphicar.motorleistung_ps} PS")
print(f"Tiefgang: {amphicar.tiefgang_meter} m")

# Aufruf der Methoden von beiden Elternteilen
print("\nAktionen:")
amphicar.fahren()
amphicar.schwimmen()
````

## Polymorphie

Polymorphie (Vielgestaltigkeit) ist einer der Grundpfeiler in der objektorientierten Programmierung.<br>
Mit Polymorphie ist gemeint, dass verschiedene Objekte beim Aufruf der gleichen Methode oder Operator unterschiedliche Aktionen durchführen, angepasst an das jeweilige Objekt.
Damit soll sichergestellt werden, dass alle Objekte aus der Vererbungshierarchie sich ähnlich (aber angepasst) verhalten.

Wir überschreiben hier die ausgabe-Methode aus der Basisklasse mit einer angepassten ausgabe-Methode in der Subklasse

````python
class Tier:
    def ausgabe(self):
        print("Ich bin ein Tier!")

class Baer(Tier):
    def ausgabe(self):
        print("Ich bin ein Bär!")

class Hund(Tier):
    def ausgabe(self):
        print("Ich bin ein Hund!")

def ausgabe_tier(tierart):
    tierart.ausgabe()


anrei = Hund()
balu = Baer()


ausgabe_tier(anrei)
ausgabe_tier(balu)
````

## Abstrakte Klassen

**Abstrakte Klassen sind Klassen, die nicht instanziert werden können**, also kein Objekt erzeugt werden kann. Sie dienen lediglich 
als Basis für andere Klassen und definieren eine allgemeine Struktur, ohne dass explizite Instanzen mit konkreten Eigenschaften 
und Verhalten entstehen. In Python sind Abstrakte Klassen durch die Verwendung der Abstract-Klausel oder einer init-Methode
mit keinerlei Implementierung definiert werden.

Wir beginnen mit der Definition der Basisklasse Tier als abstraktem Klass (abstract base class):

In der Basisklasse Tier, definieren wir die abstrakten Eigenschaften name, anzahl_beinen, and laut. Die Unterklassen Vogel und Hund implementieren diese Eigenschaften entsprechend.

````python
from abc import ABC, abstractmethod

class Tier(ABC):
    @property
    @abstractmethod
    def name(self):
        pass

    @abstractmethod
    def anzahl_beine(self):
        pass

    @abstractmethod
    def laut(self):
        pass


class Vogel(Tier):
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @property
    def anzahl_beine(self):
        return 2

    @property
    def laut(self):
        return "Pieps!"


class Hund(Tier):
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @property
    def anzahl_beine(self):
        return 4

    @property
    def laut(self):
        return "Wuff!"
````

Jetzt können wir Instanzen der konkreten Klassen erstellen:

````python
papagei = Vogel("Papagei")
print(papagei.name)
print(papagei.anzahl_beine)
print(papagei.laut)
````

In abstrakten Basisklassen kann @property mit @abstractmethod kombiniert werden, um das Vorhandensein einer Eigenschaft in Unterklassen zu erzwingen.

Wenn in der Subklasse auf das @property verzichtet wird, wird aus der Methode eine "normale" Methode, und diese muss mit Klammern ausgeführt werden.
