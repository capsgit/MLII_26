### Was ist Flask?

Flask ist ein Micro-Framework für Python, das bedeutet, dass es minimalistisch ist und nur die grundlegenden Funktionen bereitstellt, die zum Erstellen einer Webanwendung benötigt werden.
Es folgt dem WSGI-Standard (Web Server Gateway Interface) und verwendet Jinja2 als Template-Engine.<br>

#### Installation von Flask

Um Flask zu installieren, verwenden Sie pip, das Python-Paketverwaltungstool. Stellen Sie sicher, dass Sie eine virtuelle Umgebung verwenden, um Abhängigkeiten zu isolieren.<br>

<code>pip install Flask</code><br>

** Hinweis**: Der Code für die Übungen kann **nicht** innerhalb der Jupyter Notebooks ausgeführt werden.

Trotzdem ist es sinnvoll den Code in das Notebook zu kopieren um eine Unterlage zu schaffen. 


```python
 from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello():
    return "Hallo Welt!"

if __name__ == "__main__":
    app.run(debug=True)
```

Erklärung des Codes:

<code>Flask(__name__)</code>: Erstellt eine neue Flask-Anwendung. Der Parameter <code>__name__</code> hilft Flask, die Ressourcen wie Vorlagen und statische Dateien zu finden.<br>
<code>@app.route('/')</code>: Ein Dekorator, der eine URL-Route definiert. In diesem Fall ist es die Startseite der Anwendung.<br>
<code>def hello_world()</code>: Eine Funktion, die aufgerufen wird, wenn die definierte Route aufgerufen wird. Sie gibt den Text “Hello, World!” zurück. <br>
<code>app.run(debug=True)</code>: Startet den Flask-Entwicklungsserver. Der debug-Modus ermöglicht eine automatische Neustartfunktion bei Codeänderungen und eine interaktive Debugging-Konsole bei Fehlern.<br>

Nach Starten des Codes, öffnen Sie einen Webbrowser und gehen Sie zu http://127.0.0.1:5000/, um die Ausgabe “Hello, World!” zu sehen.


### Routen und Methoden

Standardmäßig unterstützt eine Route nur GET-Anfragen. Man kann jedoch auch POST, PUT und DELETE hinzufügen, um RESTful-Endpunkte zu erstellen.

```python
 from flask import Flask

app = Flask(__name__)

@app.route("/greet/<name>", methods=["GET"])
def greet(name):
    return f"Hallo {name}!"

if __name__ == "__main__":
    app.run(debug=True)
```

Bei Aufruf der Seite mit: <code>http://127.0.0.1:5000/greet/deinname</code> wird man freundlich begrüßt.

**Erklärung**:
* <code>@app.route('/greet/<name>')</code>: Dynamische Route mit einem URL-Parameter name.

### Formulare
#### Ein einfaches Formular

Um einfache Webseiten anzuzeigen braucht es kein Python. Dafür reicht einfaches HTML. 

Aber wer will das schon. Hier ein Beispiel für ein einfaches Formular.

app.py:

```python
from flask import Flask, render_template, redirect, url_for, request

app = Flask(__name__)

@app.route("/", methods=["GET"])
def index():
    return redirect(url_for("form"))

@app.route("/form", methods=["GET", "POST"])  # Decorator, dass die Funktion "form()" über http://127.0.0.1:5000/form erreichbar ist
def form():
    if request.method == "POST":  # Diese "Anforderung" wird erst nach Abschicken des Formulars aufgerufen (Schritt 2)
        name = request.form["name"]
        return f"Hallo {name}!"

    return render_template("form.html")  # Das wird direkt bei Aufruf der Route gerendert (Schritt 1)


if __name__ == "__main__":
    app.run(debug=True)
```

form.html:

```python

<!DOCTYPE html>
<html lang="de">
<head>
    <meta charset="UTF-8">
    <title>Formular</title>
</head>
<body>
    <h1>Formular</h1>
    <form action="/form" method="POST">
        <p>
            <label for="name">Name:</label>  <!-- Label "Name" wird erstellt -->
         <input type="text" name="name" id="name">  <!-- Eingabefeld "Name" wird erstellt -->
        </p>
        <input type="submit" value="Abschicken">  <!-- Knopf "Abschicken" wird erstellt -->
    </form>
</body>
</html>
```

#### Auswahlfelder mit Dropdown und Radio-Buttons

app.py:

```python
from flask import Flask, render_template, redirect, url_for, request

app = Flask(__name__)

@app.route("/", methods=["GET"])
def index():
    return redirect(url_for("form_options"))

@app.route("/form_options", methods=["GET", "POST"])
def form_options():
    # Variablen, die übermittelt werden, müssen existieren!
    selected_dropdown = None
    selected_radio = None

    if request.method == "POST":
        selected_dropdown = request.form["dropdown"]
        selected_radio = request.form["radio"]


    return render_template("form_options.html",
                           selected_dropdown=selected_dropdown,
                           selected_radio=selected_radio)


if __name__ == "__main__":
    app.run(debug=True)

```

#### Datei-Upload-Formular:
Immer wieder wichtig, gerade bei Datascience.
Das Datei-Upload-Formular: <br>

Bevor der Code funktioniert, muss innerhalb des Projekt-Verzeichnisses ein Ordner mit den Namen "uploads" erstellt werden.

<img src="img/uploads.png"> <br>

app.py:


```python
from flask import Flask, render_template, redirect, url_for, request
import os

app = Flask(__name__)
app.config["UPLOAD_FOLDER"] = "uploads/"  # Einstellung für Upload-Ordner wird gesetzt

# Weiterleitung zum aktuellen Projekt (Arbeitserleichterung für uns)
@app.route("/", methods=["GET"])
def index():
    return redirect(url_for("form_upload"))


@app.route("/upload", methods=["GET", "POST"])
def form_upload():

    if request.method == "POST":
        file = request.files["file"]  # Datei wird aus Formular übernommenn ...
        file.save(os.path.join(app.config["UPLOAD_FOLDER"], file.filename))  # ... und in den Ordner "uploads" gespeichert (s. O.)

        return f"Datei {file.filename} wurde erfolgreich hochgeladen!"  # Rückmeldung -> yay!

    return render_template("form_upload.html")


if __name__ == "__main__":
    app.run(debug=True)
```

#### Gemischtes Formular mir Weiterleitung auf einer Ergebnisseite

Um die Daten auf eine separate Seite weiterzuleiten, können wir die Daten mithilfe von Flask <code>redirect</code> und <code>url_for</code> an eine neue Route weitergeben.<br>

Hier ein Beispiel: <br>
app.py:

```python
from flask import Flask, render_template, redirect, url_for, request
import os

app = Flask(__name__)
app.config['UPLOAD_FOLDER'] = 'uploads/'


@app.route("/", methods=["GET"])
def index():
    return redirect(url_for("form_mixed"))


@app.route("/mixed", methods=["GET", "POST"])
def form_mixed():
    if request.method == "POST":
        # Daten aus dem Formular lesen:
        name = request.form["name"]
        age = request.form["age"]
        gender = request.form["gender"]
        skills = request.form.getlist("skills")
        resume = request.files["resume"]

        # Datei speichern:
        resume.save(os.path.join(app.config["UPLOAD_FOLDER"], resume.filename))

        return render_template("result.html",
                               name=name,
                               age=age,
                               gender=gender,
                               skills=skills,
                               filename = resume.filename)

    return render_template("form_mixed.html")



if __name__ == "__main__":
    app.run(debug=True)
```

form_mixed.html:

```python
<!DOCTYPE html>
<html lang="de">
<head>
    <meta charset="UTF-8">
    <title>Gemischtes Formular</title>
</head>
<body>
    <form method="post" action="/mixed" enctype="multipart/form-data">
        <label for="name">Name:</label>
        <input type="text" id="name" name="name"><br><br>

        <label for="age">Alter:</label>
        <input type="number" id="age" name="age"><br><br>

        <label>Geschlecht:</label><br>
        <input type="radio" id="male" name="gender" value="male">
        <label for="male">Männlich</label><br>
        <input type="radio" id="female" name="gender" value="female">
        <label for="female">Weiblich</label><br><br>

        <label>Fähigkeiten:</label><br>
        <input type="checkbox" id="python" name="skills" value="Python">
        <label for="python">Python</label><br>
        <input type="checkbox" id="java" name="skills" value="Java">
        <label for="java">Java</label><br><br>

        <label for="resume">Lebenslauf hochladen:</label>
        <input type="file" id="resume" name="resume" required="True"><br><br>

        <button type="submit">Absenden</button>
    </form>
</body>
</html>
```

result.html:

```python
<!DOCTYPE html>
<html lang="de">
<head>
    <meta charset="UTF-8">
    <title>Ergebnisse</title>
</head>
<body>
    <h1>Ergebnisse:</h1>
    <p><strong>Name:</strong> {{ name }}</p>
    <p><strong>Alter:</strong> {{ age }}</p>
    <p><strong>Geschlecht:</strong>  {{ gender }}</p>
    <p><strong>Fähigkeiten:</strong> {{ skills | join(", ") }}</p>  <!-- | = Pipe -->
    <p><strong>Hochgeladene Datei:</strong> {{ filename }}</p>

</body>
</html>
```

**Erklärung**:

**Datenweitergabe mit redirect und url_for**: <br>
Die eingegebenen Daten werden in der Route **/mixed** gesammelt und über POST-Methode an die Route **/result** weitergegeben.
Dadurch werden die Daten im Body der Anfrage übermittelt und nicht in der URL sichtbar.<br>
Die Methode <code>redirect(url_for())</code> generiert eine URL mit den übergebenen Parametern.<br><br>

**Ergebnisseite /result**:<br>
Nutzt <code>request.args.get()</code>, um die übergebenen Parameter auszulesen.
Die Fähigkeiten werden als CSV (durch Komma getrennt) übergeben und auf der Ergebnissseite in eine Liste umgewandelt <code>(.split(','))</code>.<br><br>

**Datei-Upload**:<br>
Die hochgeladene Datei wird im Upload-Ordner gespeichert, und ihr Dateiname wird auf der Ergebnissseite angezeigt.<br><br>

**Sicherheit**:<br>
Der Datei-Upload sollte in einer echten Anwendung durch Sicherheitschecks ergänzt werden (z. B. Dateitypen prüfen).<br><br>

### Datei lesen und ändern

Hier ein Beispiel, in dem eine Datei gelesen wird und in einer Textbox angezeigt wird. Der Text kann geändert und gespeichert werden. <br>
app.py:

```python
from flask import Flask, render_template, redirect, url_for, request, flash
import os

app = Flask(__name__)
app.secret_key = "supersecretkey"  # Geheimer Schlüssel, sollte natürlich komplexer sein, ein Hash zum Beispiel

DATEIPFAD = "data.txt"


@app.route("/", methods=["GET", "POST"])
def index():
    if request.method == "POST":
        neuer_inhalt = request.form["zeile"] # zeile = Textfeld-Variable im Formular
        neuer_inhalt = neuer_inhalt

        with open(DATEIPFAD, "w", encoding="utf-8", newline="") as datei:
            datei.write(neuer_inhalt)

        flash("Datei gespeichert!")

        return redirect(url_for("index"))

    try:
        with open(DATEIPFAD, "r", encoding="utf-8") as datei:
            inhalt = datei.read().strip()

    except FileNotFoundError:
        inhalt = ""

    return render_template("edit.html", zeile=inhalt)


if __name__ == "__main__":
    app.run(debug=True)
```

```python
<!DOCTYPE html>
<html lang="de">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Dateizeile bearbeiten</title>
    <style>
        body { font-family: Arial, sans-serif; margin: 40px; }
        form { max-width: 400px; }
        textarea { width: 100%; height: 100px; }
        button { padding: 10px 20px; }
    </style>
</head>
<body>
    <h1>Zeile aus Datei bearbeiten</h1>

    {% with messages = get_flashed_messages() %}
    {% if messages %}
    <ul style="color: green;">
        {% for message in messages %}
            <li>{{ message }}</li>
        {% endfor %}
    </ul>
    {% endif %}
    {% endwith %}

    <form method="POST">
        <textarea name="zeile">{{ zeile }}</textarea><br>
        <button type="submit">Speichern</button>
    </form>
</body>
</html>
```

### Schleifen in Jinja2
Jinja2-Schleifen sind hilfreich, um Listen oder Dictionaries zu durchlaufen und HTML dynamisch zu generieren.

app.py:

```python
from flask import Flask, render_template, redirect, url_for, request, flash

app = Flask(__name__)

@app.route('/')
def index():
    return redirect(url_for('products'))


@app.route("/products", methods=["GET", "POST"])
def products():
    product_list = [
        {"name": "Laptop", "price": 1000},
        {"name": "Smartphone", "price": 800},
        {"name": "Tablet", "price": 600}
    ]

    return render_template("products.html", products=product_list)


if __name__ == "__main__":
    app.run(debug=True)
```

products.html:

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Productos</title>
</head>
<body>
  <h1>Productos</h1>
<ul>
  {% for product in products %}
  <li>
    <strong>{{ product.name }}</strong>: {{ product.price }} €
  </li>
  {% else %}
    <li>No hay productos</li>
  {% endfor %}
</ul>
</body>
</html>
```

**Erklärung**:

* <code>{% for product in products %}</code>: Schleife durch die Liste der Produkte.
* <code>{% else %}</code>: Wird angezeigt, wenn die Liste leer ist.
* <code>{{ product.name }}</code>: Zugriff auf die Eigenschaften eines Dictionaries.

```python
from flask import Flask, render_template, redirect, url_for, request, flash

app = Flask(__name__)

@app.route('/')
def index():
    return render_template("home.html")

@app.route("/contact", methods=["GET", "POST"])
def contact():

    if request.method == "POST":
        # Daten aus dem Formular lesen:
        email = request.form["email"]
        subject = request.form["subject"]
        content = request.form["content"]

        print(email, subject, content)


        return render_template("contact.html",
                              email = email,
                               subject  = subject,
                               content = content
                               )

    return render_template("contact.html")

if __name__ == "__main__":
    app.run(debug=True)
```

### Bedingte Anweisung
Jinja2 unterstützt einfache Bedingungen wie if und else.

app.py:

```python
from flask import Flask, render_template, request, flash, redirect, url_for

app = Flask(__name__)

@app.route('/user')
def user():
    user_info = {'name': 'Max', 'logged_in': False}
    return render_template('user.html', user=user_info)

if __name__ == '__main__':
    app.run(debug=True)
```

user.html:

```python
<!DOCTYPE html>
<html lang="de">
<head>
    <meta charset="UTF-8">
    <title>Benutzerprofil</title>
</head>
<body>
    {% if user.logged_in %}
        <h1>Willkommen, {{ user.name }}!</h1>
    {% else %}
        <h1>Bitte melden Sie sich an.</h1>
    {% endif %}
</body>
</html>
```

**Erklärung**:

* <code>{% if user.logged_in %}</code>: Zeigt den Inhalt abhängig vom Wert logged_in an.
* <code>{% else %}</code>: Fallback, wenn die Bedingung nicht erfüllt ist.

### IP und Port ändern

Der Port kann beim Starten der App geändert werden:
<code>app.run(port=app_port, debug=True)</code> mit <code>app.run(host=app_IP, port=app_port, debug=True)</code> kann auch direkt der Port geändert werden.
mit der IP **0.0.0.0** lauscht der Server auf alle IPs auf die der Rechner lauscht.

```python
if __name__ == "__main__":
    app.run(host="0.0.0.0", port=80, debug=True)
```

```python
from flask import Flask

app = Flask(__name__)
app_port = 8080  # Port festlegen
app_ip = "0.0.0.0"  # IP festlegen

@app.route('/')
def home():
    return "Hallo! Diese Flask-Anwendung läuft auf einem anderen Port."

if __name__ == '__main__':
    app.run(host=app_ip, port=app_port, debug=True)  # Port auf 8080 ändern
```
